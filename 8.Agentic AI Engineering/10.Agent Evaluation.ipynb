# Chapter 10: Agent Evaluation

In the previous chapters we learned:

```text
Tool Calling

Reasoning

Planning

Memory

Architectures

LangGraph

Frameworks

Multi-Agent Systems
```

Now comes one of the most overlooked questions in Agentic AI:

> **How do we know whether an Agent is actually good?**

Unlike traditional software, an Agent does not always produce deterministic outputs.

The same input may produce:

```text
Different Plans

Different Tool Calls

Different Reasoning

Different Answers
```

This makes evaluation significantly more difficult.

---

# Why Agent Evaluation Is Hard

Traditional Software

```text
Input

↓

Output

↓

Compare
```

Simple.

---

Agent Systems

```text
Input

↓

Reasoning

↓

Planning

↓

Tool Calls

↓

Observations

↓

Final Answer
```

Multiple opportunities exist for failure.

---

## Interview Takeaway

Agent evaluation measures the entire execution process, not just the final response.

---

# Common Interview Questions

---

## Question

Why is agent evaluation harder than chatbot evaluation?

### Answer

```text
Because agents perform reasoning, planning, tool calling, and multi-step execution, creating multiple points of failure.
```

---

## Question

What is trajectory evaluation?

### Answer

```text
Evaluating the complete execution path rather than only the final response.
```

---

## Question

Why use LLM-as-a-Judge?

### Answer

```text
To automate evaluation at scale when human review is too expensive.
```

---

## Question

Offline vs Online evaluation?

### Answer

```text
Offline uses benchmark datasets.

Online measures production behavior with real users.
```

---

## Question

What metrics matter most?

### Answer

```text
Task Success Rate, Tool Accuracy, Latency, Cost, Hallucination Rate, and Safety.
```

---

## Question

Why is observability important?

### Answer

```text
It allows engineers to trace failures and understand why an agent behaved the way it did.
```
