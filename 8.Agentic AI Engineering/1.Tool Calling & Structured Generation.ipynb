# Module 8: Agentic AI Engineering

# Chapter 1: Tool Calling & Structured Generation

Before learning agents, we need to understand one fundamental concept:

> **Agents are simply LLMs that can interact with external systems.**

A normal LLM can only generate text.

An Agent can:

```text
Think

Decide

Call Tools

Observe Results

Continue Reasoning

Return Final Answer
```

The capability that enables this is **Tool Calling**.

---

# Why Do LLMs Need Tools?

A Large Language Model has only one capability:

```text
Predict Next Token
```

It **cannot**:

```text
Access Database

Call REST API

Check Weather

Execute SQL

Run Python

Read Files

Search Internet

Send Emails
```

Those are external actions.

Therefore, we give the LLM access to **Tools**.

---

## Interview Takeaway

A language model generates text.

An agent generates actions.

---

# What Is A Tool?

A Tool is simply:

```text
A Callable Function

Exposed To The LLM
```

Example:

```python
get_weather(city)

search_documents(query)

execute_sql(query)

send_email(to, subject, body)
```

The LLM never performs these actions itself.

Instead it requests:

```text
Please Call:

search_documents()

with

query = "Revenue report"
```

---

## Interview Takeaway

A tool is an external capability exposed to the model.

---

# LLM vs Agent

Traditional LLM

```text
User
   ↓

LLM
   ↓

Text Response
```

---

Agent

```text
User
    ↓

LLM
    ↓

Tool Call
    ↓

External System
    ↓

Observation
    ↓

LLM
    ↓

Final Answer
```

---

## Interview Takeaway

An agent extends an LLM by interacting with external systems.

---

# Tool Calling Lifecycle

Every tool call follows nearly the same lifecycle.

```text
User Request
      ↓

LLM Understands Task
      ↓

Choose Tool
      ↓

Generate Arguments
      ↓

Execute Tool
      ↓

Receive Observation
      ↓

Reason Again
      ↓

Final Response
```

---

Example

User

```text
What's the weather in Delhi?
```

↓

LLM

```text
Need Weather Tool
```

↓

Tool

```text
get_weather("Delhi")
```

↓

Tool Response

```text
36°C
Sunny
```

↓

LLM

```text
Today's weather in Delhi is...
```

---

## Interview Takeaway

Tool calling is an iterative reasoning process.

---

# How Does The Model Know Which Tools Exist?

When making an API request we provide:

```text
System Prompt

+

Available Tools
```

The model receives something similar to:

```json
[
  {
    "name":"get_weather",
    "description":"Get current weather",
    "parameters":{
      "city":"string"
    }
  }
]
```

The LLM reads this specification before generating.

---

## Interview Takeaway

The model does not discover tools—it receives their definitions.

---

# Tool Schema

Every tool requires a schema.

A schema defines:

```text
Tool Name

Description

Input Parameters

Parameter Types

Required Fields
```

Example

```json
{
  "name":"book_flight",
  "parameters":{
      "origin":"string",
      "destination":"string",
      "date":"string"
  }
}
```

---

## Why Schemas Matter

Without schemas

```text
LLM May Guess Arguments
```

With schemas

```text
Arguments Become Structured
```

---

## Interview Takeaway

Schemas allow reliable tool invocation.

---

# Function Calling

Tool Calling and Function Calling are closely related.

Function Calling means:

```text
Model

↓

Returns

Function Name

+

Arguments

Instead Of Free Text
```

Example

Instead of

```text
I'll check the weather...
```

Model returns

```json
{
   "function":"get_weather",
   "arguments":{
      "city":"Delhi"
   }
}
```

Application executes it.

---

## Interview Takeaway

Function calling produces executable instructions rather than natural language.

---

# Tool Calling vs Function Calling

Conceptually

```text
Tool Calling

Broader Concept
```

↓

Includes

```text
Functions

APIs

Databases

Search

Python

Custom Services
```

Function Calling

```text
Specific Mechanism

Used By API
```

---

## Interview Takeaway

Function calling is one implementation of tool calling.

---

# Structured Outputs

One of the hottest interview topics.

---

Problem

LLMs naturally generate

```text
Free Text
```

Applications require

```text
Structured Data
```

Example

Instead of

```text
The customer's sentiment appears positive.
```

Need

```json
{
   "sentiment":"positive"
}
```

---

## Interview Takeaway

Structured outputs make LLM responses machine-readable.

---

# Why Structured Outputs Matter

Production systems need:

```text
Consistency

Validation

Automation
```

Free text breaks pipelines.

Structured outputs enable:

```text
Reliable Automation
```

---

# JSON Mode

Most LLM providers support:

```text
JSON Output
```

Instead of arbitrary text.

Example

```json
{
  "priority":"high",
  "department":"finance"
}
```

---

Benefits

```text
Easy Parsing

Validation

Integration
```

---

## Interview Takeaway

JSON mode constrains model outputs into valid JSON.

---

# Schema Validation

Even JSON can be incorrect.

Need validation.

Example

Expected

```json
age : integer
```

Model returns

```json
age : "twenty"
```

Validation catches errors.

---

## Interview Takeaway

Validation ensures outputs satisfy expected data types.

---

# Pydantic

One of the most popular libraries for structured generation.

Example

```python
class Customer(BaseModel):
    name: str
    age: int
```

LLM output is automatically validated.

---

Benefits

```text
Type Safety

Validation

Serialization
```

---

## Interview Takeaway

Pydantic simplifies structured output validation.

---

# Tool Selection

Suppose the agent has:

```text
Weather Tool

Calculator

Database

Search Engine
```

Question:

```text
How Does It Choose?
```

The LLM reasons over:

```text
Tool Description

User Intent

Required Information
```

and selects the most appropriate tool.

---

## Interview Takeaway

Tool descriptions strongly influence tool selection.

---

# Tool Routing

Sometimes multiple tools exist.

Example

```text
Search Tool

SQL Tool

CRM Tool
```

The model must determine:

```text
Which Tool

Should Execute
```

This process is called routing.

---

## Interview Takeaway

Routing determines which tool receives the request.

---

# Parallel Tool Calling

Modern models can execute:

```text
Multiple Independent Tools
```

Example

```text
Weather Delhi

Weather Mumbai

Weather Chennai
```

Instead of sequential execution.

---

Benefits

```text
Lower Latency

Higher Throughput
```

---

## Interview Takeaway

Parallel tool calls reduce execution time.

---

# Sequential Tool Calling

Sometimes tools depend on previous results.

Example

```text
Search Customer
      ↓

Retrieve Customer ID
      ↓

Fetch Orders
      ↓

Generate Summary
```

Cannot execute simultaneously.

---

## Interview Takeaway

Dependent tools require sequential execution.

---

# Tool Chaining

Output of one tool becomes input to another.

Example

```text
OCR
     ↓

Text Extraction
     ↓

Summarization
     ↓

Translation
```

Multiple tools cooperate.

---

## Interview Takeaway

Tool chaining creates complex workflows.

---

# Error Handling

Tool execution may fail.

Reasons

```text
Timeout

Network Error

Permission Denied

Invalid Arguments

Service Unavailable
```

Agent should:

```text
Retry

Fallback

Ask User

Choose Another Tool
```

---

## Interview Takeaway

Robust agents gracefully recover from tool failures.

---

# Retry Strategy

Example

```text
API Failed
```

↓

Retry

↓

Success

---

Common strategies

```text
Fixed Retry

Exponential Backoff

Circuit Breaker
```

---

## Interview Takeaway

Retries improve reliability without overwhelming services.

---

# Tool Permissions

Not every tool should always be accessible.

Example

```text
Read Database

Allowed
```

↓

```text
Delete Database

Restricted
```

Permission systems prevent misuse.

---

## Interview Takeaway

Agents should follow least-privilege access principles.

---

# Dynamic Tool Discovery

Modern enterprise agents may have:

```text
Hundreds Of Tools
```

Instead of exposing every tool,

retrieve only:

```text
Relevant Tools
```

for the current task.

---

Benefits

```text
Lower Context Usage

Better Accuracy
```

---

## Interview Takeaway

Dynamic tool discovery improves scalability.

---

# Common Interview Questions

---

## Question

Why can't LLMs directly access APIs?

### Answer

```text
Because LLMs only generate tokens.

Applications execute external actions.
```

---

## Question

What is tool calling?

### Answer

```text
Allowing an LLM to invoke external functions or services during reasoning.
```

---

## Question

Function calling vs tool calling?

### Answer

```text
Function calling is a mechanism.

Tool calling is the broader concept.
```

---

## Question

Why are structured outputs important?

### Answer

```text
They enable reliable machine-readable integration.
```

---

## Question

Why use Pydantic?

### Answer

```text
To validate and enforce structured LLM outputs.
```

---

## Question

Sequential vs Parallel Tool Calling?

### Answer

```text
Sequential calls depend on previous outputs.

Parallel calls are independent and execute simultaneously.
```

---

## Question

What is tool routing?

### Answer

```text
Selecting the appropriate tool based on the user's request.
```

---

## Question

How do production agents handle tool failures?

### Answer

```text
Using retries, fallbacks, validation, and user clarification when needed.
```
