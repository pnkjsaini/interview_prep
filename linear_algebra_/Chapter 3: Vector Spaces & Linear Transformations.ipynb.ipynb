# Vector Space

## Definition

A vector space is a collection of vectors that can be combined through addition and scaling.

---

## Layman Explanation

Think of a vector space as:

```text
The World In Which Vectors Live
```

If vectors are arrows:

```text
Vector Space
```

defines all possible places those arrows can point.

---

## Example

2D Space

```text
[1,0]

[0,1]

[3,7]

[-2,4]
```

All belong to the same vector space.

---

## Why It Matters

Most machine learning models operate in:

```text
High Dimensional Vector Spaces
```

---

## Interview Takeaway

A vector space contains all valid vectors that can be created through addition and scaling.

---

# Span

## Definition

Span is the set of all vectors that can be created from a collection of vectors.

---

## Layman Explanation

Span answers:

> "What directions can I reach using these vectors?"

---

## Example

Suppose we have:

```text
v₁ = [1,0]

v₂ = [0,1]
```

By scaling and combining them we can reach:

```text
[2,3]

[-5,8]

[100,20]
```

and infinitely many others.

---

These reachable vectors form the:

```text
Span
```

---

## Geometric Interpretation

Span describes the space covered by vectors.

---

One Vector

```text
Line
```

---

Two Independent Vectors

```text
Plane
```

---

Three Independent Vectors

```text
3D Space
```

---

## Why It Matters

PCA and SVD attempt to find the most useful span.

---

## Limitation

Span says nothing about redundancy.

---

## Interview Takeaway

Span is the collection of all directions that can be formed using given vectors.

---

# Basis

## Definition

A basis is the smallest set of vectors needed to generate an entire vector space.

---

## Layman Explanation

Basis answers:

> "What is the minimum set of directions needed to describe everything?"

---

## Example

For 2D Space

```text
[1,0]

[0,1]
```

form a basis.

---

Every point in 2D can be created from them.

---

## Why Not Use More Vectors?

Suppose:

```text
[1,0]

[0,1]

[2,0]
```

The third vector adds:

```text
No New Direction
```

---

It is redundant.

---

## Why It Matters

PCA attempts to discover a new basis that captures maximum information.

---

## Limitation

Many different bases can exist for the same space.

---

## Interview Takeaway

A basis is the minimum set of independent directions needed to represent a space.

---

# Basis vs Span

## Relationship

Basis:

```text
Minimum Directions
```

---

Span:

```text
Everything Those Directions Can Generate
```

---

## Example

Basis

```text
[1,0]

[0,1]
```

---

Span

```text
Entire 2D Plane
```

---

## Interview Takeaway

Basis creates the span.

Span is generated by the basis.

---

# Linear Independence

## Definition

Vectors are linearly independent if none can be created from the others.

---

## Layman Explanation

Each vector contributes:

```text
New Information
```

---

## Example

Independent Vectors

```text
[1,0]

[0,1]
```

---

Neither can be produced from the other.

---

## Why It Matters

Independent vectors increase information.

---

## Applications

- PCA
- SVD
- Regression
- Feature Engineering

---

## Limitation

Independence becomes harder to visualize in high dimensions.

---

## Interview Takeaway

Linearly independent vectors contribute unique information.

---

# Linear Dependence

## Definition

Vectors are linearly dependent when at least one vector can be expressed using others.

---

## Layman Explanation

Dependent vectors contain:

```text
Redundant Information
```

---

## Example

```text
[1,2]

[2,4]
```

Second vector:

```text
2 × First Vector
```

---

No new information exists.

---

## Why It Matters

Dependent features create:

```text
Multicollinearity
```

in machine learning.

---

## Interview Takeaway

Linearly dependent vectors contain redundant information.

---

# Why Linear Independence Matters In ML

Suppose a housing dataset contains:

```text
Area In Square Feet

Area In Square Meters
```

Both features represent the same information.

---

Result:

```text
Redundant Features

Rank Deficiency

Unstable Regression Coefficients
```

---

PCA and SVD remove such redundancy.

---

## Interview Takeaway

Linear independence prevents duplicate information.

---

# Column Space

## Definition

Column Space is the set of all vectors that can be formed from the matrix columns.

---

## Layman Explanation

Column Space answers:

> "What outputs can this matrix produce?"

---

## Example

Matrix

```text
A
```

contains several column vectors.

All combinations of those columns create:

```text
Column Space
```

---

## Why It Matters

Column space represents:

```text
Information Preserved
```

by the matrix.

---

## Interview Takeaway

Column space contains all reachable outputs of a matrix.

---

# Row Space

## Definition

Row Space is the set of all vectors that can be formed from matrix rows.

---

## Layman Explanation

Row space represents information encoded across rows.

---

## Why It Matters

Used in:

- Rank Analysis
- Matrix Factorization
- SVD

---

## Interview Takeaway

Row space describes information contained in matrix rows.

---

# Null Space

## Definition

Null Space contains all vectors that become zero after a matrix transformation.

---

## Formula

```text
Ax = 0
```

---

## Layman Explanation

Null space answers:

> "What information disappears after transformation?"

---

## Example

Suppose a transformation removes:

```text
Height Information
```

while preserving:

```text
Weight Information
```

---

The removed direction belongs to:

```text
Null Space
```

---

## Why It Matters

Null space represents:

```text
Information Lost
```

during transformation.

---

## Applications

- PCA
- SVD
- Dimensionality Reduction

---

## Interview Takeaway

Null space contains information discarded by a transformation.

---

# Column Space vs Null Space

## Column Space

Represents:

```text
Information Preserved
```

---

## Null Space

Represents:

```text
Information Lost
```

---

## Why PCA Cares

PCA keeps directions inside:

```text
Important Column Space
```

---

and removes directions that contribute little information.

---

## Interview Takeaway

Column space preserves information.

Null space removes information.

---

# Linear Transformation

## Definition

A linear transformation maps one vector space into another while preserving linear relationships.

---

## Layman Explanation

A matrix is not just numbers.

A matrix is:

```text
A Transformation Machine
```

---

Input Vector

```text
x
```

enters.

---

Output Vector

```text
Ax
```

emerges.

---

## Example

Transformation Matrix

```text
[2 0]

[0 2]
```

---

Effect:

```text
Stretch Every Vector By 2
```

---

## Why It Matters

Neural networks are stacks of linear transformations.

---

## Interview Takeaway

Matrices transform vectors from one space into another.

---

# Types Of Linear Transformations

---

# Scaling

## Definition

Changes vector magnitude.

---

## Example

```text
[1,2]

↓

[2,4]
```

---

# Rotation

## Definition

Changes direction while preserving magnitude.

---

## Example

Vector rotates:

```text
90°
```

---

# Reflection

## Definition

Flips vectors across an axis.

---

## Example

Mirror image transformation.

---

# Projection

## Definition

Keeps information along selected directions.

---

## Example

PCA uses projections extensively.

---

# Shearing

## Definition

Slants shapes without changing area.

---

## Example

Rectangle becomes parallelogram.

---

# Geometric Interpretation Of Matrix Multiplication

Matrix multiplication can be viewed as:

```text
Input Vector
      ↓
Transformation Matrix
      ↓
Output Vector
```

---

The matrix decides:

```text
Stretch

Compress

Rotate

Project

Reflect
```

the input vector.
