# Chapter 5: Matrix Decomposition

In previous chapters we learned:

```text
Vectors

Matrices

Eigenvectors

Eigenvalues
```

A natural question arises:

> "Can a complex matrix be broken into simpler pieces that are easier to understand?"

The answer is:

```text
Yes
```

and the process is called:

```text
Matrix Decomposition
```

Matrix decomposition is one of the most powerful ideas in linear algebra.

Instead of studying a complicated matrix directly, we decompose it into simpler matrices that reveal:

```text
Structure

Patterns

Information

Redundancy

Noise
```

This chapter introduces the major matrix decomposition techniques and prepares us for:

```text
SVD
```

which is one of the most important concepts in Machine Learning.

---

# Why Matrix Decomposition Exists

Suppose we have:

```text
A Huge Matrix
```

containing millions of values.

---

Directly understanding:

```text
Relationships

Transformations

Patterns
```

can be difficult.

---

Matrix decomposition helps by breaking the matrix into:

```text
Simpler Components
```

that are easier to analyze.

---

## Why It Matters

Matrix decomposition helps with:

```text
Faster Computation

Numerical Stability

Dimensionality Reduction

Compression

Information Extraction
```

---

## Interview Takeaway

Matrix decomposition breaks a complex matrix into simpler matrices that reveal useful structure.

---

# What Is Matrix Decomposition?

## Definition

Matrix decomposition expresses a matrix as a product of simpler matrices.

---

## Layman Explanation

Imagine taking apart a machine.

Instead of studying:

```text
Entire Machine
```

you study:

```text
Individual Components
```

---

Matrix decomposition works similarly.

---

## Example

Instead of:

```text
A
```

we represent:

```text
A

=

Component 1

×

Component 2

×

Component 3
```

---

## Why It Matters

Many machine learning algorithms operate on decomposed forms rather than original matrices.

---

## Interview Takeaway

Matrix decomposition reveals hidden structure inside a matrix.

---

# Why We Decompose Matrices

Consider:

```text
Input Data
      ↓
Matrix
      ↓
Output
```

---

Questions we may ask:

```text
What Information Exists?

What Information Is Redundant?

What Information Is Important?

What Information Is Noise?
```

---

Decomposition helps answer these questions.

---

# LU Decomposition

## Definition

LU Decomposition factors a matrix into:

```text
Lower Triangular Matrix

×

Upper Triangular Matrix
```

---

## Formula

```text
A = LU
```

---

Where:

```text
L
```

Lower Triangular Matrix

---

```text
U
```

Upper Triangular Matrix

---

## Layman Explanation

Instead of solving a large matrix problem directly:

```text
Break It Into Smaller Steps
```

---

This makes computations easier.

---

## Example

A matrix:

```text
A
```

can be rewritten as:

```text
Lower Operations
```

followed by:

```text
Upper Operations
```

---

## Why It Matters

Widely used for solving:

```text
Systems Of Linear Equations
```

---

## Applications

- Scientific Computing
- Optimization
- Numerical Analysis

---

## Limitation

Not every matrix has a straightforward LU decomposition.

---

## Interview Takeaway

LU decomposition simplifies solving systems of equations.

---

# QR Decomposition

## Definition

QR Decomposition factors a matrix into:

```text
Orthogonal Matrix

×

Upper Triangular Matrix
```

---

## Formula

```text
A = QR
```

---

Where:

```text
Q
```

Orthogonal Matrix

---

```text
R
```

Upper Triangular Matrix

---

## Layman Explanation

QR separates:

```text
Direction
```

from

```text
Magnitude
```

---

Think of:

```text
Q
```

as describing:

```text
Pure Directions
```

---

and

```text
R
```

as describing:

```text
How Much Of Each Direction Exists
```

---

## Why It Matters

QR is useful when working with:

```text
Independent Directions
```

inside data.

---

## Applications

- Linear Regression
- Numerical Optimization
- PCA Computation

---

## Limitation

Less intuitive than SVD.

---

## Interview Takeaway

QR decomposition separates orthogonal directions from scaling information.

---

# Eigendecomposition

## Definition

Eigendecomposition expresses a matrix using its eigenvectors and eigenvalues.

---

## Formula

```text
A = PDP⁻¹
```

---

Where:

```text
P
```

Matrix Of Eigenvectors

---

```text
D
```

Diagonal Matrix Of Eigenvalues

---

## Layman Explanation

Instead of viewing a matrix as:

```text
Raw Numbers
```

we express it as:

```text
Important Directions

+

Their Importance
```

---

## Example

Chapter 4 introduced:

```text
Eigenvectors
```

as important directions.

---

and:

```text
Eigenvalues
```

as the strength of those directions.

---

Eigendecomposition combines them into a complete representation.

---

## Why It Matters

Foundation of:

```text
PCA
```

and many dimensionality reduction techniques.

---

## Applications

- PCA
- Covariance Analysis
- Physics
- Engineering

---

## Limitation

Only works reliably for certain square matrices.

---

## Interview Takeaway

Eigendecomposition represents a matrix using eigenvectors and eigenvalues.

---

# Why Eigendecomposition Is Not Enough

Eigendecomposition is powerful.

However it has limitations.

---

## Problem 1

Requires:

```text
Square Matrices
```

---

Real-world datasets often produce:

```text
Rectangular Matrices
```

---

## Problem 2

Some matrices do not have enough eigenvectors.

---

## Problem 3

Real-world data is noisy.

---

We need a decomposition method that is:

```text
More General

More Stable

More Useful
```

---

This leads to:

```text
Singular Value Decomposition (SVD)
```

---

# Introduction To Singular Value Decomposition (SVD)

## Why SVD Was Invented

SVD solves many limitations of eigendecomposition.

---

SVD works for:

```text
Square Matrices

Rectangular Matrices

Sparse Matrices

Dense Matrices
```

---

## Formula

```text
A = UΣVᵀ
```

---

## Components

### U Matrix

Represents important output directions.

---

### Σ Matrix

Represents importance of each direction.

---

### Vᵀ Matrix

Represents important input directions.

---

## Layman Explanation

SVD separates a matrix into:

```text
Directions

Importance

Transformation
```

---

## Why It Matters

SVD is the foundation of:

```text
PCA

Recommendation Systems

Embeddings

LSA

Model Compression

LoRA
```

---

## Interview Takeaway

SVD extracts the most important information from a matrix.

---
