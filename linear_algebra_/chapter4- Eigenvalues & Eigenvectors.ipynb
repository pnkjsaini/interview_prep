# Chapter 4: Eigenvalues & Eigenvectors

In previous chapters we learned:

```text
Vectors
```

represent data.

and

```text
Matrices
```

transform data.

A natural question arises:

> "When a matrix transforms vectors, are there some special directions that do not change direction?"

The answer leads to one of the most important concepts in Machine Learning:

```text
Eigenvectors
```

and

```text
Eigenvalues
```

These concepts power:

```text
PCA

SVD

PageRank

Recommendation Systems

Computer Vision

Deep Learning
```

---

# Why Eigenvectors Exist

Most vectors change both:

```text
Direction

Magnitude
```

after transformation.

---

Example

Imagine stretching a rubber sheet.

Most arrows drawn on it will:

```text
Rotate

Stretch

Compress
```

---

However, a few special arrows may only:

```text
Stretch

Compress
```

without changing direction.

Those special arrows are:

```text
Eigenvectors
```

---

# Eigenvector

## Definition

An eigenvector is a vector whose direction remains unchanged after a matrix transformation.

---

## Formula

```text
Av = λv
```

---

Where:

```text
A
```

Transformation Matrix

---

```text
v
```

Eigenvector

---

```text
λ
```

Eigenvalue

---

## Layman Explanation

Most vectors:

```text
Change Direction
```

after transformation.

---

Eigenvectors:

```text
Keep Same Direction
```

and only change length.

---

## Example

Suppose a matrix stretches everything horizontally.

A horizontal arrow remains horizontal.

Only its length changes.

That arrow is an eigenvector.

---

## Why It Matters

Eigenvectors reveal:

```text
Important Directions
```

inside a transformation.

---

## Interview Takeaway

An eigenvector is a direction that remains unchanged after transformation.

---

# Eigenvalue

## Definition

An eigenvalue measures how much an eigenvector is stretched or compressed.

---

## Formula

```text
Av = λv
```

---

## Layman Explanation

Eigenvector tells us:

```text
Direction
```

---

Eigenvalue tells us:

```text
Amount Of Stretching
```

---

## Example

Suppose:

```text
λ = 5
```

The eigenvector becomes:

```text
5 Times Longer
```

---

Suppose:

```text
λ = 0.5
```

The eigenvector becomes:

```text
Half As Long
```

---

Suppose:

```text
λ = -2
```

The vector:

```text
Flips Direction

And Doubles In Length
```

---

## Why It Matters

Eigenvalues measure the importance of directions.

---

## Interview Takeaway

Eigenvalues quantify how strongly a transformation acts along an eigenvector.

---

# Eigenvector & Eigenvalue Together

## Intuition

Imagine pushing a sheet of paper.

Most arrows:

```text
Rotate
```

and

```text
Stretch
```

---

A few arrows:

```text
Only Stretch
```

---

Those arrows are:

```text
Eigenvectors
```

---

The amount of stretching is:

```text
Eigenvalue
```

---

## Interview Takeaway

Eigenvectors define important directions.

Eigenvalues measure their importance.

---

# Geometric Interpretation

## Matrix Transformation

A matrix can:

```text
Rotate

Stretch

Compress

Project
```

vectors.

---

Eigenvectors identify:

```text
Stable Directions
```

inside the transformation.

---

Eigenvalues identify:

```text
Strength Of Transformation
```

along those directions.

---

## Why It Matters

Machine learning often tries to discover:

```text
Most Important Directions In Data
```

---

Eigenvectors help find them.

---

# Characteristic Equation

## Definition

The characteristic equation is used to calculate eigenvalues.

---

## Formula

```text
det(A - λI) = 0
```

---

Where:

```text
A
```

Matrix

---

```text
I
```

Identity Matrix

---

```text
λ
```

Eigenvalue

---

## Layman Explanation

The characteristic equation helps answer:

> "Which stretching factors make special directions exist?"

---

## Why It Matters

Eigenvalues are obtained by solving this equation.

---

## Interview Takeaway

The characteristic equation is used to find eigenvalues.

---

# Dominant Eigenvector

## Definition

The eigenvector associated with the largest eigenvalue.

---

## Layman Explanation

Among all important directions:

```text
One Direction Usually Dominates
```

---

The dominant eigenvector represents:

```text
Maximum Variation
```

or

```text
Maximum Influence
```

---

## Example

Suppose eigenvalues are:

```text
100

10

1
```

---

The eigenvector associated with:

```text
100
```

is dominant.

---

## Why It Matters

PCA focuses heavily on dominant eigenvectors.

---

## Interview Takeaway

The dominant eigenvector corresponds to the largest eigenvalue.

---

# Power Iteration

## Definition

Power Iteration is an algorithm used to find the dominant eigenvector.

---

## Layman Explanation

Repeatedly apply the matrix:

```text
A
```

to a random vector.

---

Eventually:

```text
All Other Directions Fade Away
```

and

```text
Dominant Direction Remains
```

---

## Example

Start:

```text
Random Vector
```

---

Apply:

```text
A

A

A

A
```

repeatedly.

---

Result:

```text
Dominant Eigenvector
```

appears.

---

## Why It Matters

Used when matrices are too large for exact decomposition.

---

## Applications

- PageRank
- Recommendation Systems
- PCA

---

## Interview Takeaway

Power Iteration finds the dominant eigenvector efficiently.

---

# Why Eigenvectors Matter In PCA

One of the most common interview questions.

---

Suppose data contains:

```text
100 Features
```

---

Some directions contain:

```text
Large Variation
```

---

Some directions contain:

```text
Mostly Noise
```

---

PCA asks:

> "Which directions contain the most information?"

---

Answer:

```text
Eigenvectors Of Covariance Matrix
```

---

The eigenvector with the largest eigenvalue becomes:

```text
Principal Component 1
```

---

The second largest becomes:

```text
Principal Component 2
```

---

and so on.

---

## Interview Takeaway

PCA uses eigenvectors to identify the most informative directions in data.

---

# Why Large Eigenvalues Matter

Large eigenvalues indicate:

```text
High Variation
```

---

High variation often means:

```text
More Information
```

---

Small eigenvalues indicate:

```text
Low Variation

Noise

Redundancy
```

---

## PCA Intuition

Keep:

```text
Large Eigenvalues
```

---

Discard:

```text
Small Eigenvalues
```

---

This reduces dimensionality while preserving information.

---

## Interview Takeaway

Large eigenvalues correspond to important directions in the data.

---
