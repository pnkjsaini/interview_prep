So far in this module we have studied:

```text
Linear Regression

Logistic Regression

Decision Trees

Random Forest

Gradient Boosting

SVM

KNN

Naive Bayes
```

All of these algorithms have one thing in common:

```text
Labels Exist
```

Examples:

```text
Spam / Not Spam

Fraud / Not Fraud

House Price

Customer Churn
```

---

The algorithm already knows:

```text
Correct Answers
```

during training.

---

But what happens when:

```text
Labels Do Not Exist?
```

---

This is where:

```text
Clustering
```

becomes important.

---

# Why Clustering Exists

Suppose a company has:

```text
10 Million Customers
```

---

Available Data:

```text
Age

Income

Purchase History

Location
```

---

But there are no labels such as:

```text
Premium Customer

Budget Customer

Occasional Customer
```

---

Question:

```text
Can We Automatically Discover Groups?
```

---

Clustering attempts to answer:

```text
Yes
```

---

## Interview Takeaway

Clustering discovers hidden groups in unlabeled data.

---

# What Is Clustering?

## Definition

Clustering groups similar observations together.

---

## Layman Explanation

Imagine entering a library where books have no labels.

---

You may naturally organize books into:

```text
Science

History

Technology

Mathematics
```

---

without knowing the categories beforehand.

---

That is clustering.

---

## Interview Takeaway

Clustering finds natural groupings within data.

---

# Supervised vs Unsupervised Learning

## Supervised Learning

Labels exist.

---

Example

```text
Fraud

Not Fraud
```

---

Goal:

```text
Predict Labels
```

---

## Unsupervised Learning

Labels do not exist.

---

Goal:

```text
Discover Structure
```

---

Examples:

```text
Customer Segments

Behavior Groups

Product Categories
```

---

## Interview Takeaway

Clustering is an unsupervised learning technique.

---

# Types Of Clustering

Most interview questions focus on:

```text
K-Means

Hierarchical Clustering

DBSCAN
```

---

# K-Means Clustering

The most important clustering algorithm.

---

# Why K-Means Exists

Suppose customer data naturally forms:

```text
Three Groups
```

---

We want to identify:

```text
Group 1

Group 2

Group 3
```

---

without manually creating rules.

---

## Interview Takeaway

K-Means partitions data into K clusters.

---

# What Does K Mean?

## Definition

K represents:

```text
Number Of Clusters
```

---

Example

```text
K = 3
```

---

Result:

```text
3 Clusters
```

---

Example

```text
K = 5
```

---

Result:

```text
5 Clusters
```

---

## Interview Takeaway

K specifies how many groups the algorithm should create.

---

# Centroid

## Definition

A centroid is the center of a cluster.

---

## Layman Explanation

Imagine a group of houses.

---

The centroid is:

```text
Average Location
```

of those houses.

---

Each cluster has:

```text
One Centroid
```

---

## Interview Takeaway

A centroid represents the center of a cluster.

---

# K-Means Algorithm

---

# Step 1

Choose:

```text
K Centroids
```

randomly.

---

# Step 2

Assign every observation to:

```text
Nearest Centroid
```

---

# Step 3

Recalculate centroids.

---

# Step 4

Repeat:

```text
Assignment

↓

Update Centroids
```

until convergence.

---

## Interview Takeaway

K-Means alternates between assigning points and updating centroids.

---

# K-Means Workflow

```text
Choose K
      ↓

Initialize Centroids
      ↓

Assign Points
      ↓

Update Centroids
      ↓

Repeat
      ↓

Final Clusters
```

---

# Example

Customer Data

```text
Income

Spending
```

---

Result

```text
Cluster 1
High Income
High Spending

Cluster 2
High Income
Low Spending

Cluster 3
Low Income
Low Spending
```

---

## Interview Takeaway

K-Means discovers customer segments automatically.

---

# How K-Means Measures Similarity

Typically uses:

```text
Euclidean Distance
```

---

Points closer to a centroid belong to that cluster.

---

## Interview Takeaway

K-Means relies on distance calculations.

---

# Choosing K

One of the most common interview questions.

---

# Problem

What value should K have?

---

Example

```text
2

3

5

10
```

---

Which one is correct?

---

# Elbow Method

The most common approach.

---

# Core Idea

As K increases:

```text
Error Decreases
```

---

Initially:

```text
Large Improvement
```

---

Eventually:

```text
Very Small Improvement
```

---

The bend in the curve is called:

```text
Elbow Point
```

---

Choose K near that point.

---

## Interview Takeaway

The Elbow Method helps determine the number of clusters.

---

# Limitations Of K-Means

## Need To Specify K

Must know cluster count beforehand.

---

## Sensitive To Initialization

Different starting centroids may produce different results.

---

## Sensitive To Outliers

Outliers can distort centroids.

---

## Assumes Roughly Spherical Clusters

May fail on irregular shapes.

---

## Interview Takeaway

K-Means is simple but has strong assumptions.

---

# Hierarchical Clustering

A completely different clustering approach.

---

# Core Idea

Instead of creating clusters immediately:

```text
Build A Hierarchy
```

of clusters.

---

## Interview Takeaway

Hierarchical clustering creates nested clusters.

---

# Agglomerative Clustering

Most common hierarchical method.

---

## Process

Start with:

```text
Every Point

As Its Own Cluster
```

---

Then repeatedly:

```text
Merge Closest Clusters
```

---

until only one cluster remains.

---

## Interview Takeaway

Agglomerative clustering follows a bottom-up approach.

---

# Divisive Clustering

Less common.

---

## Process

Start with:

```text
One Large Cluster
```

---

Then repeatedly:

```text
Split Clusters
```

---

until desired groups emerge.

---

## Interview Takeaway

Divisive clustering follows a top-down approach.

---

# Dendrogram

One of the most common interview topics.

---

## Definition

A dendrogram is a tree-like diagram showing cluster merges.

---

## Why It Matters

Helps decide:

```text
Number Of Clusters
```

---

without specifying K beforehand.

---

## Interview Takeaway

A dendrogram visualizes hierarchical cluster formation.

---

# Advantages Of Hierarchical Clustering

## No Need To Specify K Initially

Useful for exploration.

---

## Dendrogram Visualization

Provides interpretability.

---

## Flexible Cluster Shapes

More flexible than K-Means.

---

# Limitations

## Computationally Expensive

Poor scalability.

---

## Difficult On Very Large Datasets

Training becomes slow.

---

## Interview Takeaway

Hierarchical clustering is powerful but expensive.

---

# DBSCAN

One of the most important interview algorithms.

---

# Why DBSCAN Exists

Suppose clusters have:

```text
Irregular Shapes
```

---

K-Means struggles because it assumes:

```text
Circular Clusters
```

---

DBSCAN solves this problem.

---

## Interview Takeaway

DBSCAN discovers arbitrarily shaped clusters.

---

# What Is DBSCAN?

## Full Form

```text
Density Based Spatial Clustering
Of Applications With Noise
```

---

## Core Idea

Clusters are regions with:

```text
High Density
```

---

Noise points have:

```text
Low Density
```

---

## Interview Takeaway

DBSCAN groups dense regions together.

---

# Core Point

## Definition

A point with many nearby neighbors.

---

Represents:

```text
Dense Region
```

---

# Border Point

## Definition

Near a dense region but not dense itself.

---

# Noise Point

## Definition

An isolated observation.

---

Usually considered:

```text
Outlier
```

---

## Interview Takeaway

DBSCAN classifies observations into core, border, and noise points.

---

# Advantages Of DBSCAN

## No Need To Specify K

Automatically discovers cluster count.

---

## Handles Irregular Shapes

Major advantage.

---

## Detects Outliers

Built-in anomaly detection.

---

## Interview Takeaway

DBSCAN is excellent for irregular clusters and outlier detection.

---

# Limitations Of DBSCAN

## Sensitive To Density Parameters

Requires tuning.

---

## Struggles With Varying Density

Different cluster densities can be problematic.

---

## High-Dimensional Data

Performance decreases.

---

## Interview Takeaway

DBSCAN works best when cluster density is relatively consistent.

---

# K-Means vs Hierarchical vs DBSCAN

| Topic | K-Means | Hierarchical | DBSCAN |
|----------|----------|----------|----------|
| Need K | Yes | No | No |
| Outlier Handling | Poor | Moderate | Excellent |
| Irregular Shapes | Poor | Better | Excellent |
| Scalability | Excellent | Poor | Moderate |
| Dendrogram | No | Yes | No |

---

# When To Use K-Means

```text
Large Dataset

Fast Clustering Needed

Clusters Roughly Circular
```

---

# When To Use Hierarchical

```text
Need Dendrogram

Exploratory Analysis

Smaller Dataset
```

---

# When To Use DBSCAN

```text
Unknown Number Of Clusters

Outlier Detection

Irregular Cluster Shapes
```

---

# Real World Applications

```text
Customer Segmentation

Market Basket Analysis

Anomaly Detection

Fraud Detection

Image Segmentation

Recommendation Systems
```
