# Why Dimensionality Reduction Exists

Modern datasets often contain:

```text
Hundreds Of Features

Thousands Of Features

Millions Of Features
```

---

Examples:

```text
Customer Data

Sensor Data

Images

Embeddings

GenAI Vectors
```

---

Not all features are useful.

Many contain:

```text
Noise

Redundancy

Duplicate Information
```

---

Problems Created:

```text
Slower Training

Higher Memory Usage

Overfitting

Curse Of Dimensionality
```

---

## Interview Takeaway

Dimensionality reduction removes unnecessary information while preserving useful information.

---

# What Is Dimensionality Reduction?

## Definition

Dimensionality Reduction transforms data into fewer dimensions while retaining as much information as possible.

---

## Example

Original Data

```text
100 Features
```

---

Reduced Data

```text
10 Features
```

---

Goal:

```text
Keep Most Information

Lose Minimal Information
```

---

## Interview Takeaway

Dimensionality reduction compresses data without losing important structure.

---

# Curse Of Dimensionality Recap

One of the biggest reasons dimensionality reduction exists.

---

As dimensions increase:

```text
Data Becomes Sparse
```

---

Distances become:

```text
Less Meaningful
```

---

Algorithms such as:

```text
KNN

K-Means

DBSCAN
```

begin to struggle.

---

## Interview Takeaway

High-dimensional spaces make learning harder.

---

# Types Of Dimensionality Reduction

Most interview questions focus on:

```text
PCA

t-SNE

UMAP
```

---

# PCA (Principal Component Analysis)

## Recap

PCA creates:

```text
New Features
```

called:

```text
Principal Components
```

---

These components capture:

```text
Maximum Variance
```

in the data.

---

## Core Idea

Keep directions containing the most information.

---

Remove directions containing little information.

---

## Interview Takeaway

PCA preserves variance while reducing dimensions.

---

# How PCA Works

---

# Step 1

Center Data

---

# Step 2

Compute Covariance Matrix

---

# Step 3

Find Eigenvectors

---

# Step 4

Sort By Eigenvalues

---

# Step 5

Keep Important Components

---

Result:

```text
Reduced Dataset
```

---

## Interview Takeaway

PCA finds the most informative directions in data.

---

# Explained Variance

One of the most common PCA interview topics.

---

## Definition

Explained Variance measures how much information a component captures.

---

## Example

```text
PC1 → 70%

PC2 → 20%

PC3 → 5%
```

---

Keeping:

```text
PC1 + PC2
```

captures:

```text
90%
```

of information.

---

## Interview Takeaway

Explained variance helps determine how many components to keep.

---

# Advantages Of PCA

## Faster Training

Fewer features.

---

## Reduced Noise

Less irrelevant information.

---

## Lower Storage

Smaller datasets.

---

## Better Visualization

Useful for exploratory analysis.

---

# Limitations Of PCA

## Linear Method

Captures only linear relationships.

---

## Reduced Interpretability

Principal components are difficult to explain.

---

## Information Loss

Some information is discarded.

---

## Interview Takeaway

PCA is powerful but limited to linear transformations.

---

# Why t-SNE Exists

Suppose data contains:

```text
Complex Nonlinear Structure
```

---

PCA may fail to reveal hidden groups.

---

Researchers wanted:

```text
Better Visualization
```

of high-dimensional data.

---

This led to:

```text
t-SNE
```

---

# What Is t-SNE?

## Full Form

```text
t-Distributed Stochastic Neighbor Embedding
```

---

## Definition

A nonlinear dimensionality reduction technique primarily used for visualization.

---

## Core Idea

Preserve:

```text
Local Neighborhood Structure
```

---

If two points are close in high dimensions:

```text
Keep Them Close
```

in low dimensions.

---

## Interview Takeaway

t-SNE focuses on preserving local relationships.

---

# Why t-SNE Became Popular

Suppose we have:

```text
768-Dimensional Embeddings
```

---

Humans cannot visualize:

```text
768 Dimensions
```

---

t-SNE converts them into:

```text
2D

Or

3D
```

---

allowing visualization.

---

## Applications

```text
Embeddings

NLP

Image Features

Clustering Visualization
```

---

## Interview Takeaway

t-SNE is mainly used for visualization.

---

# Strength Of t-SNE

Excellent at revealing:

```text
Hidden Clusters

Local Groups

Subpopulations
```

---

## Example

Customer Embeddings

↓

t-SNE

↓

Distinct Customer Segments

---

## Interview Takeaway

t-SNE often reveals clusters invisible to PCA.

---

# Limitations Of t-SNE

One of the most frequently asked interview questions.

---

## Slow

Computationally expensive.

---

## Poor Scalability

Struggles with large datasets.

---

## Distorts Global Structure

Distances between clusters may not be meaningful.

---

## Not Ideal For Production Pipelines

Primarily a visualization tool.

---

## Interview Takeaway

t-SNE is excellent for visualization but not typically used for feature engineering.

---

# Why UMAP Exists

Researchers wanted:

```text
PCA Speed

+

t-SNE Visualization Quality
```

---

This led to:

```text
UMAP
```

---

# What Is UMAP?

## Full Form

```text
Uniform Manifold Approximation And Projection
```

---

## Definition

A nonlinear dimensionality reduction algorithm that preserves both local and global structure.

---

## Interview Takeaway

UMAP is a modern alternative to t-SNE.

---

# Core Idea Of UMAP

UMAP attempts to preserve:

```text
Local Structure

And

Global Structure
```

simultaneously.

---

This makes visualizations more meaningful.

---

## Interview Takeaway

UMAP balances local and global information.

---

# Why UMAP Is Popular

Compared to t-SNE:

```text
Faster

Scales Better

Preserves More Structure
```

---

Result:

```text
Increasing Industry Adoption
```

---

## Applications

```text
Embeddings

Vector Search

Clustering

Visualization

Recommendation Systems
```

---

## Interview Takeaway

UMAP is often preferred over t-SNE for large datasets.

---

# PCA vs t-SNE vs UMAP

| Topic | PCA | t-SNE | UMAP |
|----------|----------|----------|----------|
| Linear | Yes | No | No |
| Visualization | Good | Excellent | Excellent |
| Speed | Fastest | Slowest | Faster |
| Scalability | Excellent | Poor | Good |
| Global Structure | Good | Weak | Better |
| Feature Engineering | Yes | Rarely | Sometimes |

---

---

## Examples

```text
Preprocessing

Noise Reduction

Feature Compression
```

---

# When To Use t-SNE

Choose t-SNE when:

```text
Goal Is Visualization
```

---

Examples:

```text
Embeddings

Cluster Exploration

NLP Visualization
```

---

# When To Use UMAP

Choose UMAP when:

```text
Large Dataset

Visualization

Embedding Analysis

Clustering
```

---

## Interview Takeaway

UMAP is often the best visualization choice for modern ML workflows.

# Common Interview Questions

---

## Question

Why does dimensionality reduction exist?

### Answer

```text
To reduce complexity while preserving important information.
```

---

## Question

What problem does PCA solve?

### Answer

```text
Reducing dimensions while preserving variance.
```

---

## Question

What is explained variance?

### Answer

```text
The percentage of information captured by a principal component.
```

---

## Question

Why is PCA considered linear?

### Answer

```text
Because it uses linear combinations of features.
```

---

## Question

What is t-SNE mainly used for?

### Answer

```text
Visualization of high-dimensional data.
```

---

## Question

Why is t-SNE not commonly used in production?

### Answer

```text
Because it is computationally expensive and primarily designed for visualization.
```

---

## Question

What advantage does UMAP have over t-SNE?

### Answer

```text
Better scalability and preservation of global structure.
```

---

## Question

When should I use PCA instead of t-SNE?

### Answer

```text
When dimensionality reduction is needed for modeling rather than visualization.
```
