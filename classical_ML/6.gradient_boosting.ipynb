# Module 5: Classical Machine Learning

# Chapter 6: Gradient Boosting

In the previous chapters we learned:

```text
Decision Tree

Bagging

Random Forest
```

Random Forest improves performance by:

```text
Training Many Trees Independently
```

and combining their predictions.

This reduces:

```text
Variance
```

and improves stability.

However, Random Forest still has a limitation.

---

# Why Random Forest Is Not Enough

Suppose the true relationship is very complex.

Even after averaging many trees:

```text
Systematic Errors
```

may still remain.

---

In other words:

```text
Bias Still Exists
```

---

Random Forest says:

```text
Let's Build Many Trees
```

---

Gradient Boosting says:

```text
Let's Learn From Our Mistakes
```

---

This is a fundamentally different philosophy.

---

## Interview Takeaway

Random Forest reduces variance.

Gradient Boosting primarily reduces bias.

---

# Why Gradient Boosting Exists

Imagine a student taking a test.

---

Attempt 1

```text
60%
```

---

Teacher highlights mistakes.

---

Attempt 2

Student focuses only on mistakes.

---

Score:

```text
75%
```

---

Attempt 3

Student fixes remaining mistakes.

---

Score:

```text
90%
```

---

This process continues.

---

Gradient Boosting works exactly like this.

---

## Interview Takeaway

Gradient Boosting improves predictions by repeatedly correcting mistakes.

---

# Core Idea

Instead of training trees independently:

```text
Train Sequentially
```

---

Each new tree learns:

```text
Errors Made By Previous Trees
```

---

Result:

```text
Progressively Better Predictions
```

---

## Interview Takeaway

Every tree in Gradient Boosting focuses on remaining errors.

---

# What Is A Weak Learner?

## Definition

A weak learner performs only slightly better than random guessing.

---

## In Practice

Weak learners are usually:

```text
Small Decision Trees
```

---

Example:

```text
Depth = 1

Depth = 2

Depth = 3
```

---

These trees are intentionally simple.

---

## Why?

Simple trees:

```text
Learn Basic Patterns

Avoid Memorization
```

---

## Interview Takeaway

Gradient Boosting builds many small trees rather than one large tree.

---

# How Gradient Boosting Works

Suppose we want to predict:

```text
House Prices
```

---

# Step 1

Train First Tree.

---

Prediction:

```text
₹50 Lakh
```

---

Actual:

```text
₹60 Lakh
```

---

Error:

```text
₹10 Lakh
```

---

# Step 2

Train Second Tree

to predict:

```text
Residual Error
```

---

# Step 3

Add correction.

---

Updated Prediction:

```text
₹58 Lakh
```

---

# Step 4

Train another tree

to correct remaining error.

---

Prediction:

```text
₹59.5 Lakh
```

---

Continue until error becomes very small.

---

## Interview Takeaway

Gradient Boosting learns residuals instead of learning the target directly.

---

# Residual Learning

## Definition

Residuals are prediction errors.

---

## Formula

```text
Residual

=

Actual - Predicted
```

---

## Example

Actual Price

```text
₹100 Lakh
```

---

Prediction

```text
₹80 Lakh
```

---

Residual

```text
₹20 Lakh
```

---

Next tree learns:

```text
₹20 Lakh
```

instead of learning the full price again.

---

## Why It Matters

Learning residuals is easier than learning the entire problem repeatedly.

---

## Interview Takeaway

Each boosting stage predicts remaining errors.

---

# Sequential Learning

This is the biggest difference from Random Forest.

---

# Random Forest

```text
Tree 1

Tree 2

Tree 3

Tree 4
```

All trained independently.

---

# Gradient Boosting

```text
Tree 1
      ↓
Tree 2
      ↓
Tree 3
      ↓
Tree 4
```

Each tree depends on previous trees.

---

## Why It Matters

New trees become specialists at correcting mistakes.

---

## Interview Takeaway

Boosting is sequential learning.

---

# Additive Learning

Gradient Boosting builds the final model gradually.

---

Instead of:

```text
One Large Tree
```

---

Build:

```text
Tree 1

+

Tree 2

+

Tree 3

+

Tree 4
```

---

Final Prediction:

```text
Sum Of All Trees
```

---

## Interview Takeaway

Gradient Boosting creates a strong learner by adding weak learners together.

---

# Why It Is Called "Gradient" Boosting

One of the most common interview questions.

---

The model improves itself by moving in the direction that reduces loss.

---

This idea comes from:

```text
Gradient Descent
```

which we studied earlier.

---

Instead of updating:

```text
Coefficients
```

like Linear Regression,

Gradient Boosting updates:

```text
Trees
```

to reduce loss.

---

## Interview Takeaway

Gradient Boosting uses gradient-based optimization to reduce prediction error.

---

# Learning Rate

## Definition

Learning Rate controls how much each tree contributes.

---

## Layman Explanation

Suppose a tree predicts:

```text
Correction = 10
```

---

Learning Rate:

```text
0.1
```

---

Actual Update:

```text
1
```

---

instead of:

```text
10
```

---

## Why?

Large corrections can:

```text
Overfit
```

---

Small corrections create:

```text
Stable Learning
```

---

## Interview Takeaway

Learning Rate controls the contribution of each tree.

---

# Number Of Trees

## Definition

Total boosting iterations.

---

Few Trees

```text
Underfitting
```

---

Too Many Trees

```text
Overfitting
```

---

Goal:

```text
Find Balance
```

---

## Interview Takeaway

More trees increase complexity and learning capacity.

---

# Tree Depth

## Definition

Maximum depth of each weak learner.

---

Shallow Trees

```text
Simple Patterns
```

---

Deep Trees

```text
Complex Patterns
```

---

Most boosting algorithms use:

```text
Small Trees
```

---

## Interview Takeaway

Boosting usually relies on shallow trees.

---

# Loss Function

Gradient Boosting minimizes a loss function.

---

Examples:

```text
MSE
```

for regression.

---

```text
Log Loss
```

for classification.

---

The model repeatedly asks:

> "How can I reduce this loss?"

---

## Interview Takeaway

Boosting learns by minimizing loss.

---

# Why Gradient Boosting Is Powerful

Each tree specializes in:

```text
Remaining Errors
```

---

Over time:

```text
Small Improvements
```

accumulate into:

```text
Large Performance Gains
```

---

## Interview Takeaway

Gradient Boosting builds accuracy through many small corrections.

---

# Advantages

## High Predictive Accuracy

Often outperforms Random Forest.

---

## Learns Complex Relationships

Handles nonlinear patterns well.

---

## Works With Mixed Features

Numerical and categorical data.

---

## Strong Competition Performance

Widely used in Kaggle competitions.

---

# Limitations

## Slower Training

Sequential learning prevents parallel training.

---

## Sensitive To Hyperparameters

Requires careful tuning.

---

## Can Overfit

Especially with deep trees and large iterations.

---

## More Complex

Harder to interpret than a single tree.

---

# Random Forest vs Gradient Boosting

| Topic | Random Forest | Gradient Boosting |
|----------|----------|----------|
| Training | Parallel | Sequential |
| Goal | Reduce Variance | Reduce Bias |
| Tree Dependency | Independent | Dependent |
| Speed | Faster | Slower |
| Accuracy | Good | Often Better |
| Tuning Required | Lower | Higher |

---

# Gradient Boosting Workflow

```text
Dataset
      ↓

Tree 1
      ↓

Residuals
      ↓

Tree 2
      ↓

Residuals
      ↓

Tree 3
      ↓

Residuals
      ↓

Final Prediction
```
