# Module 5: Classical Machine Learning

# Chapter 9: K-Nearest Neighbors (KNN)

So far we have studied:

```text
Linear Regression

Logistic Regression

Decision Trees

Random Forest

Gradient Boosting

XGBoost

SVM
```

All these algorithms share one common idea:

```text
Learn A Model During Training
```

---

KNN is different.

KNN says:

```text
Don't Learn Anything During Training
```

Instead:

```text
Store The Data

And Learn During Prediction
```

This makes KNN one of the simplest yet most intuitive machine learning algorithms.

---

# Why KNN Exists

Imagine you move into a new city and want to estimate:

```text
House Price
```

---

You might ask:

```text
What Are Similar Houses Selling For?
```

---

Not:

```text
What Is The Mathematical Formula?
```

---

KNN follows exactly this logic.

---

For a new observation:

```text
Find Similar Observations
```

and use them to make a prediction.

---

## Interview Takeaway

KNN predicts using nearby observations rather than learning explicit rules.

---

# Core Idea

Suppose a new customer arrives.

---

Instead of building a model:

```text
Find Similar Customers
```

---

Then ask:

```text
What Happened To Them?
```

---

Use that information to make a prediction.

---

## Interview Takeaway

KNN assumes similar observations produce similar outcomes.

---

# What Does "Nearest Neighbor" Mean?

The most important question in KNN is:

```text
Who Is Similar?
```

---

To answer this:

```text
Distance Metrics
```

are used.

---

The closer two observations are:

```text
More Similar They Are Considered
```

---

## Interview Takeaway

KNN defines similarity using distance.

---

# K Value

## Definition

K represents the number of neighbors considered.

---

## Example

```text
K = 3
```

---

Use:

```text
3 Closest Observations
```

---

```text
K = 10
```

---

Use:

```text
10 Closest Observations
```

---

## Why It Matters

K controls model complexity.

---

## Interview Takeaway

K determines how many nearby observations influence prediction.

---

# Classification Using KNN

Suppose:

```text
K = 5
```

---

Nearest Neighbors

```text
Fraud

Fraud

Fraud

Not Fraud

Not Fraud
```

---

Majority Class

```text
Fraud
```

---

Prediction

```text
Fraud
```

---

## Interview Takeaway

Classification uses majority voting among neighbors.

---

# Regression Using KNN

KNN can also perform regression.

---

Example

Nearest House Prices

```text
₹40 Lakh

₹45 Lakh

₹50 Lakh
```

---

Prediction

```text
Average

=

₹45 Lakh
```

---

## Interview Takeaway

Regression uses the average value of neighboring observations.

---

# Euclidean Distance

The most common distance metric.

---

## Formula

```text
Straight Line Distance
```

between two points.

---

## Layman Explanation

Imagine measuring distance on a map using a ruler.

---

That direct distance is:

```text
Euclidean Distance
```

---

## Example

Points

```text
(0,0)

(3,4)
```

---

Distance

```text
5
```

---

## Why It Matters

Most KNN implementations use Euclidean Distance.

---

## Interview Takeaway

Euclidean Distance measures straight-line similarity.

---

# Manhattan Distance

## Definition

Distance traveled using horizontal and vertical movement.

---

## Layman Explanation

Imagine driving through city streets.

---

You cannot travel diagonally.

---

Instead:

```text
Move Horizontally

And Vertically
```

---

## Example

From:

```text
(0,0)
```

to:

```text
(3,4)
```

---

Distance

```text
7
```

---

## Why It Matters

More robust in some high-dimensional problems.

---

## Interview Takeaway

Manhattan Distance measures grid-based distance.

---

# Euclidean vs Manhattan

| Topic | Euclidean | Manhattan |
|----------|----------|----------|
| Path | Straight Line | Grid Path |
| Geometry | Diagonal Allowed | Horizontal/Vertical |
| Most Common | Yes | Less Common |

---

## Interview Takeaway

Euclidean is the default choice, but Manhattan may perform better in certain datasets.

---

# Choosing K

One of the most common interview topics.

---

# Small K

Example

```text
K = 1
```

---

Prediction depends on:

```text
Single Observation
```

---

Result:

```text
High Variance

Overfitting
```

---

# Large K

Example

```text
K = 100
```

---

Prediction depends on:

```text
Many Observations
```

---

Result:

```text
High Bias

Underfitting
```

---

# Goal

Find balance.

---

Typically done using:

```text
Cross Validation
```

---

## Interview Takeaway

Small K overfits, large K underfits.

---

# Feature Scaling

One of the most important KNN concepts.

---

# Problem

Suppose features are:

```text
Age

Income
```

---

Age Range

```text
20 - 60
```

---

Income Range

```text
₹1 Lakh - ₹1 Crore
```

---

Income dominates distance calculations.

---

Result:

```text
Age Becomes Irrelevant
```

---

# Solution

Scale features.

---

Common techniques:

```text
Standardization

Normalization
```

---

## Interview Takeaway

Feature scaling is critical for KNN.

---

# Training Phase

One of the easiest interview questions.

---

What happens during KNN training?

---

Answer:

```text
Almost Nothing
```

---

KNN simply stores:

```text
Training Data
```

---

No coefficients.

No trees.

No optimization.

---

## Interview Takeaway

KNN is called a lazy learner because it performs almost no training.

---

# Prediction Phase

Most computation happens here.

---

Process:

```text
New Observation
       ↓

Calculate Distances
       ↓

Find K Nearest Neighbors
       ↓

Vote Or Average
       ↓

Prediction
```

---

## Interview Takeaway

KNN shifts computation from training to prediction.

---

# Why KNN Can Become Slow

Suppose dataset contains:

```text
10 Million Records
```

---

Every prediction requires:

```text
Distance Calculation

Against All Records
```

---

Result:

```text
Slow Predictions
```

---

## Interview Takeaway

KNN scales poorly with large datasets.

---

# Curse Of Dimensionality

One of the most important KNN interview topics.

---

# Problem

As dimensions increase:

```text
Distance Becomes Less Meaningful
```

---

Example

In:

```text
2 Dimensions
```

neighbors are easy to identify.

---

In:

```text
1000 Dimensions
```

almost all points become similarly distant.

---

Result:

```text
Similarity Breaks Down
```

---

## Why It Matters

KNN performs poorly in very high-dimensional spaces.

---

## Interview Takeaway

KNN struggles when feature dimensions become very large.

---

# Advantages

## Extremely Simple

Easy to understand.

---

## No Training Required

Fast model building.

---

## Works For Classification And Regression

Flexible algorithm.

---

## Naturally Nonlinear

No assumptions about relationships.

---

# Limitations

## Slow Predictions

Must compute distances at prediction time.

---

## Sensitive To Scaling

Feature scaling is mandatory.

---

## Sensitive To Noise

Outliers can distort predictions.

---

## Suffers From Curse Of Dimensionality

Performance decreases in high dimensions.

---

# KNN vs Logistic Regression

| Topic | Logistic Regression | KNN |
|----------|----------|----------|
| Training | Learns Model | Stores Data |
| Decision Boundary | Linear | Nonlinear |
| Prediction Speed | Fast | Slow |
| Interpretability | Higher | Lower |
| Scaling Needed | Helpful | Critical |

---

# KNN vs SVM

| Topic | SVM | KNN |
|----------|----------|----------|
| Training | Expensive | Minimal |
| Prediction | Fast | Slow |
| High Dimensions | Strong | Weak |
| Decision Boundary | Learned | Neighbor Based |

---
