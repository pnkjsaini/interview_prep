# Module 5: Classical Machine Learning

# Chapter 8: Support Vector Machines (SVM)

So far we have studied:

```text
Logistic Regression

Decision Trees

Random Forest

Gradient Boosting

XGBoost
```

Most of these algorithms try to:

```text
Minimize Prediction Error
```

SVM approaches the problem differently.

Instead of asking:

> "How can I classify these points correctly?"

SVM asks:

> "What is the safest possible boundary that separates the classes?"

This idea makes SVM one of the most elegant algorithms in Machine Learning.

---

# Why SVM Exists

Suppose we want to classify:

```text
Spam

Not Spam
```

---

Many different lines can separate the classes.

Example:

```text
Line A

Line B

Line C
```

All may classify training data correctly.

---

Question:

```text
Which Line Should We Choose?
```

---

SVM says:

```text
Choose The Boundary

That Maximizes Separation
```

---

## Interview Takeaway

SVM chooses the safest decision boundary rather than just any valid boundary.

---

# Core Idea

Imagine two groups of points.

```text
Class A

Class B
```

---

Many boundaries can separate them.

---

SVM searches for:

```text
Maximum Margin Boundary
```

---

The larger the margin:

```text
Better Generalization
```

---

## Why?

Future data is less likely to fall near the boundary.

---

## Interview Takeaway

SVM maximizes the distance between classes.

---

# Hyperplane

## Definition

A hyperplane is the decision boundary used to separate classes.

---

## Layman Explanation

In:

```text
2 Dimensions
```

a hyperplane is:

```text
A Line
```

---

In:

```text
3 Dimensions
```

it becomes:

```text
A Plane
```

---

In:

```text
100 Dimensions
```

it becomes:

```text
A Hyperplane
```

---

## Example

Email Classification

```text
Spam

Not Spam
```

---

Hyperplane:

```text
Decision Boundary
```

between the two classes.

---

## Interview Takeaway

A hyperplane is the separating boundary learned by SVM.

---

# Margin

## Definition

Margin is the distance between the decision boundary and the nearest data points.

---

## Layman Explanation

Suppose two roads exist.

---

Road A

```text
Very Narrow
```

---

Road B

```text
Very Wide
```

---

Which is safer?

```text
Road B
```

---

Similarly:

```text
Large Margin

=

Safer Boundary
```

---

## Why It Matters

Larger margins generally improve:

```text
Generalization
```

---

## Interview Takeaway

Margin measures separation between classes.

---

# Support Vectors

## Definition

Support Vectors are the data points closest to the decision boundary.

---

## Layman Explanation

These points are:

```text
Most Difficult To Classify
```

---

and therefore determine:

```text
Where The Boundary Should Be
```

---

## Example

Suppose we remove:

```text
Far Away Points
```

---

The boundary changes very little.

---

Remove:

```text
Support Vectors
```

---

Boundary changes significantly.

---

## Why It Matters

Support vectors define the hyperplane.

---

## Interview Takeaway

Support vectors are the most influential observations in SVM.

---

# Maximum Margin Principle

This is the heart of SVM.

---

Among all possible boundaries:

```text
Choose The One

With Largest Margin
```

---

Example

```text
Boundary A

Margin = 1
```

---

```text
Boundary B

Margin = 5
```

---

SVM chooses:

```text
Boundary B
```

---

## Why?

Because larger margins usually lead to:

```text
Better Performance On Unseen Data
```

---

## Interview Takeaway

SVM maximizes margin to improve generalization.

---

# Linear SVM

## Definition

Linear SVM uses a straight hyperplane.

---

## Example

Suppose data can be separated using:

```text
One Straight Line
```

---

Then Linear SVM works well.

---

## Advantages

```text
Fast

Simple

Works Well In High Dimensions
```

---

## Limitation

Cannot handle complex nonlinear patterns.

---

## Interview Takeaway

Linear SVM works when classes are linearly separable.

---

# The Problem With Linear SVM

Consider:

```text
Circular Data
```

---

Example

```text
Class A

Inside Circle
```

---

```text
Class B

Outside Circle
```

---

No straight line can separate them.

---

Linear SVM fails.

---

## Interview Takeaway

Many real-world datasets are not linearly separable.

---

# Kernel Trick

One of the most frequently asked interview topics.

---

# Problem

Data is not linearly separable.

---

# Solution

Transform data into a higher-dimensional space.

---

In the new space:

```text
Linear Separation

May Become Possible
```

---

This transformation is performed using:

```text
Kernels
```

---

## Layman Explanation

Imagine looking at a shadow.

---

From one angle:

```text
Objects Overlap
```

---

Change perspective:

```text
Objects Separate Clearly
```

---

Kernel transformation creates a new perspective.

---

## Interview Takeaway

The Kernel Trick allows SVM to learn nonlinear boundaries.

---

# Why Is It Called A Trick?

Direct transformation may require:

```text
Thousands

Or

Millions

Of Dimensions
```

---

Very expensive.

---

Kernel functions compute similarity directly.

---

Without explicitly creating the higher-dimensional space.

---

## Interview Takeaway

The Kernel Trick avoids expensive feature transformations.

---

# Common Kernels

---

# Linear Kernel

## When To Use

```text
Linearly Separable Data

High Dimensional Data
```

---

## Example

Text Classification

Spam Detection

---

## Interview Takeaway

Linear Kernel is simple and efficient.

---

# Polynomial Kernel

## Idea

Create curved decision boundaries.

---

## Example

Interactions between features become important.

---

## Limitation

Can overfit easily.

---

## Interview Takeaway

Polynomial kernels model nonlinear relationships.

---

# RBF Kernel (Gaussian Kernel)

The most commonly used kernel.

---

## Idea

Create flexible nonlinear boundaries.

---

## Layman Explanation

Points closer together have stronger influence.

---

Points farther apart have weaker influence.

---

## Why It Matters

Can model very complex patterns.

---

## Interview Takeaway

RBF Kernel is the most popular kernel in practice.

---

# Soft Margin SVM

Real-world datasets contain:

```text
Noise

Outliers

Mislabels
```

---

Perfect separation may be impossible.

---

# Hard Margin

Requires:

```text
Perfect Separation
```

---

Rarely practical.

---

# Soft Margin

Allows:

```text
Small Classification Errors
```

---

to achieve:

```text
Better Generalization
```

---

## Interview Takeaway

Soft Margin SVM balances separation and flexibility.

---

# Regularization Parameter (C)

One of the most important SVM hyperparameters.

---

# Large C

Model prioritizes:

```text
Correct Classification
```

---

Result:

```text
Smaller Margin

Higher Overfitting Risk
```

---

# Small C

Model prioritizes:

```text
Larger Margin
```

---

Result:

```text
More Generalization
```

---

## Interview Takeaway

C controls the tradeoff between margin size and classification error.

---

# Advantages

## Effective In High Dimensions

Works well when:

```text
Features >> Samples
```

---

## Strong Theoretical Foundation

Maximum margin principle.

---

## Handles Nonlinear Data

Through kernels.

---

## Works Well On Small Datasets

Often better than deep learning.

---

# Limitations

## Training Can Be Slow

Large datasets become expensive.

---

## Kernel Selection Matters

Poor kernel choice reduces performance.

---

## Hard To Interpret

Compared to trees.

---

## Memory Intensive

Large datasets create scalability issues.

---

# SVM vs Logistic Regression

| Topic | Logistic Regression | SVM |
|----------|----------|----------|
| Output | Probability | Class Boundary |
| Objective | Predict Probability | Maximize Margin |
| Nonlinear Support | Limited | Strong Via Kernels |
| Interpretability | Higher | Lower |
| Small Dataset Performance | Good | Often Better |

---

# SVM vs Random Forest

| Topic | SVM | Random Forest |
|----------|----------|----------|
| Decision Boundary | Margin Based | Tree Based |
| Nonlinear Learning | Kernels | Trees |
| Interpretability | Lower | Higher |
| Large Dataset Scaling | Poorer | Better |
| High Dimensional Data | Excellent | Good |

---
