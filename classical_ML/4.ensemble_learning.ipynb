# Module 5: Classical Machine Learning

# Chapter 4: Ensemble Learning

In the previous chapter we learned:

```text
Decision Trees
```

Decision Trees are powerful because they can learn:

```text
Nonlinear Relationships

Complex Decision Rules
```

However they have a major weakness:

```text
High Variance
```

A small change in training data can produce a completely different tree.

This makes predictions unstable.

Ensemble Learning was created to solve this problem.

---

# Why Ensemble Learning Exists

Imagine asking:

```text
One Doctor
```

for a diagnosis.

---

Now imagine asking:

```text
100 Doctors
```

and taking the combined opinion.

---

Which prediction would you trust more?

Usually:

```text
100 Doctors
```

---

This is the core idea behind:

```text
Ensemble Learning
```

---

## Definition

Ensemble Learning combines multiple models to create a stronger model.

---

## Interview Takeaway

Ensemble Learning improves prediction quality by combining multiple models.

---

# Why Ensembles Work

A single model may make mistakes because of:

```text
Noise

Limited Data

Randomness
```

---

Different models often make:

```text
Different Mistakes
```

---

When predictions are combined:

```text
Individual Errors Tend To Cancel Out
```

---

Result:

```text
Better Accuracy

Better Stability

Better Generalization
```

---

## Interview Takeaway

Ensembles work because combining diverse models reduces overall error.

---

# The Bias-Variance Problem

Before understanding ensembles, we must understand:

```text
Bias

Variance
```

---

# Bias

## Definition

Bias measures how strongly a model oversimplifies reality.

---

## Layman Explanation

A high-bias model makes strong assumptions.

---

Example:

```text
Linear Regression
```

trying to model highly nonlinear data.

---

Result:

```text
Underfitting
```

---

## Interview Takeaway

High bias causes underfitting.

---

# Variance

## Definition

Variance measures sensitivity to training data.

---

## Layman Explanation

If small changes in data create large prediction changes:

```text
Variance Is High
```

---

Example:

```text
Deep Decision Tree
```

---

Result:

```text
Overfitting
```

---

## Interview Takeaway

High variance causes overfitting.

---

# Bias-Variance Tradeoff

Most models balance:

```text
Bias

And

Variance
```

---

High Bias

```text
Underfit
```

---

High Variance

```text
Overfit
```

---

Goal:

```text
Low Bias

Low Variance
```

---

## Interview Takeaway

Ensemble methods attempt to reduce bias, variance, or both.

---

# Three Major Ensemble Techniques

Most ensemble algorithms belong to:

```text
Bagging

Boosting

Stacking
```

---

# Bagging

## Full Form

```text
Bootstrap Aggregating
```

---

## Core Idea

Train multiple models independently.

---

Combine predictions.

---

## Layman Explanation

Instead of one Decision Tree:

```text
Train Many Trees
```

---

Each tree sees slightly different data.

---

Final prediction becomes:

```text
Majority Vote

Or

Average Prediction
```

---

## Why It Works

Different trees make different mistakes.

---

Averaging reduces variance.

---

## Interview Takeaway

Bagging reduces variance by combining independently trained models.

---

# Bootstrap Sampling

Bagging depends on:

```text
Bootstrap Sampling
```

---

## Definition

Sampling with replacement.

---

## Example

Original Data

```text
A B C D E
```

---

Bootstrap Sample

```text
A B B D E
```

---

Notice:

```text
B Appears Twice

C Is Missing
```

---

This is allowed.

---

## Why It Matters

Each model receives a different dataset.

---

This creates diversity.

---

## Interview Takeaway

Bootstrap sampling creates different training datasets for ensemble members.

---

# Aggregation

After training:

```text
Model 1

Model 2

Model 3

...
```

---

Predictions are combined.

---

## Classification

Use:

```text
Majority Voting
```

---

Example

```text
Fraud

Fraud

Not Fraud

Fraud
```

---

Final Prediction:

```text
Fraud
```

---

## Regression

Use:

```text
Average Prediction
```

---

## Interview Takeaway

Aggregation combines predictions from multiple models.

---

# Bagging Workflow

```text
Original Dataset
        ↓
Bootstrap Samples
        ↓
Train Multiple Models
        ↓
Combine Predictions
        ↓
Final Prediction
```

---

# Strengths Of Bagging

## Reduces Variance

Major benefit.

---

## More Stable Predictions

Less sensitive to data changes.

---

## Lower Overfitting

Compared to a single tree.

---

# Limitation Of Bagging

Bagging primarily reduces:

```text
Variance
```

---

It does not significantly reduce:

```text
Bias
```

---

## Interview Takeaway

Bagging improves stability but may not solve underfitting.

---

# Boosting

## Core Idea

Instead of training models independently:

```text
Train Sequentially
```

---

Each new model focuses on:

```text
Previous Mistakes
```

---

## Layman Explanation

Imagine a student solving a test.

---

After each attempt:

```text
Review Mistakes
```

---

Then improve.

---

Boosting works similarly.

---

## Interview Takeaway

Boosting learns from previous errors.

---

# How Boosting Works

Step 1

Train model.

---

Step 2

Find mistakes.

---

Step 3

Train next model to focus on mistakes.

---

Step 4

Repeat.

---

Result:

```text
Progressively Better Models
```

---

# Example

Suppose first tree misclassifies:

```text
Fraud Cases
```

---

Next tree gives more attention to:

```text
Fraud Cases
```

---

Eventually:

```text
Difficult Cases Receive More Focus
```

---

## Interview Takeaway

Boosting prioritizes difficult observations.

---

# Why Boosting Works

Boosting reduces:

```text
Bias
```

by continuously improving weak models.

---

Result:

```text
Higher Accuracy
```

---

## Interview Takeaway

Boosting primarily reduces bias.

---

# Strengths Of Boosting

## Extremely Accurate

Often wins machine learning competitions.

---

## Learns Complex Patterns

Very powerful.

---

## Handles Difficult Cases Well

Focuses on mistakes.

---

# Limitations Of Boosting

## Sensitive To Noise

May learn noisy observations.

---

## Slower Training

Models depend on previous models.

---

## Higher Overfitting Risk

If not regularized.

---

# Bagging vs Boosting

| Topic | Bagging | Boosting |
|----------|----------|----------|
| Training | Parallel | Sequential |
| Main Goal | Reduce Variance | Reduce Bias |
| Focus On Mistakes | No | Yes |
| Training Speed | Faster | Slower |
| Overfitting Risk | Lower | Higher |

---

## Interview Takeaway

Bagging reduces variance.

Boosting reduces bias.

---

# Stacking

## Definition

Stacking combines predictions from multiple different algorithms.

---

## Layman Explanation

Instead of using:

```text
Many Trees
```

use:

```text
Decision Tree

Logistic Regression

SVM

Random Forest
```

together.

---

Then train another model to learn:

```text
Which Model To Trust
```

---

## Example

Level 1 Models

```text
Tree

SVM

Logistic Regression
```

---

Predictions

↓

Meta Model

↓

Final Prediction

---

## Interview Takeaway

Stacking learns how to combine multiple algorithms.

---

# Why Ensemble Learning Dominates Competitions

Most winning Kaggle solutions use:

```text
Multiple Models
```

rather than a single model.

---

Reason:

```text
Different Models Capture Different Patterns
```

---

Combining them usually improves performance.

---
