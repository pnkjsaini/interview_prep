# Module 5: Classical Machine Learning

# Chapter 5: Random Forest

In the previous chapter we learned:

```text
Bagging
```

and discovered that a major weakness of Decision Trees is:

```text
High Variance
```

A small change in training data can create a completely different tree.

Bagging helps by training:

```text
Multiple Trees
```

and combining their predictions.

Random Forest takes this idea further and becomes one of the most successful machine learning algorithms ever created.

---

# Why Random Forest Exists

Suppose we train:

```text
100 Decision Trees
```

using Bagging.

---

Problem:

All trees may still look very similar.

---

Example:

If:

```text
Income
```

is the strongest feature,

most trees will split on:

```text
Income
```

first.

---

Result:

```text
Highly Correlated Trees
```

---

If all trees think similarly:

```text
Bagging Benefit Reduces
```

---

Random Forest solves this problem by introducing:

```text
Random Feature Selection
```

---

## Interview Takeaway

Random Forest improves Bagging by creating more diverse trees.

---

# What Is Random Forest?

## Definition

Random Forest is an ensemble of Decision Trees trained using:

```text
Bootstrap Sampling

+

Random Feature Sampling
```

---

Predictions from all trees are combined to produce the final prediction.

---

## Layman Explanation

Instead of consulting:

```text
One Expert
```

consult:

```text
Hundreds Of Experts
```

---

and ensure they do not all think in exactly the same way.

---

## Interview Takeaway

Random Forest combines many diverse Decision Trees.

---

# Core Idea

Random Forest introduces randomness in two places:

---

## Randomness 1

Bootstrap Sampling

---

Different trees receive different datasets.

---

## Randomness 2

Feature Sampling

---

Different trees see different features.

---

Result:

```text
More Diversity

Less Correlation

Better Generalization
```

---

## Interview Takeaway

Random Forest creates diversity through data sampling and feature sampling.

---

# Step 1: Bootstrap Sampling

## Recap

Sampling with replacement.

---

Original Data

```text
A B C D E
```

---

Sample 1

```text
A B B D E
```

---

Sample 2

```text
A C D D E
```

---

Sample 3

```text
B C C E E
```

---

Every tree receives a slightly different dataset.

---

## Why It Matters

Different trees learn different patterns.

---

## Interview Takeaway

Bootstrap sampling reduces variance by creating multiple training datasets.

---

# Step 2: Random Feature Sampling

The real innovation of Random Forest.

---

Suppose dataset contains:

```text
Income

Age

Credit Score

Experience

Loan Amount

Savings
```

---

At every split:

Decision Tree normally evaluates:

```text
All Features
```

---

Random Forest evaluates:

```text
Random Subset Of Features
```

---

Example

At one node:

```text
Income

Age

Savings
```

---

At another node:

```text
Credit Score

Loan Amount

Experience
```

---

## Why It Matters

Trees become:

```text
Less Similar
```

---

This improves ensemble performance.

---

## Interview Takeaway

Random feature sampling creates decorrelated trees.

---

# Random Forest Training Workflow

```text
Original Dataset
          ↓

Bootstrap Sampling
          ↓

Multiple Datasets
          ↓

Decision Trees
          ↓

Random Feature Selection
          ↓

Independent Trees
          ↓

Combine Predictions
          ↓

Final Prediction
```

---

# Prediction In Random Forest

After training:

```text
Tree 1

Tree 2

Tree 3

...

Tree N
```

---

Each tree produces a prediction.

---

# Classification

Use:

```text
Majority Voting
```

---

Example

```text
Fraud

Fraud

Not Fraud

Fraud

Fraud
```

---

Final Prediction

```text
Fraud
```

---

# Regression

Use:

```text
Average Prediction
```

---

Example

```text
₹10 Lakh

₹12 Lakh

₹11 Lakh

₹13 Lakh
```

---

Prediction

```text
₹11.5 Lakh
```

---

## Interview Takeaway

Classification uses voting.

Regression uses averaging.

---

# Why Random Forest Works

Individual trees are:

```text
Weak

Noisy

Unstable
```

---

However:

```text
Average Of Many Trees
```

is usually:

```text
Stable

Accurate

Robust
```

---

This is the:

```text
Wisdom Of Crowds
```

principle.

---

## Interview Takeaway

Combining many imperfect trees produces a strong model.

---

# Out Of Bag (OOB) Samples

One of the most frequently asked Random Forest interview topics.

---

# What Is OOB?

During bootstrap sampling:

```text
Some Observations

Are Not Selected
```

---

Example

Original Data

```text
A B C D E
```

---

Bootstrap Sample

```text
A B B D E
```

---

Missing Observation

```text
C
```

---

C becomes:

```text
Out Of Bag Sample
```

---

# Why OOB Matters

The tree never saw:

```text
C
```

during training.

---

Therefore:

```text
C Can Be Used For Validation
```

---

Without creating a separate validation set.

---

## Interview Takeaway

OOB samples provide an internal estimate of model performance.

---

# OOB Error

## Definition

Prediction error computed using Out Of Bag samples.

---

## Why It Matters

Provides:

```text
Model Validation

Without Cross Validation
```

---

## Interview Takeaway

OOB Error estimates generalization performance.

---

# Feature Importance

One of the biggest advantages of Random Forest.

---

## Definition

Feature Importance measures how much a feature contributes to predictions.

---

## Example

Features

```text
Income

Age

Credit Score

Loan Amount
```

---

Suppose:

```text
Income
```

appears in many important splits.

---

Importance becomes:

```text
High
```

---

## Why It Matters

Provides explainability.

---

## Applications

```text
Feature Selection

Business Insights

Model Interpretation
```

---

## Interview Takeaway

Feature importance identifies influential features.

---

# Hyperparameters Of Random Forest

Interviewers often ask about tuning.

---

# Number Of Trees

```text
n_estimators
```

---

More Trees

```text
Better Stability

Longer Training
```

---

# Maximum Depth

```text
max_depth
```

---

Controls tree complexity.

---

# Minimum Samples Split

```text
min_samples_split
```

---

Controls when splitting occurs.

---

# Maximum Features

```text
max_features
```

---

Controls feature sampling.

---

## Interview Takeaway

Random Forest performance depends heavily on tree count, depth, and feature sampling.

---

# Advantages

## High Accuracy

Usually better than a single tree.

---

## Handles Nonlinear Relationships

No linear assumptions.

---

## Robust To Outliers

More stable than individual trees.

---

## Feature Importance

Built-in interpretability.

---

## Less Overfitting

Compared to deep Decision Trees.

---

# Limitations

## Slower Training

Many trees must be trained.

---

## Larger Memory Usage

Stores multiple trees.

---

## Less Interpretable

Harder to explain than a single tree.

---

## Cannot Extrapolate

Limited to observed patterns.

---

# Random Forest vs Decision Tree

| Topic | Decision Tree | Random Forest |
|----------|----------|----------|
| Trees | One | Many |
| Variance | High | Lower |
| Overfitting | Higher | Lower |
| Accuracy | Lower | Higher |
| Stability | Lower | Higher |
| Interpretability | Better | Lower |

# Common Interview Questions

---

## Question

Why is Random Forest better than a Decision Tree?

### Answer

```text
Because averaging multiple trees reduces variance and improves generalization.
```

---

## Question

What are the two sources of randomness?

### Answer

```text
Bootstrap Sampling

Feature Sampling
```

---

## Question

What is Bootstrap Sampling?

### Answer

```text
Sampling with replacement.
```

---

## Question

Why do we randomly select features?

### Answer

```text
To reduce correlation between trees.
```

---

## Question

What is OOB Error?

### Answer

```text
An internal validation error computed using observations not selected during bootstrap sampling.
```

---

## Question

How does Random Forest perform classification?

### Answer

```text
Using majority voting across trees.
```

---

## Question

How does Random Forest perform regression?

### Answer

```text
By averaging predictions across trees.
```
