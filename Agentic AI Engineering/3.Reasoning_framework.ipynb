# Module 8: Agentic AI Engineering

# Chapter 3: Reasoning Frameworks

In the previous chapter we learned:

```text
What An Agent Is

Reasoning

Planning

Actions

Observations

Memory
```

Now comes the most important question in Agentic AI:

> **How does an agent decide what to do next?**

The answer is **Reasoning**.

Reasoning is the intelligence that transforms an LLM from a text generator into a decision-making system.

This chapter covers the reasoning paradigms behind modern agents.

---

# What Is Reasoning?

Reasoning is the process of:

```text
Understanding

Analyzing

Planning

Making Decisions

Choosing Actions
```

Instead of immediately producing an answer.

---

Example

Question

```text
How many solar panels are required for my house?
```

Without reasoning

```text
Random Estimate
```

With reasoning

```text
Understand Electricity Usage

↓

Estimate Daily Consumption

↓

Estimate Panel Output

↓

Calculate Number Of Panels
```

---

## Interview Takeaway

Reasoning is the process of deriving an answer through intermediate logical steps.

---

# Why LLMs Need Reasoning

Many tasks cannot be solved in one step.

Examples

```text
Financial Analysis

Coding

Research

Trip Planning

Medical Diagnosis

Enterprise Automation
```

Need:

```text
Multi-Step Thinking
```

---

## Interview Takeaway

Reasoning improves performance on complex tasks requiring multiple decisions.

---

# Evolution Of LLM Reasoning

```text
Prompt
      ↓

Direct Answer
      ↓

Chain Of Thought
      ↓

Self Consistency
      ↓

Tree Of Thought
      ↓

Reflection
      ↓

ReAct
      ↓

Planning Agents
      ↓

Reasoning Models
```

---

# Direct Prompting

Simplest approach.

```text
Question
     ↓

Answer
```

---

Advantages

```text
Fast

Cheap
```

---

Problems

```text
Poor Multi-Step Reasoning

Higher Error Rate
```

---

## Interview Takeaway

Direct prompting works well for simple questions but struggles with complex reasoning.

---

# Chain Of Thought (CoT)

One of the most influential prompting techniques.

Instead of:

```text
Answer Directly
```

Model generates:

```text
Reasoning Steps

↓

Final Answer
```

---

Example

```text
Problem

↓

Step 1

↓

Step 2

↓

Step 3

↓

Answer
```

---

## Why CoT Works

Complex problems become:

```text
Smaller Problems
```

Reducing reasoning errors.

---

## Interview Takeaway

Chain of Thought improves reasoning by explicitly generating intermediate steps.

---

# Zero-Shot CoT

Instead of demonstrations,

simply prompt:

```text
Let's Think Step By Step
```

Surprisingly effective.

---

## Interview Takeaway

Zero-shot CoT enables reasoning without few-shot examples.

---

# Few-Shot CoT

Provide:

```text
Examples

+

Reasoning

+

Solutions
```

before asking the real question.

---

Benefits

```text
Higher Accuracy

More Stable Reasoning
```

---

## Interview Takeaway

Few-shot CoT teaches the model how to reason through examples.

---

# Limitations Of Chain Of Thought

Reasoning path:

```text
Single

Linear
```

If one step is wrong:

```text
Entire Solution

May Fail
```

---

## Interview Takeaway

CoT explores only one reasoning path.

---

# Self Consistency

Extension of Chain of Thought.

Instead of:

```text
One Reasoning Path
```

Generate:

```text
Multiple Reasoning Paths
```

Then select:

```text
Most Common Answer
```

---

Example

```text
Path A

↓

42

Path B

↓

42

Path C

↓

39
```

Final answer:

```text
42
```

---

## Interview Takeaway

Self-consistency improves robustness through multiple reasoning attempts.

---

# Why Self Consistency Works

Different reasoning paths may contain:

```text
Different Mistakes
```

Voting reduces errors.

---

## Interview Takeaway

Consensus often outperforms individual reasoning chains.

---

# Tree Of Thought (ToT)

Major advancement over CoT.

Instead of:

```text
One Chain
```

Explore:

```text
Multiple Branches
```

Like:

```text
Decision Tree
```

---

Visualization

```text
Question

     |

 -------------

 |     |      |

Idea1 Idea2 Idea3

 |      |

More Exploration

```

---

## Interview Takeaway

Tree of Thought searches multiple reasoning possibilities.

---

# Why ToT Is Better

Can:

```text
Backtrack

Explore Alternatives

Discard Bad Paths
```

---

Useful for:

```text
Planning

Math

Search Problems

Strategy Games
```

---

## Interview Takeaway

ToT enables search-based reasoning.

---

# Graph Of Thought (GoT)

Generalizes Tree of Thought.

Instead of:

```text
Tree
```

Use:

```text
Graph
```

Ideas can:

```text
Merge

Reuse

Reconnect
```

---

Benefits

```text
More Flexible Reasoning
```

---

## Interview Takeaway

GoT models reasoning as interconnected ideas rather than a strict tree.

---

# Reflection

One of the hottest agentic concepts.

Pipeline

```text
Generate

↓

Critique

↓

Improve
```

Instead of trusting the first answer.

---

Example

```text
Write Code

↓

Review Code

↓

Fix Bugs
```

---

## Interview Takeaway

Reflection improves quality through self-review.

---

# Reflexion

An extension of reflection.

Agent remembers:

```text
Previous Mistakes
```

Uses them in future reasoning.

Pipeline

```text
Task

↓

Attempt

↓

Evaluate

↓

Store Lesson

↓

Retry
```

---

## Interview Takeaway

Reflexion combines reasoning with learning from failures.

---

# Plan And Execute

Separate:

```text
Planning

Execution
```

Pipeline

```text
Goal

↓

Planner

↓

Task List

↓

Executor

↓

Results
```

---

Advantages

```text
Structured

Predictable

Scalable
```

---

## Interview Takeaway

Plan-and-Execute separates strategic planning from execution.

---

# ReAct

The most important reasoning framework.

ReAct means:

```text
Reason

+

Act
```

---

Instead of only reasoning,

the agent:

```text
Reasons

↓

Calls Tool

↓

Observes

↓

Reasons Again
```

---

Pipeline

```text
Thought

↓

Action

↓

Observation

↓

Thought

↓

Action

↓

Answer
```

---

## Why ReAct Changed Everything

Earlier prompting:

```text
Reason Only
```

ReAct:

```text
Reason

+

Interact With Environment
```

---

## Interview Takeaway

ReAct combines reasoning and tool usage in an iterative loop.

---

# ReAct Example

Question

```text
Who is the CEO of the company that acquired GitHub?
```

Reasoning

```text
Need Company

↓

Search GitHub Acquisition
```

Observation

```text
Microsoft
```

Reasoning

```text
Need Microsoft CEO
```

Action

```text
Search CEO
```

Observation

```text
Satya Nadella
```

Final Answer

```text
Satya Nadella
```

---

## Interview Takeaway

ReAct alternates between thinking and acting.

---

# Planning vs ReAct

Plan And Execute

```text
Plan First

Execute Later
```

---

ReAct

```text
Think

Act

Observe

Repeat
```

---

## Interview Takeaway

Planning is proactive, while ReAct adapts after every observation.

---

# Test-Time Compute

Modern reasoning models allocate:

```text
More Compute

During Inference
```

Instead of larger models.

Examples

```text
Reflection

Multiple CoT

Search

Verification
```

---

## Interview Takeaway

Reasoning quality can improve by increasing inference-time computation.

---

# Reasoning Models

Modern frontier models:

```text
OpenAI o-series

DeepSeek R1

Gemini Thinking

Claude Reasoning
```

Instead of answering immediately,

they:

```text
Think

Plan

Verify

Answer
```

---

## Interview Takeaway

Reasoning models trade latency for higher-quality solutions.

---

# Choosing The Right Framework

| Problem | Framework |
|----------|-----------|
| Simple QA | Direct Prompt |
| Multi-Step Logic | Chain Of Thought |
| Higher Accuracy | Self Consistency |
| Search Problems | Tree Of Thought |
| Dynamic Tasks | ReAct |
| Long Workflows | Plan And Execute |
| Self Correction | Reflection |
| Learning From Mistakes | Reflexion |

---

# Production Agent Reasoning

Typical enterprise loop

```text
Goal
    ↓

Reason

↓

Select Tool

↓

Execute

↓

Observe

↓

Update State

↓

Continue

↓

Goal Achieved
```

---

# Common Mistakes

---

Using CoT for every problem

---

Planning when ReAct is sufficient

---

Too many reasoning loops

---

Ignoring tool observations

---

No verification step

---

## Interview Takeaway

The best reasoning strategy depends on the task.

# Common Interview Questions

---

## Question

What is Chain Of Thought?

### Answer

```text
Generating intermediate reasoning steps before producing the final answer.
```

---

## Question

Why does CoT improve performance?

### Answer

```text
Because it decomposes complex problems into smaller logical steps.
```

---

## Question

What is Self Consistency?

### Answer

```text
Generating multiple reasoning paths and selecting the most consistent answer.
```

---

## Question

Tree Of Thought vs Chain Of Thought?

### Answer

```text
CoT explores one reasoning path.

ToT explores multiple reasoning branches.
```

---

## Question

What is Reflection?

### Answer

```text
Reviewing and improving a generated solution before returning it.
```

---

## Question

What is Reflexion?

### Answer

```text
Reflection combined with learning from previous failures.
```

---

## Question

What is ReAct?

### Answer

```text
A reasoning framework that alternates between Thought, Action, and Observation.
```

---

## Question

Plan And Execute vs ReAct?

### Answer

```text
Plan-and-Execute creates the plan first.

ReAct plans incrementally after each observation.
```

---

## Question

Why are reasoning models becoming popular?

### Answer

```text
Because they improve complex reasoning without necessarily increasing model size.
```
