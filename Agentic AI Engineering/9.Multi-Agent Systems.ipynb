# Module 8: Agentic AI Engineering

# Chapter 9: Multi-Agent Systems

In the previous chapter we explored the ecosystem of Agent frameworks.

However, regardless of the framework used, one question remains:

> **When should we use multiple agents instead of one?**

Many companies are currently experimenting with Multi-Agent systems, but one of the biggest misconceptions is:

```text
More Agents

=

Better System
```

This is **not true**.

A poorly designed multi-agent system is often:

```text
Slower

More Expensive

Harder To Debug

Less Reliable
```

This chapter explains when Multi-Agent architectures are justified and how they are designed in production.

---

# What Is A Multi-Agent System?

A Multi-Agent System (MAS) consists of:

```text
Multiple Specialized Agents

Collaborating

To Solve A Common Goal
```

Instead of one LLM performing every task,

multiple agents divide responsibilities.

---

## Interview Takeaway

A Multi-Agent System distributes work among specialized agents.

---

# Why Multi-Agent Systems Exist

Consider a request:

```text
Prepare Investment Report
```

The system must:

```text
Collect News

↓

Retrieve Financial Data

↓

Perform Risk Analysis

↓

Generate Charts

↓

Write Report

↓

Review Report
```

One agent can perform everything.

But should it?

Not always.

---

## Interview Takeaway

Multi-Agent systems improve specialization for complex workflows.

---

# Single Agent vs Multi-Agent

Single Agent

```text
User

↓

Agent

↓

Tools

↓

Answer
```

---

Multi-Agent

```text
User

↓

Coordinator

↓

Research

SQL

Python

Writer

Reviewer

↓

Answer
```

---

## Interview Takeaway

Single-agent systems centralize intelligence, while multi-agent systems distribute it.

---

# Characteristics Of Multi-Agent Systems

Every MAS contains:

```text
Multiple Agents

Communication

Coordination

Task Allocation

Shared Goal
```

Without these,

it's simply multiple LLM calls.

---

## Interview Takeaway

Communication and coordination distinguish a true multi-agent system.

---

# Agent Specialization

Each agent should own one responsibility.

Example

```text
Research Agent

↓

Find Information
```

---

SQL Agent

```text
Database Queries
```

---

Python Agent

```text
Calculations
```

---

Reviewer Agent

```text
Verify Output
```

---

## Interview Takeaway

Specialized agents reduce prompt complexity and improve maintainability.

---

# Coordinator Agent

Most systems require one coordinating agent.

Responsibilities

```text
Receive User Request

↓

Assign Tasks

↓

Collect Results

↓

Return Final Response
```

The coordinator does **not** solve every problem itself.

---

## Interview Takeaway

The coordinator manages collaboration rather than execution.

---

# Supervisor Pattern

The most common production architecture.

```text
User

↓

Supervisor

↓

Worker A

↓

Worker B

↓

Worker C

↓

Supervisor

↓

Response
```

Supervisor decides:

```text
Who Works Next
```

based on current state.

---

## Interview Takeaway

Supervisor-worker is the dominant enterprise multi-agent architecture.

---

# Planner-Worker Pattern

Architecture

```text
Goal

↓

Planner

↓

Task List

↓

Workers

↓

Results
```

Planner never performs execution.

Workers never perform planning.

---

## Interview Takeaway

Separating planning from execution improves scalability.

---

# Peer-To-Peer Agents

No central coordinator.

Architecture

```text
Agent A

↔

Agent B

↔

Agent C
```

Agents communicate directly.

---

Advantages

```text
Flexible

Distributed
```

---

Problems

```text
Hard To Coordinate

Loop Risk

Complex Debugging
```

---

## Interview Takeaway

Peer-to-peer systems provide flexibility but increase coordination complexity.

---

# Blackboard Pattern

Agents share:

```text
Common Workspace
```

Architecture

```text
Shared Memory

↓

Research Agent

↓

Writer

↓

Reviewer
```

Agents never communicate directly.

Instead they read and update shared state.

---

## Interview Takeaway

Blackboard systems coordinate through shared memory.

---

# Committee Pattern

Several agents solve:

```text
Same Problem
```

Example

```text
Agent A

Agent B

Agent C
```

All generate answers.

Another agent chooses:

```text
Best Solution
```

---

Applications

```text
Critical Decisions

Scientific Research

Medical AI
```

---

## Interview Takeaway

Committee architectures improve robustness through diversity.

---

# Debate Pattern

Instead of agreeing,

agents challenge each other.

Architecture

```text
Agent A

↓

Argument

↓

Agent B

↓

Counterargument

↓

Judge
```

---

Applications

```text
Legal

Finance

Scientific Reasoning
```

---

## Interview Takeaway

Debate encourages deeper reasoning by exposing weaknesses.

---

# Reviewer Pattern

Architecture

```text
Solver

↓

Reviewer

↓

Improved Answer
```

Instead of:

```text
Multiple Solvers
```

---

Applications

```text
Coding

Report Writing

Math

Contracts
```

---

## Interview Takeaway

A reviewer agent improves quality through verification.

---

# Shared State

```text
Shared State
```

```text
Research Complete

↓

State Updated

↓

Writer Reads State
```

## Interview Takeaway

---

```text
Research

SQL

Python

↓

Merge
```

## Interview Takeaway

# Synchronization

Parallel agents eventually need to synchronize.

Example

```text
Research

↓

Done

SQL

↓

Done

↓

Writer Starts
```

Without synchronization,

results may be incomplete.

---

## Interview Takeaway

Synchronization coordinates parallel execution.

---

# Common Problems

---

## Infinite Loops

Agent A

↓

Agent B

↓

Agent A

↓

Agent B

Forever.

---

Solution

```text
Maximum Iterations

Loop Detection
```

---

## Deadlocks

Agent A waits for B.

Agent B waits for A.

Nobody proceeds.

---

Solution

```text
Timeouts

Dependency Analysis
```

---

## Token Explosion

Too many agents exchanging messages.

Results in:

```text
Huge Prompts

High Cost

Slow Response
```

---

Solution

```text
Summaries

Shared State

Compression
```

---

## Duplicate Work

Multiple agents solve:

```text
Same Task
```

Unnecessarily.

---

Solution

```text
Task Registry

Coordinator
```

---

## Hallucination Propagation

One agent hallucinates.

Other agents trust it.

Entire workflow fails.

---

Solution

```text
Verification

Reviewer

Grounding
```

---

## Interview Takeaway

Most multi-agent failures are coordination problems rather than model problems.

---

# Cost Considerations

Question

Does the additional intelligence justify the cost?

## Interview Takeaway

# Common Interview Questions

---

## Question

What is a Multi-Agent System?

### Answer

```text
A system where multiple specialized AI agents collaborate to solve a shared objective.
```

---

## Question

Why not always use multiple agents?

### Answer

```text
Because they increase complexity, latency, coordination overhead, and cost.
```

---

## Question

What is the Supervisor pattern?

### Answer

```text
A supervisor coordinates worker agents by assigning tasks and combining results.
```

---

## Question

How do agents communicate?

### Answer

```text
Through direct messaging, shared memory, supervisors, or event-driven mechanisms.
```

---

## Question

What causes token explosion?

### Answer

```text
Excessive communication and repeated context sharing between agents.
```

---

## Question

What is the Blackboard pattern?

### Answer

```text
Agents collaborate by reading from and writing to a shared state rather than communicating directly.
```

---

## Question

When is a single agent sufficient?

### Answer

```text
For simple, deterministic tasks that do not require specialized expertise.
```
