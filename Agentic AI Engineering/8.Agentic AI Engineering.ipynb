# Module 8: Agentic AI Engineering

# Chapter 8: Agent Framework Ecosystem

In the previous chapter we learned how LangGraph provides a powerful framework for building production-grade AI agents.

However, LangGraph is **not the only framework** available.

Today, the Agentic AI ecosystem contains multiple frameworks, each designed with different goals.

Choosing the right framework is an architectural decision.

This chapter explains the most popular frameworks, their internal philosophy, strengths, weaknesses, and when to use each.

---

# Why Agent Frameworks Exist

Without frameworks, developers would have to build:

```text
State Management

Tool Calling

Memory

Planning

Retries

Execution Engine

Observability

Persistence
```

from scratch.

Frameworks provide these building blocks.

---

## Interview Takeaway

Agent frameworks abstract common agent infrastructure, allowing developers to focus on business logic.

---

# LangChain

The framework that popularized LLM application development.

Designed for:

```text
Chains

RAG

Prompt Pipelines

Tool Calling
```

---

Architecture

```text
Prompt

↓

LLM

↓

Parser

↓

Output
```

---

Strengths

```text
Large Ecosystem

Many Integrations

Easy To Learn

Strong Community
```

---

Limitations

```text
Linear Execution

Limited State

Complex Agent Workflows Become Difficult
```

---

Best For

```text
RAG

Chatbots

Simple AI Applications
```

---

## Interview Takeaway

LangChain is excellent for building LLM applications but is less suited for complex stateful agents.

---

# LangGraph

Built by the LangChain team.

Purpose:

```text
Production Agent Systems
```

---

Core Philosophy

```text
Everything

Is

A Graph
```

---

Strengths

```text
State Management

Persistence

Loops

Checkpointing

Human Approval

Multi-Agent

Conditional Routing
```

---

Limitations

```text
Steeper Learning Curve

Requires Architectural Thinking
```

---

Best For

```text
Enterprise Agents

Long Running Workflows

Production AI Systems
```

---

## Interview Takeaway

LangGraph is currently one of the strongest frameworks for enterprise-grade agent orchestration.

---

# CrewAI

CrewAI focuses on:

```text
Role Playing Agents
```

Instead of building graphs,

you define:

```text
Agents

Roles

Goals

Tasks
```

---

Example

```text
Researcher

↓

Writer

↓

Reviewer
```

Each behaves like a team member.

---

Strengths

```text
Simple

Fast Development

Natural Multi-Agent Design
```

---

Limitations

```text
Less Control

Limited Workflow Customization

Smaller Ecosystem
```

---

Best For

```text
Research Automation

Content Generation

Rapid Prototyping
```

---

## Interview Takeaway

CrewAI emphasizes collaborative role-based agents.

---

# AutoGen

Developed by Microsoft.

Purpose:

```text
Agent To Agent Communication
```

Architecture

```text
Planner

↓

Coder

↓

Executor

↓

Reviewer
```

Conversation itself becomes:

```text
Execution Engine
```

---

Strengths

```text
Excellent Multi-Agent Conversations

Code Generation

Research Tasks
```

---

Limitations

```text
Token Intensive

Harder Debugging

Conversation Explosion
```

---

Best For

```text
Collaborative Agent Systems

Research

Software Development
```

---

## Interview Takeaway

AutoGen models workflows as conversations between agents.

---

# PydanticAI

One of the fastest-growing frameworks.

Built around:

```text
Python Types
```

Instead of prompts.

Example

```python
class Report(BaseModel):
    summary: str
    risk: float
```

---

Strengths

```text
Type Safety

Validation

Developer Productivity

Structured Outputs
```

---

Limitations

```text
Smaller Community

New Ecosystem
```

---

Best For

```text
Enterprise APIs

Backend AI Services
```

---

## Interview Takeaway

# Can Frameworks Be Combined?

Absolutely.

Example

```text
LangGraph

↓

Uses LangChain Tools

↓

Calls OpenAI SDK

↓

Uses Pydantic Models

↓

Stores Memory In Redis
```

Frameworks are **not mutually exclusive**.

---

## Interview Takeaway

Production systems often combine multiple frameworks instead of relying on one.

---

# Framework Lock-In

Avoid tightly coupling business logic to a framework.

Instead:

```text
Business Logic

↓

Framework Adapter

↓

Framework
```

Benefits

```text
Easy Migration

Better Testing

Vendor Independence
```

---

## Interview Takeaway

Keep business logic independent of framework-specific APIs.

---

# Common Interview Questions

---

## Question

Why was LangGraph introduced?

### Answer

```text
To support stateful, graph-based workflows with persistence, loops, and multi-agent orchestration beyond LangChain's linear chains.
```

---

## Question

LangChain vs LangGraph?

### Answer

```text
LangChain focuses on chains and application development.

LangGraph focuses on stateful agent orchestration.
```

---

## Question

When should you use CrewAI?

### Answer

```text
When building collaborative, role-based multi-agent systems.
```

---

## Question

How is AutoGen different from CrewAI?

### Answer

```text
AutoGen emphasizes conversation between agents, while CrewAI focuses on role-based task execution.
```

---

## Question

Why would an enterprise choose Semantic Kernel?

### Answer

```text
Because it integrates well with the Microsoft ecosystem and enterprise security requirements.
```

---

## Question

Why is PydanticAI becoming popular?

### Answer

```text
Because it provides strong typing, validation, and structured outputs for AI applications.
```

---

## Question

Can multiple frameworks be used together?

### Answer

```text
Yes. Production systems often combine frameworks to leverage their respective strengths.
```
