# Module 8: Agentic AI Engineering

# Chapter 5: Agent Memory Systems

In the previous chapters we learned:

```text
Tool Calling

Agent Fundamentals

Reasoning

Planning
```

One question still remains:

> **How does an AI Agent remember information?**

A normal LLM forgets everything outside its context window.

An Agent, however, can remember:

```text
Previous Conversations

Past Tasks

User Preferences

Tool Outputs

Long-Term Knowledge
```

Memory is one of the biggest differences between a chatbot and a production-grade AI agent.

---

# Why Do Agents Need Memory?

Imagine asking:

```text
Book my usual flight.
```

Without memory, the agent asks:

```text
Which city?

Which airline?

Which seat?

Which airport?
```

Every single time.

With memory:

```text
Home Airport

Preferred Airline

Window Seat

Business Class
```

The task becomes effortless.

---

## Interview Takeaway

Memory allows an agent to use information beyond the current prompt.

---

# LLM Context Is Not Memory

This is one of the most misunderstood concepts.

Many people believe:

```text
Context Window

=

Memory
```

This is incorrect.

---

Context Window

```text
Temporary

Limited

Lost After Conversation
```

---

Memory

```text
Persistent

Retrievable

Reusable
```

---

Example

Current Context

```text
Previous 20 Messages
```

Memory

```text
User likes Python

Works at BlackRock

Prefers SQL over NoSQL
```

Weeks later.

---

## Interview Takeaway

Context is temporary; memory is persistent.

---

# Memory vs RAG

Another common interview question.

RAG retrieves:

```text
External Knowledge
```

Memory retrieves:

```text
Personal Or Task Information
```

---

Example

RAG

```text
Company Policy
```

Memory

```text
User Always Wants PDF Reports
```

---

## Interview Takeaway

RAG retrieves knowledge, while memory retrieves experience.

---

# Why Memory Exists

Without memory:

```text
Agent Starts From Zero

Every Conversation
```

With memory:

```text
Agent Learns

Over Time
```

---

Benefits

```text
Personalization

Efficiency

Long-Term Tasks

Consistency
```

---

# Types Of Memory

Modern agents usually implement multiple memory systems.

---

# Working Memory

Temporary memory.

Contains:

```text
Current Goal

Current Plan

Current Tool Outputs
```

Example

```text
Search Result

↓

Immediately Used
```

Not stored permanently.

---

## Interview Takeaway

Working memory supports the current reasoning process.

---

# Short-Term Memory

Stores information during a conversation.

Example

```text
User

↓

Upload CSV

↓

Ask Questions

↓

Generate Charts
```

The agent remembers:

```text
Uploaded File
```

throughout the session.

---

## Interview Takeaway

Short-term memory survives within a session.

---

# Long-Term Memory

Persists across sessions.

Examples

```text
Preferred Language

Favorite IDE

Business Domain

Frequently Used Reports
```

Stored in:

```text
Database

Vector Store

Knowledge Graph
```

---

## Interview Takeaway

Long-term memory enables personalization across conversations.

---

# Semantic Memory

Stores:

```text
Facts

Knowledge

Concepts
```

Example

```text
User Works In Finance

Uses Azure

Uses Databricks
```

Not tied to a specific event.

---

## Interview Takeaway

Semantic memory stores generalized knowledge.

---

# Episodic Memory

Stores:

```text
Experiences

Events

Previous Tasks
```

Example

```text
Yesterday

Generated Quarterly Report
```

---

## Interview Takeaway

Episodic memory remembers past experiences.

---

# Procedural Memory

Stores:

```text
How To Perform Tasks
```

Example

```text
Deployment Workflow

Company Approval Process

Internal SOP
```

---

## Interview Takeaway

Procedural memory stores learned processes.

---

# Memory Hierarchy

```text
Working Memory
        ↓

Short-Term Memory
        ↓

Long-Term Memory
        ↓

Semantic / Episodic / Procedural
```

---

# Memory Lifecycle

Every memory follows a lifecycle.

```text
Experience

↓

Store

↓

Retrieve

↓

Update

↓

Archive

↓

Forget
```

---

## Interview Takeaway

Memory is dynamic rather than static.

---

# Memory Storage

Where should memories live?

Options

```text
SQL Database

Vector Database

Graph Database

Document Store

Redis Cache
```

Each has different strengths.

---

# SQL Memory

Good for:

```text
Structured Data

Profiles

Settings

Configurations
```

Example

```text
Preferred Language

Theme

Timezone
```

---

## Interview Takeaway

SQL is ideal for structured user preferences.

---

# Vector Memory

Store:

```text
Embeddings
```

Retrieve using:

```text
Semantic Similarity
```

Useful for:

```text
Past Conversations

Notes

Documents
```

---

## Interview Takeaway

Vector databases enable semantic memory retrieval.

---

# Graph Memory

Stores:

```text
Relationships
```

Example

```text
Alice

↓

Works At

↓

BlackRock
```

Supports:

```text
Multi-Hop Retrieval
```

---

## Interview Takeaway

Graph memory captures relationships between entities.

---

# Hybrid Memory

Modern enterprise agents use:

```text
SQL

+

Vector DB

+

Graph DB
```

Each memory type stores different information.

---

## Interview Takeaway

Hybrid memory combines multiple storage technologies.

---

# Memory Retrieval

Question:

```text
Which Memories

Should Be Loaded?
```

Agent cannot retrieve:

```text
Everything
```

Need:

```text
Relevant Memories
```

---

Pipeline

```text
User Query

↓

Retrieve Memories

↓

Rank

↓

Inject Into Prompt
```

---

## Interview Takeaway

Memory retrieval is a search problem.

---

# Memory Ranking

Suppose there are:

```text
1000 Memories
```

Need only:

```text
Top Relevant Memories
```

Ranking considers:

```text
Similarity

Recency

Importance

Frequency
```

---

## Interview Takeaway

Not every memory has equal importance.

---

# Memory Compression

Long conversations become expensive.

Need:

```text
Summaries
```

Instead of:

```text
Entire History
```

Example

100 messages

↓

One summary.

---

## Interview Takeaway

Compression reduces token usage while preserving context.

---

# Memory Consolidation

Several similar memories become:

```text
One Memory
```

Example

```text
User Likes Python

User Uses Python Daily

User Prefers Python
```

↓

```text
User Prefers Python
```

---

## Interview Takeaway

Consolidation removes redundancy.

---

# Memory Forgetting

Not every memory should remain forever.

Need:

```text
Expiration

Deletion

Decay
```

Example

```text
Temporary OTP

↓

Delete
```

---

## Interview Takeaway

Forgetting prevents stale information accumulation.

---

# Memory Importance

Agents score memories.

High Importance

```text
Permanent Preferences
```

Low Importance

```text
Today's Weather
```

---

Memory score may depend on:

```text
Frequency

Recency

User Explicitly Saved

Business Value
```

---

## Interview Takeaway

Memory importance determines retention.

---

# Memory Update

Memories change.

Old

```text
Works At Company A
```

New

```text
Works At Company B
```

Agent updates instead of duplicating.

---

## Interview Takeaway

Memory must remain consistent over time.

# Common Interview Questions

---

## Question

Why isn't context window considered memory?

### Answer

```text
Because context is temporary and disappears after the conversation ends.
```

---

## Question

Memory vs RAG?

### Answer

```text
Memory stores user- or task-specific information.

RAG retrieves external knowledge.
```

---

## Question

Short-term vs Long-term Memory?

### Answer

```text
Short-term memory lasts for a session.

Long-term memory persists across sessions.
```

---

## Question

What is semantic memory?

### Answer

```text
General facts and knowledge about users or the world.
```

---

## Question

What is episodic memory?

### Answer

```text
Memories of specific past events or interactions.
```

---

## Question

Why use vector databases for memory?

### Answer

```text
They enable semantic retrieval using embeddings.
```

---

## Question

Why compress memories?

### Answer

```text
To reduce token usage while preserving important information.
```

---

## Question

What is hybrid memory?

### Answer

```text
Using multiple storage systems such as SQL, vector databases, and graph databases together.
```
