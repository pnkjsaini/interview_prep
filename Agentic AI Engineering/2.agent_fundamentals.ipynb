# Module 8: Agentic AI Engineering

# Chapter 2: Agent Fundamentals

In the previous chapter we learned:

```text
Tool Calling

Function Calling

Structured Outputs

Tool Routing

Tool Execution
```

Now an important question arises:

> If an LLM can call tools, does that automatically make it an Agent?

**No.**

Tool calling is only one capability.

An Agent requires:

```text
Reasoning

Planning

Decision Making

Tool Usage

Observation

Iteration
```

This chapter explains what actually makes an AI system an Agent.

---

# What Is An AI Agent?

An AI Agent is an autonomous software system that:

```text
Perceives

Reasons

Plans

Acts

Observes

Repeats

Until Goal Is Achieved
```

Unlike a chatbot,

an agent focuses on:

```text
Completing Tasks
```

rather than merely:

```text
Generating Text
```

---

## Interview Takeaway

An AI agent is a goal-oriented system capable of interacting with its environment to accomplish tasks.

---

# LLM vs Workflow vs Agent

One of the most common interview questions.

---

# LLM

```text
User
    ↓

LLM
    ↓

Response
```

Characteristics:

```text
Single Prompt

Single Response

No Decisions
```

---

Example

```text
Explain Gradient Descent.
```

↓

Answer.

---

## Interview Takeaway

An LLM generates responses but does not autonomously perform tasks.

---

# AI Workflow

A workflow follows:

```text
Fixed Steps
```

Example

```text
Read PDF
      ↓

Extract Text
      ↓

Summarize
      ↓

Translate
```

Every execution follows:

```text
Same Sequence
```

---

## Interview Takeaway

A workflow is deterministic and predefined.

---

# AI Agent

Agent execution is dynamic.

```text
User Request
       ↓

Reason
       ↓

Choose Tool
       ↓

Observe
       ↓

Reason Again
       ↓

Continue
```

No predefined number of steps.

---

## Interview Takeaway

Agents dynamically decide what actions to perform.

---

# Workflow vs Agent

| Workflow | Agent |
|------------|------------|
| Fixed Steps | Dynamic Decisions |
| Deterministic | Adaptive |
| No Reasoning | Uses Reasoning |
| Limited Flexibility | High Flexibility |

---

## Interview Takeaway

Workflows execute predefined logic while agents adapt to changing situations.

---

# Goal-Oriented Systems

Agents are designed around:

```text
Goals
```

Example

User says:

```text
Book me the cheapest flight.
```

The goal is:

```text
Find Best Flight
```

Not:

```text
Generate Text
```

---

## Interview Takeaway

Agents optimize for goal completion rather than response generation.

---

# Agent Components

Every agent contains these building blocks.

```text
Goal
     ↓

Reasoning
     ↓

Planning
     ↓

Actions
     ↓

Observation
     ↓

Memory
```

---

Let's study each.

---

# Goal

The objective the agent must accomplish.

Examples

```text
Schedule Meeting

Generate Report

Analyze Portfolio

Book Flight
```

Everything else revolves around the goal.

---

## Interview Takeaway

Goals drive agent behavior.

---

# Reasoning

Reasoning answers:

```text
What Should I Do Next?
```

Example

```text
Need Customer ID

↓

Search CRM
```

Reasoning decides the next step.

---

## Interview Takeaway

Reasoning determines the next action.

---

# Planning

Planning answers:

```text
How Can I Reach The Goal?
```

Example

```text
Find Customer
      ↓

Retrieve Orders
      ↓

Generate Summary
      ↓

Email Customer
```

Planning determines:

```text
Overall Strategy
```

---

## Interview Takeaway

Planning determines the sequence of actions.

---

# Action

Actions change the environment.

Examples

```text
Call API

Run SQL

Execute Python

Search Documents

Send Email
```

---

## Interview Takeaway

Actions interact with external systems.

---

# Observation

After every action,

the agent receives:

```text
Observation
```

Example

SQL Tool

↓

```text
Customer Found
```

Observation becomes:

```text
New Information
```

---

## Interview Takeaway

Observations provide feedback from the environment.

---

# Memory

Agents often need:

```text
Past Information
```

Example

```text
Customer Name

Previous Conversation

Preferences
```

Memory prevents:

```text
Repeated Questions
```

---

## Interview Takeaway

Memory allows agents to retain useful information across steps.

---

# The Agent Loop

Every modern agent follows a loop.

```text
Goal
     ↓

Reason
     ↓

Plan
     ↓

Action
     ↓

Observation
     ↓

Reason Again
     ↓

Repeat
```

Until:

```text
Goal Achieved
```

---

## Interview Takeaway

Agents continuously reason based on observations.

---

# ReAct Loop

Most production agents use:

```text
Reason

↓

Act

↓

Observe

↓

Repeat
```

We'll study ReAct deeply in Chapter 3.

---

## Interview Takeaway

ReAct is the dominant reasoning loop used in production agents.

---

# Agent Environment

Agents do not operate in isolation.

Environment includes:

```text
Databases

APIs

Files

Humans

Search Engines

Other Agents
```

Agent interacts continuously.

---

## Interview Takeaway

The environment provides information and execution targets.

---

# Agent State

State represents:

```text
Current Situation
```

Example

```text
Current Goal

Completed Tasks

Pending Tasks

Tool Results

Conversation
```

State evolves during execution.

---

## Interview Takeaway

Agent state captures execution progress.

---

# Stateless vs Stateful Agents

Stateless

```text
Forget Everything

After Response
```

---

Stateful

```text
Remember Previous Information
```

---

## Interview Takeaway

Stateful agents support multi-step reasoning.

---

# Autonomy Levels

Not every agent is fully autonomous.

---

# Level 0

```text
Chatbot
```

Only answers questions.

---

# Level 1

```text
Tool Calling
```

Uses tools when instructed.

---

# Level 2

```text
Task Execution
```

Chooses tools automatically.

---

# Level 3

```text
Planning
```

Creates execution plans.

---

# Level 4

```text
Multi-Agent Collaboration
```

Coordinates multiple agents.

---

# Level 5

```text
Highly Autonomous
```

Long-running goal execution.

---

## Interview Takeaway

Autonomy exists on a spectrum rather than being binary.

---

# Reactive vs Proactive Agents

Reactive

```text
Wait

↓

Respond
```

---

Proactive

```text
Predict

↓

Act
```

Example

```text
Monitor Stock

Notify User

Automatically
```

---

## Interview Takeaway

Proactive agents initiate actions without explicit user requests.

---

# Single-Agent Systems

Architecture

```text
User
    ↓

Agent
    ↓

Tools
```

Simple.

Easy to manage.

---

Advantages

```text
Low Cost

Simple Design
```

---

Limitations

```text
Limited Parallelism

Single Decision Maker
```

---

## Interview Takeaway

Single-agent systems are easier to build but have limited scalability.

---

# Multi-Agent Systems

Architecture

```text
Planner
      ↓

Research Agent

Code Agent

Reviewer

Executor
```

Agents collaborate.

---

Advantages

```text
Parallel Work

Specialization
```

---

Limitations

```text
Coordination

Higher Cost

Communication Overhead
```

---

## Interview Takeaway

Multi-agent systems improve specialization at the cost of complexity.

---

# Agent Failure Modes

Production interview favorite.

---

# Hallucination

Wrong reasoning.

---

# Wrong Tool Selection

Incorrect API chosen.

---

# Infinite Loops

Agent keeps repeating.

---

# Invalid Planning

Poor task decomposition.

---

# Tool Failure

External service unavailable.

---

# Context Loss

Forgets earlier information.

---

# Permission Errors

No access.

---

## Interview Takeaway

Most production failures originate outside the LLM itself.

---

# How Production Agents Recover

Common strategies

```text
Retries

Fallback Tools

Human Approval

Timeouts

Loop Detection

Checkpoint Recovery
```

---

## Interview Takeaway

Recovery mechanisms are essential for reliable agents.
