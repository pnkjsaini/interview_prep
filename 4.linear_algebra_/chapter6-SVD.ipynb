# Module 4: Linear Algebra for Machine Learning

# Chapter 6: Singular Value Decomposition (SVD)

Singular Value Decomposition (SVD) is one of the most important concepts in Machine Learning.

Many seemingly unrelated techniques are built on SVD:

```text
PCA

Recommendation Systems

Embeddings

Latent Semantic Analysis (LSA)

Image Compression

LoRA

Model Compression
```

If there is one linear algebra concept every ML engineer should deeply understand, it is:

```text
SVD
```

---

# Why SVD Exists

Before SVD, we learned:

```text
Eigendecomposition
```

which expresses a matrix using:

```text
Eigenvectors

Eigenvalues
```

---

However eigendecomposition has limitations.

---

## Limitation 1

Requires:

```text
Square Matrices
```

---

Real-world datasets often produce:

```text
Rectangular Matrices
```

---

Example

```text
1000 Customers

50 Features
```

Matrix Shape:

```text
1000 × 50
```

---

Not square.

---

## Limitation 2

Some matrices do not have enough eigenvectors.

---

## Limitation 3

Real-world data is noisy and messy.

---

We need something:

```text
More General

More Stable

More Useful
```

---

This leads to:

```text
Singular Value Decomposition
```

---

# What Is SVD?

## Definition

SVD decomposes any matrix into three simpler matrices.

---

## Formula

```text
A = UΣVᵀ
```

---

Where:

```text
U
```

Output Directions

---

```text
Σ
```

Importance Of Directions

---

```text
Vᵀ
```

Input Directions

---

## Layman Explanation

Imagine a complex machine.

Instead of studying the machine directly:

```text
Break It Into Components
```

---

SVD breaks a matrix into:

```text
Directions

Importance

Transformation
```

---

## Why It Matters

SVD reveals:

```text
Hidden Structure

Redundancy

Noise

Important Patterns
```

inside data.

---

## Interview Takeaway

SVD decomposes any matrix into directions and their importance.

---

# The Big Picture

Original Matrix

```text
A
```

↓

Decompose

↓

```text
U
```

Output Directions

---

```text
Σ
```

Importance

---

```text
Vᵀ
```

Input Directions

---

# Understanding U Matrix

## Definition

U contains orthogonal output directions.

---

## Layman Explanation

Think of U as:

```text
Important Directions
```

in the transformed space.

---

After SVD is performed:

```text
U
```

tells us where information eventually flows.

---

## Why It Matters

U represents important output patterns.

---

## Interview Takeaway

U contains important output directions.

---

# Understanding Σ Matrix

## Definition

Σ contains singular values.

---

## Layman Explanation

This is the most important matrix in SVD.

---

Singular values tell us:

```text
How Important A Direction Is
```

---

Large Singular Value

```text
Important Information
```

---

Small Singular Value

```text
Noise

Redundancy
```

---

## Example

Singular Values

```text
100

20

5

0.1

0.01
```

---

Interpretation:

```text
100

20

5
```

contain most information.

---

```text
0.1

0.01
```

contribute very little.

---

## Why It Matters

Most dimensionality reduction comes from analyzing singular values.

---

## Interview Takeaway

Singular values measure the importance of information.

---

# Understanding Vᵀ Matrix

## Definition

Vᵀ contains orthogonal input directions.

---

## Layman Explanation

Vᵀ identifies:

```text
Important Directions
```

in the original data.

---

If features contain hidden patterns:

```text
Vᵀ
```

 helps uncover them.

---

## Why It Matters

Vᵀ reveals the structure of the original data.

---

## Interview Takeaway

Vᵀ contains important input directions.

---

# What SVD Is Really Doing

SVD can be viewed as:

```text
Input Directions
      ↓
Vᵀ

Importance
      ↓
Σ

Output Directions
      ↓
U
```

---

Rather than viewing a matrix as:

```text
Millions Of Numbers
```

SVD views it as:

```text
Important Directions

+

Their Importance
```

---

# Geometric Interpretation Of SVD

One of the most important intuitions.

---

SVD performs three operations.

---

## Step 1

Rotate Data

```text
Vᵀ
```

---

## Step 2

Stretch Or Compress

```text
Σ
```

---

## Step 3

Rotate Again

```text
U
```

---

Therefore:

```text
SVD

=

Rotate

↓

Stretch

↓

Rotate
```

---

## Interview Takeaway

SVD can be interpreted as rotating, scaling, and rotating data again.

---

# Why Singular Values Matter

Suppose singular values are:

```text
500

200

100

2

0.5

0.01
```

---

Large values represent:

```text
Signal
```

---

Small values represent:

```text
Noise

Redundancy
```

---

This observation powers:

```text
PCA

Compression

Recommendation Systems
```

---

## Interview Takeaway

Large singular values usually correspond to important information.

---

# Rank And Singular Values

## Full Rank Matrix

All singular values are non-zero.

---

## Rank Deficient Matrix

Some singular values are:

```text
0
```

---

## Layman Explanation

Zero singular values indicate:

```text
Redundant Information
```

---

## Why It Matters

Rank can be determined directly from singular values.

---

## Interview Takeaway

Non-zero singular values determine matrix rank.

---

# Low Rank Approximation

## Definition

Low rank approximation keeps only the largest singular values.

---

## Layman Explanation

Suppose singular values are:

```text
100

50

20

1

0.01
```

---

Most information exists in:

```text
100

50

20
```

---

Discard:

```text
1

0.01
```

---

Result:

```text
Smaller Matrix

Nearly Same Information
```

---

## Why It Works

Real-world datasets often contain:

```text
Redundancy
```

---

Many features provide overlapping information.

---

## Interview Takeaway

Low rank approximation preserves important information while reducing complexity.

---

# SVD For Compression

## Definition

Compression reduces storage requirements while preserving information.

---

## Example

Image

```text
1000 × 1000 Pixels
```

---

Original:

```text
1 Million Values
```

---

After SVD:

Keep only:

```text
Top Singular Values
```

---

Result:

```text
Much Smaller Representation
```

---

## Applications

- Image Compression
- Embedding Compression
- LLM Compression

---

## Interview Takeaway

SVD enables compression by retaining only important information.

---

# SVD For Noise Removal

## Definition

Noise removal discards unimportant information.

---

## Layman Explanation

Small singular values often represent:

```text
Noise

Measurement Error

Random Variation
```

---

Removing them creates:

```text
Cleaner Data
```

---

## Example

Sensor data containing random fluctuations.

---

Keep:

```text
Large Singular Values
```

---

Discard:

```text
Tiny Singular Values
```

---

Result:

```text
Signal Preserved

Noise Reduced
```

---

## Interview Takeaway

SVD separates signal from noise.

---

# SVD And PCA

One of the most important interview questions.

---

## Traditional PCA

Compute:

```text
Covariance Matrix
```

---

Then perform:

```text
Eigendecomposition
```

---

## Modern PCA

Directly perform:

```text
SVD
```

on centered data.

---

## Why?

SVD is:

```text
More Stable

More Efficient

More General
```

---

## Interview Takeaway

Most PCA implementations internally use SVD.

---

# SVD In Recommendation Systems

## Problem

Netflix-like systems contain:

```text
User × Movie Matrix
```

---

Most entries are missing.

---

## Goal

Discover:

```text
Hidden Preferences
```

---

## Solution

SVD identifies:

```text
Latent Factors
```

such as:

```text
Comedy Preference

Action Preference

Drama Preference
```

---

## Interview Takeaway

SVD discovers hidden factors in recommendation systems.

---

# SVD In NLP

## Problem

Term-document matrices are enormous.

---

## Solution

Apply:

```text
SVD
```

---

Result:

```text
Latent Semantic Analysis (LSA)
```

---

## Benefit

Words with similar meanings become closer in latent space.

---

## Interview Takeaway

SVD helps discover hidden semantic relationships.

---

# SVD In Embeddings

Embeddings often contain:

```text
Hundreds

Or Thousands

Of Dimensions
```

---

SVD can reduce dimensionality while preserving meaning.

---

## Applications

- Semantic Search
- Vector Databases
- Recommendation Systems

---

## Interview Takeaway

SVD reduces embedding dimensions while preserving information.

---

# SVD In Deep Learning

## Model Compression

Large weight matrices contain redundancy.

---

SVD approximates them using:

```text
Lower Rank Matrices
```

---

Result:

```text
Smaller Models

Faster Inference
```

---

# LoRA Intuition

LoRA assumes:

```text
Updates Are Low Rank
```

---

Instead of storing huge matrices:

```text
Store Small Low-Rank Updates
```

---

This idea comes directly from SVD intuition.

---

## Interview Takeaway

SVD forms the mathematical foundation of many model compression techniques.

---

# SVD vs Eigendecomposition

| Topic | Eigendecomposition | SVD |
|----------|----------|----------|
| Matrix Type | Square Only | Any Matrix |
| Stability | Lower | Higher |
| ML Usage | PCA Foundation | Everywhere |
| Noise Handling | Limited | Excellent |
| Compression | Limited | Excellent |
