# Module 4: Linear Algebra for Machine Learning

# Chapter 8: Matrix Factorization & Recommendation Systems

One of the biggest challenges in Machine Learning is:

> "How do we recommend the right item to the right user?"

Examples:

```text
Netflix Movies

Amazon Products

YouTube Videos

Spotify Songs

Instagram Reels
```

Recommendation systems solve this problem.

The mathematical foundation behind many recommendation systems is:

```text
Matrix Factorization
```

Matrix factorization uses concepts from:

```text
Linear Algebra

SVD

Low Rank Approximation
```

to discover hidden patterns in user behavior.

---

# Why Recommendation Systems Exist

Suppose Netflix has:

```text
100 Million Users

10,000 Movies
```

---

Creating recommendations manually is impossible.

---

Instead we want to discover:

```text
Hidden Preferences
```

automatically.

---

Examples:

```text
Comedy Lovers

Action Lovers

Sci-Fi Lovers

Romance Lovers
```

---

These hidden preferences are called:

```text
Latent Factors
```

---

## Interview Takeaway

Recommendation systems attempt to discover hidden user preferences.

---

# User-Item Matrix

## Definition

A User-Item Matrix stores interactions between users and items.

---

## Example

| User | Batman | Titanic | Avengers |
|--------|--------|--------|--------|
| User A | 5 | ? | 5 |
| User B | ? | 5 | 1 |
| User C | 4 | ? | 5 |

---

Rows:

```text
Users
```

---

Columns:

```text
Items
```

---

Values:

```text
Ratings

Clicks

Purchases

Watch Time
```

---

## Problem

Most entries are missing.

---

This creates a:

```text
Sparse Matrix
```

---

## Interview Takeaway

Recommendation systems begin with a sparse User-Item Matrix.

---

# Why Traditional Approaches Fail

Suppose:

```text
10 Million Users

100 Thousand Movies
```

---

User-Item Matrix Size:

```text
10 Million × 100 Thousand
```

---

Most entries are:

```text
Unknown
```

---

Traditional similarity methods struggle.

---

We need a way to discover:

```text
Hidden Structure
```

inside sparse data.

---

# Matrix Factorization

## Definition

Matrix Factorization decomposes a large User-Item Matrix into smaller matrices.

---

## Formula

```text
User-Item Matrix

≈

User Factors

×

Item Factors
```

---

## Layman Explanation

Instead of remembering:

```text
Every User-Movie Rating
```

we learn:

```text
User Preferences

Movie Characteristics
```

---

These smaller representations can recreate the original matrix.

---

## Why It Matters

The hidden factors are often more informative than raw ratings.

---

## Interview Takeaway

Matrix factorization learns hidden patterns from user behavior.

---

# Latent Factors

## Definition

Latent factors are hidden characteristics discovered automatically from data.

---

## Layman Explanation

Latent means:

```text
Hidden
```

---

The system never explicitly sees:

```text
Comedy Preference

Action Preference

Romance Preference
```

---

Instead it discovers them from ratings.

---

## Example

User A

```text
Likes:

Batman

Avengers

Iron Man
```

---

System may learn:

```text
Action Preference = High
```

---

without anyone explicitly providing that feature.

---

## Why It Matters

Latent factors power recommendation systems.

---

## Interview Takeaway

Latent factors represent hidden user preferences and item characteristics.

---

# Matrix Factorization Intuition

Suppose:

```text
1000 Movies
```

---

Each movie may be represented using:

```text
Action Score

Comedy Score

Romance Score

Drama Score
```

---

Instead of storing:

```text
1000 Movie Ratings
```

for each user.

---

Store:

```text
A Few Latent Preferences
```

---

This dramatically reduces complexity.

---

## Interview Takeaway

Matrix factorization compresses behavior into a few hidden factors.

---

# Relationship To SVD

One of the most common interview questions.

---

SVD decomposes:

```text
A = UΣVᵀ
```

---

Matrix factorization uses a similar idea.

---

Instead of:

```text
Raw Ratings
```

discover:

```text
Hidden User Factors

Hidden Item Factors
```

---

SVD provides the mathematical intuition behind this process.

---

## Interview Takeaway

Matrix factorization is heavily inspired by SVD and low-rank approximation.

---

# Low Rank Approximation

## Definition

Low Rank Approximation keeps only the most important latent factors.

---

## Layman Explanation

Suppose:

```text
1000 Movies
```

---

Actually depend on:

```text
20 Hidden Preferences
```

---

Instead of storing all information:

```text
Store Only Important Factors
```

---

## Benefits

```text
Compression

Noise Reduction

Generalization
```

---

## Interview Takeaway

Recommendation systems often use low-rank approximations to reduce complexity.

---

# Collaborative Filtering

## Definition

Collaborative Filtering recommends items using behavior of similar users.

---

## Layman Explanation

The idea is:

```text
People Similar To You

May Like Similar Things
```

---

## Example

User A likes:

```text
Batman

Avengers

Iron Man
```

---

User B likes:

```text
Batman

Avengers
```

---

System predicts:

```text
User B May Like Iron Man
```

---

## Why It Matters

Collaborative Filtering became one of the most successful recommendation techniques.

---

## Interview Takeaway

Collaborative Filtering learns recommendations from user behavior patterns.

---

# User-Based vs Item-Based Collaborative Filtering

## User-Based

Find similar users.

---

Example

```text
Users With Similar Viewing History
```

---

Recommend what similar users enjoyed.

---

## Item-Based

Find similar items.

---

Example

```text
People Who Watched Batman

Also Watched Avengers
```

---

Recommend similar items.

---

## Interview Takeaway

User-based focuses on similar users.

Item-based focuses on similar items.

---

# Alternating Least Squares (ALS)

## Definition

ALS is a popular algorithm used to learn user and item latent factors.

---

## Layman Explanation

ALS alternates between:

```text
Learning User Factors
```

and

```text
Learning Item Factors
```

until ratings are reconstructed accurately.

---

## Why It Matters

ALS is widely used in large-scale recommendation systems.

---

## Applications

- Netflix
- Spark MLlib
- E-Commerce Recommendations

---

## Interview Takeaway

ALS learns latent factors by repeatedly improving user and item representations.

---
