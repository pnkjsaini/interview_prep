# Module 4: Linear Algebra for Machine Learning

# Chapter 7: Principal Component Analysis (PCA)

Modern datasets often contain:

```text
Hundreds Of Features

Thousands Of Features

Millions Of Features
```

Not all features contribute useful information.

Many features contain:

```text
Noise

Redundancy

Duplicate Information
```

PCA was created to solve this problem.

Its goal is:

```text
Keep Maximum Information

Using Fewer Dimensions
```

PCA is one of the most important dimensionality reduction techniques in Machine Learning.

---

# Why PCA Exists

Suppose we have a housing dataset:

```text
Area (sq ft)

Area (sq meter)

Number Of Rooms

Property Size

Built Up Area
```

Several features describe:

```text
The Same Underlying Information
```

---

Result:

```text
Redundancy

Higher Computation

Multicollinearity
```

---

PCA tries to find:

```text
A Smaller Set Of Features

Containing Most Information
```

---

## Interview Takeaway

PCA reduces dimensionality while preserving maximum information.

---

# Curse Of Dimensionality

## Definition

As the number of features increases, data becomes sparse and learning becomes harder.

---

## Layman Explanation

Imagine:

```text
2 Features
```

Easy to visualize.

---

Now imagine:

```text
1000 Features
```

Distances become less meaningful.

Patterns become harder to discover.

---

## Problems Caused

```text
Overfitting

Slower Training

Storage Cost

Noise
```

---

## Why PCA Helps

PCA reduces feature count while preserving important structure.

---

## Interview Takeaway

PCA helps combat the curse of dimensionality.

---

# Core Idea Of PCA

PCA asks:

> "Which directions contain the most information?"

---

Instead of keeping original features:

```text
Age

Income

Experience

Education
```

---

PCA creates:

```text
Principal Component 1

Principal Component 2

Principal Component 3
```

---

These components are:

```text
New Directions

Created From Existing Features
```

---

## Interview Takeaway

PCA creates new features that capture maximum information.

---

# Variance

## Definition

Variance measures how much values spread around their mean.

---

## Layman Explanation

Higher variance often indicates:

```text
More Information
```

---

Low variance often indicates:

```text
Less Information
```

---

## Example

Feature A

```text
10

10

10

10
```

Variance:

```text
0
```

Little information.

---

Feature B

```text
5

50

100

150
```

High variance.

More information.

---

## Why PCA Cares

PCA assumes:

```text
Important Directions

↓

High Variance
```

---

## Interview Takeaway

PCA searches for directions with maximum variance.

---

# Covariance

## Definition

Covariance measures how two variables change together.

---

## Layman Explanation

Covariance answers:

> "Do these features move together?"

---

## Example

Income ↑

Spending ↑

---

Positive covariance.

---

## Why PCA Cares

PCA begins by analyzing relationships among features.

---

## Interview Takeaway

Covariance reveals relationships between features.

---

# Covariance Matrix

## Definition

A covariance matrix contains covariances between all feature pairs.

---

## Example

Features:

```text
Age

Income

Spending
```

---

Covariance Matrix:

```text
Relationship Of Every Feature
With Every Other Feature
```

---

## Why It Matters

The covariance matrix is the starting point of PCA.

---

## Interview Takeaway

PCA extracts important directions from the covariance matrix.

---

# Principal Components

## Definition

Principal Components are new directions that capture maximum variance.

---

## Layman Explanation

Imagine shining a flashlight onto a cloud of points.

Some viewing angles reveal:

```text
Most Information
```

---

Some reveal:

```text
Very Little Information
```

---

PCA finds the most informative viewing directions.

---

## Principal Component 1 (PC1)

Captures:

```text
Maximum Variance
```

---

## Principal Component 2 (PC2)

Captures:

```text
Second Highest Variance
```

---

Must be:

```text
Orthogonal To PC1
```

---

## Principal Component 3

Captures:

```text
Third Highest Variance
```

---

and so on.

---

## Interview Takeaway

Principal components are directions containing the most information.

---

# PCA Using Eigenvectors

One of the most common interview questions.

---

## Step 1

Compute:

```text
Covariance Matrix
```

---

## Step 2

Perform:

```text
Eigendecomposition
```

---

## Step 3

Find:

```text
Eigenvectors

Eigenvalues
```

---

## Step 4

Sort eigenvalues from largest to smallest.

---

## Step 5

Keep the corresponding eigenvectors.

---

Result:

```text
Principal Components
```

---

## Why?

Eigenvectors represent:

```text
Important Directions
```

---

Eigenvalues represent:

```text
Importance Of Directions
```

---

## Interview Takeaway

PCA uses eigenvectors of the covariance matrix as principal components.

---

# PCA Using SVD

Modern PCA implementations rarely use eigendecomposition directly.

---

Instead:

```text
Centered Data
      ↓
SVD
      ↓
Principal Components
```

---

## Why?

SVD is:

```text
More Stable

More Efficient

Works Better On Large Datasets
```

---

## Interview Takeaway

Most modern PCA implementations use SVD internally.

---

# Explained Variance

## Definition

Explained variance measures how much information a principal component captures.

---

## Layman Explanation

Suppose:

```text
Total Information = 100%
```

---

PC1 captures:

```text
70%
```

---

PC2 captures:

```text
20%
```

---

PC3 captures:

```text
5%
```

---

Remaining:

```text
5%
```

---

## Why It Matters

Explained variance helps decide:

```text
How Many Components To Keep
```

---

## Interview Takeaway

Explained variance measures information captured by each component.

---

# Choosing Number Of Components

## Example

Explained Variance

```text
PC1 → 60%

PC2 → 25%

PC3 → 10%

PC4 → 3%

PC5 → 2%
```

---

Keeping:

```text
PC1 + PC2
```

captures:

```text
85%
```

of information.

---

Many practitioners keep enough components to explain:

```text
90%–95%
```

of variance.

---

## Interview Takeaway

Keep components that capture most of the variance.

---

# PCA Transformation

Original Features

```text
100 Features
```

---

PCA

↓

Principal Components

```text
10 Components
```

---

Result:

```text
Less Noise

Lower Dimension

Faster Training
```

---

## Interview Takeaway

PCA converts many correlated features into fewer informative components.

---

# Benefits Of PCA

## Dimensionality Reduction

Fewer features.

---

## Faster Training

Smaller datasets.

---

## Reduced Noise

Less irrelevant information.

---

## Reduced Multicollinearity

Removes redundant features.

---

## Better Visualization

High-dimensional data becomes easier to visualize.

---

# Limitations Of PCA

## Limitation 1

Components lose interpretability.

---

Original Feature:

```text
Income
```

Easy to understand.

---

Principal Component:

```text
Combination Of Many Features
```

Harder to interpret.

---

## Limitation 2

Assumes important information exists in:

```text
High Variance Directions
```

---

Not always true.

---

## Limitation 3

Captures only:

```text
Linear Relationships
```

---

Cannot model complex nonlinear structures.

---

## Interview Takeaway

PCA improves efficiency but may reduce interpretability.

---

# PCA Workflow

```text
Original Features
        ↓
Center Data
        ↓
Covariance Matrix
        ↓
Eigenvectors / SVD
        ↓
Principal Components
        ↓
Keep Important Components
        ↓
Reduced Dataset
```

---

# PCA vs Feature Selection

## Feature Selection

Keeps:

```text
Original Features
```

---

## PCA

Creates:

```text
New Features
```

called principal components.

---

## Interview Takeaway

PCA creates new features rather than selecting existing ones.

---

# Common Interview Questions

---

## Question

Why does PCA use eigenvectors?

### Answer

```text
Because eigenvectors identify directions of maximum variance.
```

---

## Question

Why does PCA reduce dimensionality?

### Answer

```text
Because most information often exists in a small number of directions.
```

---

## Question

Why does PCA use variance?

### Answer

```text
Because high variance directions usually contain more information.
```

---

## Question

Why is SVD used in PCA?

### Answer

```text
Because SVD is numerically more stable and efficient.
```

---

## Question

What is explained variance?

### Answer

```text
The proportion of information captured by a principal component.
```
