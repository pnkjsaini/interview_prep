# Module 4: Linear Algebra for Machine Learning

# Chapter 1: Vectors & Geometry

Linear Algebra is the language of Machine Learning.

Almost everything in ML can ultimately be represented as:

```text
Vectors

Matrices

Transformations
```

Examples:

```text
Word Embeddings

Image Pixels

Feature Vectors

Neural Network Weights

Attention Mechanisms
```

are all built on linear algebra.

This chapter introduces the geometric intuition required to understand modern machine learning.

---

# Scalar

## Definition

A scalar is a single numerical value.

---

## Layman Explanation

A scalar only represents magnitude.

It has:

```text
Size
```

but no direction.

---

## Example

Temperature:

```text
35°C
```

---

Salary:

```text
₹15 Lakh
```

---

Age:

```text
30 Years
```

---

## Why It Matters

Scalars are the building blocks of vectors and matrices.

---

## Limitation

Scalars cannot represent direction or relationships.

---

## Interview Takeaway

A scalar is a single numerical value.

---

# Vector

## Definition

A vector is an ordered collection of numbers representing magnitude and direction.

---

## Layman Explanation

A vector can be thought of as:

```text
An Arrow
```

pointing in a particular direction.

---

## Example

Customer Features:

```text
[Age, Income, Spending]

=

[30, 1000000, 50000]
```

---

Word Embedding:

```text
[0.12, -0.81, 0.44, ...]
```

---

## Why It Matters

Almost all ML data is represented as vectors.

---

## Limitation

High-dimensional vectors are difficult to visualize.

---

## Interview Takeaway

Vectors represent data points in machine learning.

---

# Vector Magnitude (Norm)

## Definition

The norm measures the length of a vector.

---

## Layman Explanation

Magnitude answers:

> "How large is this vector?"

---

## Formula (L2 Norm)

```text
||v||

=

√(x₁² + x₂² + ... + xₙ²)
```

---

## Example

Vector:

```text
[3,4]
```

Magnitude:

```text
√(3² + 4²)

=

5
```

---

## Why It Matters

Used in:

- Distance Calculations
- Similarity Search
- Regularization
- Embeddings

---

## Limitation

Magnitude alone does not tell direction.

---

## Interview Takeaway

The norm measures the length of a vector.

---

# Unit Vector

## Definition

A unit vector is a vector whose magnitude equals 1.

---

## Layman Explanation

A unit vector preserves:

```text
Direction
```

while removing:

```text
Magnitude
```

---

## Example

Vector:

```text
[3,4]
```

Magnitude:

```text
5
```

Unit Vector:

```text
[3/5 , 4/5]
```

---

## Why It Matters

Used in:

- Cosine Similarity
- Normalization
- Embeddings

---

## Limitation

Magnitude information is lost.

---

## Interview Takeaway

A unit vector preserves direction while having length 1.

---

# Distance Metrics

Distance metrics measure how far two vectors are from each other.

---

# Euclidean Distance

## Definition

Euclidean Distance is the straight-line distance between two points.

---

## Formula

```text
Distance

=

√Σ(xᵢ-yᵢ)²
```

---

## Layman Explanation

Imagine drawing a straight line between two locations.

That shortest path is Euclidean distance.

---

## Example

Points:

```text
A = [1,1]

B = [4,5]
```

Distance:

```text
5
```

---

## Why It Matters

Used in:

- KNN
- Clustering
- Similarity Search

---

## Limitation

Sensitive to feature scale.

---

## Interview Takeaway

Euclidean distance measures straight-line distance.

---

# Manhattan Distance

## Definition

Manhattan Distance measures travel distance along grid lines.

---

## Formula

```text
Distance

=

Σ|xᵢ-yᵢ|
```

---

## Layman Explanation

Imagine moving through city streets.

You cannot move diagonally.

You must move:

```text
Up

Down

Left

Right
```

---

## Example

Points:

```text
A = [1,1]

B = [4,5]
```

Distance:

```text
|4-1| + |5-1|

=

7
```

---

## Why It Matters

Often more robust in high-dimensional spaces.

---

## Limitation

Ignores geometric straight-line distance.

---

## Interview Takeaway

Manhattan distance measures grid-based movement distance.

---

# Minkowski Distance

## Definition

Minkowski Distance generalizes both Euclidean and Manhattan distance.

---

## Formula

```text
Distance

=

(Σ|xᵢ-yᵢ|ᵖ)^(1/p)
```

---

## Layman Explanation

Changing:

```text
p
```

changes the distance behavior.

---

```text
p = 1
```

Manhattan Distance

---

```text
p = 2
```

Euclidean Distance

---

## Why It Matters

Provides a flexible distance framework.

---

## Limitation

Choosing p can be difficult.

---

## Interview Takeaway

Minkowski Distance generalizes multiple distance metrics.

---

# Dot Product

## Definition

The dot product measures the alignment between two vectors.

---

## Formula

```text
A · B

=

Σ(AᵢBᵢ)
```

---

## Layman Explanation

Dot product answers:

> "How much are these vectors pointing in the same direction?"

---

## Example

Vectors:

```text
A = [1,2]

B = [3,4]
```

Dot Product:

```text
(1×3)

+

(2×4)

=

11
```

---

## Interpretation

Large Positive Value:

```text
Similar Direction
```

---

Zero:

```text
Perpendicular
```

---

Negative:

```text
Opposite Direction
```

---

## Why It Matters

Used everywhere:

- Neural Networks
- Attention Mechanisms
- Embeddings
- Recommendation Systems

---

## Interview Takeaway

Dot product measures directional alignment.

---

# Cosine Similarity

## Definition

Cosine Similarity measures similarity based on angle rather than magnitude.

---

## Formula

```text
Cosine Similarity

=

(A · B)

/

(||A|| ||B||)
```

---

## Layman Explanation

Instead of comparing sizes:

```text
Compare Directions
```

---

## Example

Sentence Embeddings:

```text
"Car"

"Automobile"
```

may point in similar directions.

Result:

```text
High Cosine Similarity
```

---

## Interpretation

```text
1
```

Identical Direction

---

```text
0
```

Unrelated

---

```text
-1
```

Opposite Direction

---

## Why It Matters

Used extensively in:

- Semantic Search
- Vector Databases
- Recommendation Systems
- RAG Systems

---

## Limitation

Ignores magnitude.

---

## Interview Takeaway

Cosine Similarity compares vector direction rather than size.

---

# Orthogonality

## Definition

Two vectors are orthogonal if they are perpendicular.

---

## Formula

```text
A · B = 0
```

---

## Layman Explanation

Orthogonal vectors share:

```text
No Information
```

about one another.

---

## Example

Vectors:

```text
[1,0]

[0,1]
```

Dot Product:

```text
0
```

---

## Why It Matters

Used in:

- PCA
- SVD
- Feature Engineering

---

## Limitation

Perfect orthogonality is uncommon in real-world data.

---

## Interview Takeaway

Orthogonal vectors are independent directions in space.

---

# Projection

## Definition

Projection measures how much one vector lies in the direction of another.

---

## Layman Explanation

Projection answers:

> "How much of vector A points toward vector B?"

---

## Example

Suppose:

```text
Employee Performance
```

depends on:

```text
Experience

Education
```

Projection helps isolate the contribution of one direction.

---

## Why It Matters

Projection appears in:

- PCA
- Regression
- Embeddings
- Attention

---

## Limitation

Interpretation becomes harder in high dimensions.

---

## Interview Takeaway

Projection measures how much one vector aligns with another.

---

# Vector Space Intuition

## Definition

A vector space is a collection of vectors that can be combined through addition and scaling.

---

## Layman Explanation

Think of a vector space as:

```text
The World In Which Vectors Live
```

---

## Example

Word Embedding Space

```text
King

Queen

Man

Woman
```

all exist inside the same vector space.

---

## Why It Matters

Most ML models operate within high-dimensional vector spaces.

---

## Interview Takeaway

Machine learning largely involves learning meaningful vector spaces.
