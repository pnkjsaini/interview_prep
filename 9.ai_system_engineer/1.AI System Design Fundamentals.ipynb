# What is AI System Design?

AI System Design is the process of building an AI application that satisfies:

```text
Business Requirements

↓

Technical Requirements

↓

Operational Requirements
```

Instead of asking

```text
Which LLM should I use?
```

we ask

```text
How should the entire system work?
```

---

## Interview Takeaway

System Design is about building the complete ecosystem around the AI model.

---

# AI Model ≠ AI System

One of the biggest misconceptions.

People often think:

```text
LLM

=

AI Application
```

Wrong.

Production AI looks like:

```text
Users

↓

API Gateway

↓

Authentication

↓

Business Logic

↓

Memory

↓

Retriever

↓

LLM

↓

Guardrails

↓

Observability

↓

Monitoring
```

Notice:

```text
LLM

Is Just One Component
```

---

## Interview Takeaway

The AI model is only one service inside a much larger architecture.

# Step 1 — Requirement Gathering

Never jump into architecture immediately.

Always clarify requirements.

---

Example

Design:

```text
Enterprise RAG System
```

Questions:

```text
Who are the users?

Number of users?

Expected latency?

Document size?

Real-time updates?

Security requirements?

Budget?
```

---

## Interview Takeaway

Requirement gathering prevents solving the wrong problem.

---

# Functional Requirements

These describe:

```text
What The System Should Do
```

Example

```text
Upload Documents

Semantic Search

Question Answering

Citations

Conversation History
```

---

## Interview Takeaway

Functional requirements describe system capabilities.

---

# Non-Functional Requirements

Describe:

```text
How Well

The System

Should Perform
```

Examples

```text
Latency

Availability

Security

Scalability

Reliability

Cost
```

---

## Interview Takeaway

Non-functional requirements often drive architecture decisions.

---

# Latency

Latency means:

```text
Time

To Produce

One Response
```

Example

```text
User

↓

Response

800 ms
```

Lower latency

↓

Better user experience.

---

## Interview Takeaway

Latency measures response time.

---

# Throughput

Throughput measures:

```text
How Many Requests

Per Second
```

Example

```text
100 Requests

Per Second
```

High throughput

does not necessarily imply

low latency.

---

## Interview Takeaway

Throughput measures system capacity.

---

# Latency vs Throughput

| Latency | Throughput |
|----------|------------|
| Time per request | Requests per second |
| User experience | Overall capacity |
| Measured in ms | Measured in RPS/QPS |

---

# Scalability

Question:

```text
Can The System

Handle Growth?
```

Growth may come from:

```text
Users

Documents

Models

Requests
```

---

## Interview Takeaway

Scalability is the ability to handle increasing workload.

---

# Vertical Scaling

Increase resources.

```text
4 CPU

↓

16 CPU
```

Advantages

```text
Simple
```

Problems

```text
Upper Hardware Limit
```

---

## Interview Takeaway

Vertical scaling increases the power of one machine.

---

# Horizontal Scaling

Instead of:

```text
One Server
```

Deploy:

```text
Many Servers
```

Behind:

```text
Load Balancer
```

---

Advantages

```text
Fault Tolerant

Highly Scalable
```

---

## Interview Takeaway

Horizontal scaling increases the number of machines.

---

# Availability

Question

```text
Can Users

Access

The System?
```

Measured as:

```text
99%

99.9%

99.99%

99.999%
```

Higher availability

↓

Higher cost.

---

## Interview Takeaway

Availability measures system uptime.

---

# Reliability

Reliability asks:

```text
Can The System

Continue Working

Correctly?
```

Example

```text
Retry

Fallback

Recovery
```

---

## Interview Takeaway

Reliable systems continue operating despite failures.

---

# Fault Tolerance

Suppose

```text
One Server

Fails
```

System should continue.

Achieved using:

```text
Redundancy

Replication

Failover
```

---

## Interview Takeaway

Fault tolerance minimizes service interruption.

---

# Stateless Services

Every request is independent.

Example

```text
Request A

↓

Complete

↓

Forget
```

Advantages

```text
Easy Scaling
```

---

## Interview Takeaway

Stateless services simplify horizontal scaling.

---

# Stateful Services

Need memory.

Example

```text
Conversation

Shopping Cart

Agent State
```

Must preserve information.

---

## Interview Takeaway

Stateful services require state management.

---

# Monolith vs Microservices

Monolith

```text
Everything

One Application
```

---

Advantages

```text
Simple Deployment
```

Problems

```text
Hard To Scale

Hard To Maintain
```

---

Microservices

```text
Retriever

↓

LLM

↓

Memory

↓

Authentication

↓

Monitoring
```

Independent services.

---

## Interview Takeaway

Enterprise AI systems usually adopt microservice architectures.

---

# API Gateway

Every request enters through:

```text
Gateway
```

Responsibilities

```text
Authentication

Routing

Rate Limiting

Logging

Load Balancing
```

---

## Interview Takeaway

API Gateway is the entry point into the system.

---

# Load Balancer

Distributes requests.

Instead of

```text
All Users

↓

One Server
```

Use

```text
Users

↓

Load Balancer

↓

Server A

Server B

Server C
```

---

## Interview Takeaway

Load balancers distribute traffic across multiple instances.

---

# Caching

Question

```text
Should We Compute

Again?
```

If answer already exists,

return from cache.

---

Example

```text
LLM Response

↓

Cache

↓

Future Requests
```

---

Benefits

```text
Lower Latency

Lower Cost
```

---

## Interview Takeaway

Caching avoids repeated computation.

---

# Queues

Suppose

```text
Generate

500 Page Report
```

Instead of blocking,

place task into:

```text
Queue
```

Worker processes later.

Examples

```text
Kafka

RabbitMQ

Azure Service Bus
```

---

## Interview Takeaway

Queues decouple request submission from execution.

---

# Database Selection

Different data

↓

Different databases.

Structured Data

↓

SQL

---

Embeddings

↓

Vector Database

---

Relationships

↓

Graph Database

---

Temporary State

↓

Redis

---

## Interview Takeaway

Choose storage based on access patterns rather than popularity.

---

# Consistency

Question

```text
Should Everyone

See

Exactly Same Data?
```

Strong consistency

↓

Latest data always visible.

Eventual consistency

↓

Data synchronizes over time.

---

## Interview Takeaway

Consistency often trades off against availability and scalability.

---

# CAP Theorem

Distributed systems cannot simultaneously guarantee:

```text
Consistency

Availability

Partition Tolerance
```

You can fully guarantee only two.

Most cloud systems choose:

```text
Availability

+

Partition Tolerance
```

while accepting eventual consistency in some scenarios.

---

## Interview Takeaway

CAP theorem explains trade-offs in distributed system design.

---

# Observability

Question

```text
Can We Understand

What Happened?
```

Observability consists of:

```text
Logs

Metrics

Traces
```

---

Logs

```text
Events
```

Metrics

```text
Numbers
```

Traces

```text
Entire Request Path
```

---

## Interview Takeaway

Observability makes distributed systems debuggable.

---

# Monitoring

Continuously track:

```text
Latency

Availability

Failures

GPU Usage

Token Usage

Cost

Throughput
```

---

## Interview Takeaway

Monitoring detects issues before users notice them.

---

# Security

Every AI system needs:

```text
Authentication

Authorization

Encryption

Guardrails

Audit Logs

Secrets Management
```

Never rely solely on the LLM.

---

## Interview Takeaway

Security is a system property, not a model feature.

---

# Cost

Every architectural decision has a cost.

Examples

```text
Larger LLM

↓

Higher Cost

--------------------------------

Caching

↓

Lower Cost

--------------------------------

Smaller Model

↓

Lower Latency
```

System Design is about:

```text
Trade-offs
```

---

## Interview Takeaway

There is no perfect architecture—only trade-offs.

---

# Trade-off Thinking

Every interview expects you to justify decisions.

Examples

```text
Latency

vs

Accuracy

--------------------------------

Cost

vs

Quality

--------------------------------

Availability

vs

Consistency

--------------------------------

Single Agent

vs

Multi-Agent
```

---

## Interview Takeaway

Good architects justify decisions instead of searching for perfect solutions.

# Common Interview Questions

---

## Question

What is System Design?

### Answer

```text
Designing a complete software architecture that satisfies functional and non-functional requirements.
```

---

## Question

Latency vs Throughput?

### Answer

```text
Latency measures response time.

Throughput measures requests processed per unit time.
```

---

## Question

Horizontal vs Vertical Scaling?

### Answer

```text
Vertical scaling upgrades a machine.

Horizontal scaling adds more machines.
```

---

## Question

Why use an API Gateway?

### Answer

```text
To centralize routing, authentication, rate limiting, and request management.
```

---

## Question

Why are microservices preferred?

### Answer

```text
They allow independent deployment, scaling, and maintenance of system components.
```

---

## Question

Why use caching?

### Answer

```text
To reduce latency, cost, and repeated computation.
```

---

## Question

What are the three pillars of observability?

### Answer

```text
Logs

Metrics

Traces
```

---

## Question

Why are trade-offs important?

### Answer

```text
Every architecture optimizes some qualities while sacrificing others.
```
