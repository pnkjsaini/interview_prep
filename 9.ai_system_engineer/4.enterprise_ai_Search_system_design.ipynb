# What is AI Search?

AI Search is the process of retrieving the most relevant information for a user's intent.

Instead of asking

```text
Does the document contain the exact words?
```

AI Search asks

```text
What is the user trying to find?
```

---

## Interview Takeaway

Modern search optimizes for user intent rather than exact text matching.

# Common Interview Questions

---

## Question

Why is Hybrid Search preferred?

### Answer

```text
It combines the strengths of lexical and semantic retrieval, improving both recall and precision.
```

---

## Question

Why rerank results?

### Answer

```text
Retrievers maximize recall, while rerankers improve the final ordering of results.
```

---

## Question

Why use BM25 if we already have embeddings?

### Answer

```text
BM25 is excellent for exact keyword matching, identifiers, product names, and rare terms that embedding models may miss.
```

---

## Question

Why perform query expansion?

### Answer

```text
To improve recall by searching for semantically related terms.
```

---

## Question

What metrics evaluate search quality?

### Answer

```text
Precision@K, Recall@K, MRR, NDCG, MAP, CTR, and search latency.
```

---

## Question

Where should security filtering happen?

### Answer

```text
During retrieval, before ranking or sending context to the LLM.
```
