# Common Interview Questions

---

## Question

Why use KV Cache?

### Answer

```text
To reuse previous attention computations and significantly reduce autoregressive inference latency.
```

---

## Question

Continuous vs Dynamic Batching?

### Answer

```text
Dynamic batching forms batches before execution.

Continuous batching allows new requests to join an already executing batch, improving GPU utilization.
```

---

## Question

Why quantize models?

### Answer

```text
To reduce GPU memory usage and increase inference speed while maintaining acceptable accuracy.
```

---

## Question

Data vs Tensor Parallelism?

### Answer

```text
Data parallelism serves more requests by replicating the model.

Tensor parallelism splits a single large model across multiple GPUs.
```

---

## Question

What metrics matter most?

### Answer

```text
TTFT, Tokens Per Second, GPU Utilization, Latency, Throughput, Queue Length, and Cost.
```
