# What is a Recommendation System?

A Recommendation System predicts which items are most relevant to a user.

Instead of matching a query,

it predicts:

```text
Probability

User

Will Like

An Item
```

---

## Interview Takeaway

Recommendation systems predict user preferences rather than retrieving matching documents.

---

# Recommendation Pipeline

Production systems generally follow two stages.

```text
Candidate Generation

↓

Ranking

↓

Business Rules

↓

Final Recommendation
```

This architecture allows recommendations over millions of items with low latency.

---

# Why Two Stages?

Suppose Amazon has

```text
500 Million Products
```

Should we rank all of them?

No.

Instead

```text
500 Million

↓

500 Candidates

↓

Top 20
```

---

## Interview Takeaway

Candidate generation reduces the search space before expensive ranking.

# Collaborative Filtering

One of the oldest recommendation techniques.

Idea

```text
Users

With Similar Behavior

Like Similar Items
```

Example

```text
Alice

Likes A B C

Bob

Likes A B

↓

Recommend C
```

---

```text
Simple

Effective

Personalized
```

---

Limitations

```text
Cold Start

Sparse Data
```

---

## Interview Takeaway

---

# Common Interview Questions

---

## Question

Why use Candidate Generation?

### Answer

```text
Ranking millions of items is computationally infeasible, so candidate generation narrows the search space.
```

---

## Question

Collaborative vs Content-Based Filtering?

### Answer

```text
Collaborative filtering uses user interactions, while content-based filtering uses item attributes.
```

---

## Question

What is the Cold Start problem?

### Answer

```text
Making recommendations for new users or new items with little or no historical interaction data.
```

---

## Question

Why use a Feature Store?

### Answer

```text
To provide consistent, reusable features for both model training and online inference.
```

---

## Question

Why apply Business Rules after ranking?

### Answer

```text
Business constraints such as inventory, compliance, and promotions should refine the ranked list before serving it to users.
```
