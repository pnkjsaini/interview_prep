# Common Interview Questions

---

## Question

Why are stateless services preferred?

### Answer

```text
Because they can be replicated and scaled horizontally without sharing local state.
```

---

## Question

Why use a circuit breaker?

### Answer

```text
To stop repeated calls to a failing service and prevent cascading failures.
```

---

## Question

Semantic Cache vs Traditional Cache?

### Answer

```text
Traditional caches require exact matches, while semantic caches reuse responses for semantically similar requests.
```

---

## Question

Why use queues?

### Answer

```text
Queues decouple request submission from processing, improving scalability and reliability.
```

---

## Question

What should be monitored in an AI platform?

### Answer

```text
Infrastructure metrics, AI metrics, business metrics, logs, traces, and costs.
```
