# A System Design Interview Framework

Every AI system design interview should follow the same structured approach.

```text
1. Clarify Requirements

↓

2. Estimate Scale

↓

3. Design High-Level Architecture

↓

4. Design Individual Components

↓

5. Database & Storage Choices

↓

6. Scalability

↓

7. Reliability

↓

8. Security

↓

9. Monitoring

↓

10. Trade-offs
```

---

## Interview Takeaway

Never jump directly into architecture. Always clarify requirements first.

---

# Step 1 — Requirement Gathering

Suppose the interviewer asks

```text
Design an Enterprise RAG System.
```

Start asking questions.

Functional

```text
Who are the users?

Document types?

Conversation support?

Source citations?

Authentication?

Real-time updates?
```

Non-functional

```text
Latency

Availability

Concurrent users

Security

Budget
```

---

## Interview Takeaway

Good system design begins with understanding the problem.

---

# Step 2 — Capacity Estimation

Estimate the workload.

Example

```text
100,000 Users

10,000 Concurrent Users

5 Requests/Minute

2 KB Average Prompt
```

Estimate

```text
Requests Per Second

Storage

Bandwidth

GPU Count

Database Size
```

---

## Interview Takeaway

Rough estimates justify infrastructure decisions.

---

# Step 3 — High-Level Architecture

Design the major components first.

Example

```text
Users

↓

API Gateway

↓

Authentication

↓

Application Layer

↓

AI Components

↓

Storage

↓

Monitoring
```

Do not immediately discuss implementation details.

# Whiteboard Strategy

When drawing architecture

Always move

```text
Left

↓

Right
```

Example

```text
Users

↓

Gateway

↓

Authentication

↓

Business Logic

↓

Retriever

↓

LLM

↓

Storage

↓

Monitoring
```

Avoid jumping randomly between components.

---

# Communication Strategy

During interviews

Explain

```text
Why

Before

How
```

Instead of saying

```text
Use Redis
```

say

```text
I'll introduce Redis here to cache embeddings and frequent responses, reducing latency and inference cost.
```

# Common Interview Questions

---

## Question

How do you start a system design interview?

### Answer

```text
Clarify functional and non-functional requirements before discussing architecture.
```

---

## Question

What is the biggest mistake candidates make?

### Answer

```text
Jumping directly into technical implementation without understanding the problem.
```

---

## Question

Why discuss trade-offs?

### Answer

```text
Because engineering decisions depend on business priorities, scale, latency, cost, and operational constraints.
```

---

## Question

What topics should appear in every AI system design?

### Answer

```text
Requirements, architecture, storage, scalability, security, reliability, observability, cost optimization, and trade-offs.
```
