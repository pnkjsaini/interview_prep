# What Is An Agentic AI System?

An Agentic AI System is an AI platform capable of autonomously completing tasks.

Unlike RAG,

which answers questions,

an Agent can

```text
Understand Goal

↓

Create Plan

↓

Execute Plan

↓

Observe Results

↓

Modify Plan

↓

Complete Goal
```

---

## Interview Takeaway

RAG retrieves information.

Agentic systems perform work.

# Step 2 — Planner

Planner converts goals into executable tasks.

Example

```text
Analyze Tesla
```

↓

```text
Collect News

↓

Retrieve Financial Statements

↓

Calculate Ratios

↓

Generate Summary
```

## Interview Takeaway

# Common Interview Questions

---

## Question

Why separate Planner and Workers?

### Answer

```text
It separates decision-making from execution, improving modularity, scalability, and maintainability.
```

---

## Question

Why use Shared Memory?

### Answer

```text
Shared memory reduces direct coupling between agents and enables coordinated execution.
```

---

## Question

Why have a Reflection layer?

### Answer

```text
To verify reasoning, improve answer quality, and reduce hallucinations before returning results.
```

---

## Question

Why maintain a Tool Registry?

### Answer

```text
It centralizes tool metadata, permissions, and schemas, enabling dynamic and secure tool selection.
```

---

## Question

How do you scale an Agent platform?

### Answer

```text
Scale planners, workers, memory services, vector databases, and inference independently using stateless microservices and autoscaling.
```

---

## Question

How is an Agentic system different from a RAG system?

### Answer

```text
RAG retrieves information to answer questions, while an Agentic system plans, executes actions, uses tools, and adapts based on observations.
```
