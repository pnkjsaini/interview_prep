# Common Interview Questions

---

## Question

Why parse code into an AST?

### Answer

```text
AST captures the structural representation of code, enabling accurate analysis beyond plain text matching.
```

---

## Question

Why use Repository RAG?

### Answer

```text
To ground code generation using the project's own source code, documentation, and architecture.
```

---

## Question

Why combine LLMs with static analysis?

### Answer

```text
LLMs provide reasoning and generation, while static analysis ensures correctness, style, and security.
```

---

## Question

Why maintain a Symbol Graph?

### Answer

```text
It captures dependencies and relationships between functions, classes, and modules, improving retrieval and reasoning.
```

---

## Question

Why validate generated code?

### Answer

```text
Generated code may compile incorrectly, violate project conventions, or introduce security issues.
```
