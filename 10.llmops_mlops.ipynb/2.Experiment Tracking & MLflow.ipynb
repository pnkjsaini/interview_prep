# Common Interview Questions

---

## Question

What is the difference between an Experiment and a Run?

### Answer

```text
An Experiment groups related runs, while a Run represents one execution of a model or workflow.
```

---

## Question

What are Parameters, Metrics, and Artifacts?

### Answer

```text
Parameters are configuration values, Metrics measure performance, and Artifacts are files generated during execution.
```

---

## Question

Why use a Model Registry?

### Answer

```text
To manage model versions, lifecycle stages, approvals, and deployments.
```

---

## Question

What is Model Lineage?

### Answer

```text
Lineage tracks the relationships between datasets, code, experiments, models, and deployments for reproducibility and governance.
```

---

## Question

How is MLflow used for LLM applications?

### Answer

```text
MLflow tracks prompts, traces, evaluations, models, agents, and deployments in addition to traditional ML experiments.
```
