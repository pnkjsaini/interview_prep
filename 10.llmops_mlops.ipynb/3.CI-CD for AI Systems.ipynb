# Common Interview Questions

---

## Question

Why is CI/CD more challenging for AI systems?

### Answer

```text
AI deployments require validating code, models, prompts, RAG pipelines, safety, and evaluation metrics—not just application logic.
```

---

## Question

What is an Evaluation Gate?

### Answer

```text
An automated quality checkpoint that blocks deployment if predefined performance or safety thresholds are not met.
```

---

## Question

Why use Canary Deployment?

### Answer

```text
To gradually expose a new model or prompt to production traffic while monitoring quality before a full rollout.
```

---

## Question

What is the purpose of Shadow Deployment?

### Answer

```text
To evaluate a new model with real production traffic without affecting end users.
```

---

## Question

Why use Infrastructure as Code?

### Answer

```text
To make infrastructure reproducible, version-controlled, and auditable across environments.
```
