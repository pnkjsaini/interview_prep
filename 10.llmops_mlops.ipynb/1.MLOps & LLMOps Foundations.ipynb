# What is MLOps?

MLOps stands for

```text
Machine Learning Operations
```

It applies software engineering and DevOps principles to Machine Learning.

Instead of asking

```text
How do we train a model?
```

MLOps asks

```text
How do we deploy,

monitor,

version,

retrain,

and operate

the model?
```

---

## Interview Takeaway

MLOps manages the complete lifecycle of machine learning systems.

---

# Blue-Green Deployment

```text
Blue

↓

Current Production

Green

↓

New Version
```

---

# Canary Deployment

```text
5%

↓

New Model

95%

↓

Old Model
```

---

# Shadow Deployment

New model receives

```text
Real Traffic
```

```text
Not Returned
```

---

## Interview Takeaway

# Common Interview Questions

---

## Question

Why is LLMOps different from MLOps?

### Answer

```text
LLMOps manages prompts, foundation models, RAG, agents, token costs, and hallucination evaluation in addition to traditional ML lifecycle management.
```

---

## Question

Why version prompts?

### Answer

```text
Prompt changes can significantly alter model behavior and should be reproducible and reversible.
```

---

## Question

Why use a model registry?

### Answer

```text
To manage model versions, deployment stages, ownership, and lifecycle.
```

---

## Question

Why use Canary deployments?

### Answer

```text
To safely validate new models using a small percentage of production traffic before full rollout.
```
