# Module 10: LLMOps / MLOps

# Chapter 6: Data Drift, Model Drift & Continuous Retraining

So far we have learned how to

```text
Build AI Systems

↓

Deploy Models

↓

Serve Models

↓

Monitor Models
```

Now comes one of the biggest production challenges.

> **How do AI systems stay accurate after months or years in production?**

Unlike traditional software,

AI systems become outdated because

```text
Data Changes

↓

User Behavior Changes

↓

Business Changes

↓

Environment Changes
```

This phenomenon is called **Model Drift**.

---

# Why Models Fail

Suppose we train a fraud detection model in

```text
January
```

It achieves

```text
99% Accuracy
```

Six months later

```text
New Fraud Patterns

↓

Customer Behavior Changes

↓

Transactions Change
```

Accuracy becomes

```text
82%
```

No bug exists.

The world changed.

---

## Interview Takeaway

Production ML systems fail because reality changes faster than models.

---

# The Production ML Lifecycle

A production ML system never stops evolving.

```text
Collect Data

↓

Train Model

↓

Deploy

↓

Monitor

↓

Detect Drift

↓

Retrain

↓

Redeploy

↓

Monitor Again
```

This continuous loop is the foundation of modern MLOps.

---

# What is Drift?

Drift means

```text
Production Data

≠

Training Data
```

or

```text
Relationships Learned

↓

No Longer Valid
```

---

# Types of Drift

The major categories are

```text
Data Drift

↓

Concept Drift

↓

Label Drift

↓

Feature Drift

↓

Prediction Drift
```

Each requires different monitoring strategies.

---

# Data Drift

Definition

The statistical distribution of input features changes.

Example

Training

```text
Average Customer Age

35 Years
```

Production

```text
Average Customer Age

52 Years
```

The input data has changed.

---

## Interview Takeaway

Data Drift affects the input distribution but not necessarily the relationship between features and labels.

---

# Example of Data Drift

Training

```text
Income

₹30K–₹70K
```

Production

```text
Income

₹80K–₹200K
```

The model is seeing unfamiliar inputs.

---

# Concept Drift

The relationship between inputs and outputs changes.

Example

Training

```text
Large Transactions

↓

Fraud
```

Production

```text
Large Transactions

↓

Normal
```

because business behavior changed.

---

## Interview Takeaway

Concept Drift changes the meaning of the data.

---

# Example of Concept Drift

COVID-19

↓

Consumer behavior changes

↓

Retail forecasting models fail

The features remain similar,

but their relationship with sales changes dramatically.

---

# Label Drift

Definition

The distribution of labels changes.

Example

Training

```text
Spam

10%
```

Production

```text
Spam

40%
```

The input distribution may remain stable,

but class frequencies change.

---

# Feature Drift

One or more important features change significantly.

Example

```text
Average Order Value

↓

₹500

↓

₹1800
```

A single feature drifts,

while the remaining features stay stable.

---

# Prediction Drift

Monitor

```text
Model Predictions
```

Example

Training

```text
Approved Loans

70%
```

Production

```text
Approved Loans

98%
```

The prediction distribution changes unexpectedly.

---

## Interview Takeaway

Prediction drift often indicates underlying data or concept drift.

---

# Causes of Drift

Common reasons include

```text
Seasonality

Business Changes

Economic Changes

Policy Updates

New Products

Changing User Behavior

External Events
```

---

# Detecting Drift

Drift must be measured,

not guessed.

Pipeline

```text
Training Data

↓

Production Data

↓

Statistical Comparison

↓

Drift Alert
```

---

# Statistical Drift Detection

Common methods

```text
Kolmogorov–Smirnov Test

Chi-Square Test

Population Stability Index (PSI)

Jensen-Shannon Divergence

KL Divergence

Wasserstein Distance
```

---

# Population Stability Index (PSI)

Very common in banking.

Interpretation

```text
PSI < 0.1

↓

No Drift

----------------

0.1–0.25

↓

Moderate Drift

----------------

>0.25

↓

Significant Drift
```

---

## Interview Takeaway

PSI is one of the most widely used production drift metrics.

---

# Monitoring Pipeline

```text
Production Data

↓

Feature Statistics

↓

Drift Detection

↓

Alert

↓

Investigation
```

---

# Performance Monitoring

Even if no drift is detected,

monitor

```text
Accuracy

Precision

Recall

F1

AUC

Latency
```

Performance degradation may reveal hidden problems.

---

# Retraining Strategies

Several strategies exist.

---

## Time-Based Retraining

Retrain

```text
Weekly

Monthly

Quarterly
```

Advantages

```text
Simple

Predictable
```

Disadvantages

```text
May Retrain Too Early

Or Too Late
```

---

# Drift-Based Retraining

Retrain only when

```text
Drift Threshold

Exceeded
```

Pipeline

```text
Monitor

↓

Drift

↓

Retrain

↓

Deploy
```

---

## Interview Takeaway

Drift-triggered retraining is more efficient than fixed schedules.

---

# Performance-Based Retraining

Retrain when

```text
Accuracy

↓

Below Threshold
```

Example

```text
Accuracy

<90%

↓

Trigger Pipeline
```

---

# Event-Based Retraining

Triggered by

```text
New Product

Policy Change

Market Crash

Regulation

Business Event
```

---

# Continuous Retraining Pipeline

```text
Production Data

↓

Validation

↓

Feature Engineering

↓

Training

↓

Evaluation

↓

Registry

↓

Deployment

↓

Monitoring
```

Fully automated.

---

# Champion-Challenger Strategy

Current Model

↓

Champion

New Model

↓

Challenger

Compare both

↓

Promote only if

```text
Better
```

---

## Interview Takeaway

Champion-Challenger reduces deployment risk.

---

# Shadow Evaluation

New model

↓

Receives production traffic

↓

Predictions logged

↓

Not shown to users.

Useful before promotion.

---

# Active Learning

Instead of labeling everything,

select

```text
Most Uncertain Samples
```

for human review.

Benefits

```text
Lower Labeling Cost

Higher Model Improvement
```

---

# Human-in-the-Loop

Certain predictions require review.

Example

```text
Medical Diagnosis

Loan Approval

Insurance Claims
```

Workflow

```text
Model

↓

Human Review

↓

Final Decision
```

---

# Data Versioning

Version

```text
Training Dataset

Validation Dataset

Test Dataset
```

```text
Delta Lake

DVC

LakeFS
```

---

# Feature Versioning

```text
Feature Definition

Transformation

Version
```

# Common Interview Questions

---

## Question

What is Data Drift?

### Answer

```text
A change in the statistical distribution of production input features compared to training data.
```

---

## Question

What is Concept Drift?

### Answer

```text
A change in the relationship between input features and target labels.
```

---

## Question

What is the Champion-Challenger strategy?

### Answer

```text
The production model (Champion) is compared against a new model (Challenger), and the challenger is promoted only if it performs better.
```

---

## Question

Why use Drift-Based Retraining?

### Answer

```text
It retrains models only when meaningful changes occur, reducing unnecessary computation while maintaining performance.
```

---

## Question

How do you detect drift?

### Answer

```text
Using statistical methods such as PSI, KS Test, Chi-Square Test, KL Divergence, or Jensen-Shannon Divergence.
```
