# Knowledge Base Lifecycle

Enterprise knowledge constantly changes.

Example

```text
Policies

Contracts

Wiki Pages

Reports

Emails
```

Every change should trigger

```text
Versioning

↓

Embedding

↓

Index Update

↓

Evaluation
```

---

## Interview Takeaway

Knowledge bases are living systems that require continuous synchronization.

---

# Knowledge Versioning

Every document should have

```text
Document ID

Version

Owner

Timestamp

Access Policy
```

Example

```text
HR Policy

v3.2
```

instead of

```text
HR_Final_Final_v12.pdf
```

---

# Incremental Indexing

Avoid rebuilding the entire vector database.

Instead

```text
Changed Documents

↓

Chunk

↓

Embed

↓

Update Index
```

Benefits

```text
Lower Cost

Lower Downtime

Faster Updates
```

---

## Interview Takeaway

Incremental indexing is preferred over full re-indexing for production systems.

---

# Common Interview Questions

---

## Question

Why version embeddings?

### Answer

```text
Different embedding models produce vectors in different semantic spaces, making versioning essential for reproducibility and safe migration.
```

---

## Question

Why prefer incremental indexing?

### Answer

```text
It updates only changed documents, reducing cost, latency, and operational downtime compared to rebuilding the entire index.
```

---

## Question

Should metadata updates always trigger re-embedding?

### Answer

```text
No. Metadata changes can often be updated independently without regenerating embeddings.
```

---

## Question

How do you evaluate a production RAG system?

### Answer

```text
Measure retrieval quality, grounding, faithfulness, correctness, latency, cost, and user feedback continuously.
```

---

## Question

What additional components require versioning in LLMOps?

### Answer

```text
Prompts, embeddings, chunking strategies, retrievers, tools, agents, memory policies, and knowledge bases.
```
