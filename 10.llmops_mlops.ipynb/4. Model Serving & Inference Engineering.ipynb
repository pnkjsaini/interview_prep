# What is Model Serving?

Model Serving is the process of exposing a trained model as a service that applications can use.

Instead of

```text
Notebook

↓

Prediction
```

we build

```text
Application

↓

REST API

↓

Inference Server

↓

Model

↓

Prediction
```

---

## Interview Takeaway

Model Serving converts trained models into production services.

---

# Offline vs Online Inference

## Offline Inference

Predictions generated ahead of time.

Example

```text
Nightly Recommendation Generation

↓

Store Predictions

↓

Serve Later
```

Applications

```text
Recommendation Systems

Fraud Reports

Forecasting

Customer Segmentation
```

Advantages

```text
Cheap

Efficient

Large Scale
```

---

## Online Inference

Predictions generated in real time.

Pipeline

```text
User Request

↓

Model

↓

Prediction

↓

Response
```

Applications

```text
Chatbots

Fraud Detection

Search Ranking

Personalization
```

---

## Interview Takeaway

Online inference optimizes latency, while offline inference optimizes throughput.

---

# Common Interview Questions

---

## Question

Online vs Offline Inference?

### Answer

```text
Online inference serves predictions in real time, while offline inference computes predictions ahead of time in batches.
```

---

## Question

Why use a Feature Store?

### Answer

```text
To ensure consistent feature computation across training and production while reducing duplicate engineering effort.
```

---

## Question

What is Training-Serving Skew?

### Answer

```text
A mismatch between feature generation during training and production inference that leads to incorrect predictions.
```

---

## Question

What is Data Drift?

### Answer

```text
A change in the statistical distribution of production input data compared to training data.
```

---

## Question

Why use Shadow Deployment?

### Answer

```text
To evaluate a new model using real production traffic without affecting end users.
```
