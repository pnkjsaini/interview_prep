# Common Interview Questions

---

## Question

Monitoring vs Observability?

### Answer

```text
Monitoring detects problems, while observability helps explain their root cause using logs, metrics, and traces.
```

---

## Question

What should be monitored in a production LLM system?

### Answer

```text
Latency, token usage, cost, hallucination rate, groundedness, prompt versions, safety, and user feedback.
```

---

## Question

What is a Golden Dataset?

### Answer

```text
A curated benchmark dataset with expected outputs used to detect regressions before deployment.
```

---

## Question

How do you evaluate a RAG system?

### Answer

```text
Evaluate retrieval quality, context relevance, generation correctness, groundedness, latency, and user satisfaction independently.
```

---

## Question

How do you monitor an Agent?

### Answer

```text
Track planning quality, tool selection, execution success, recovery behavior, latency, and task completion.
```
