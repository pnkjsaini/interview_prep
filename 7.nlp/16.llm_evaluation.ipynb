# Chapter 16: LLM Evaluation

In previous chapters we studied:

```text
LLMs

RAG

Fine-Tuning

LoRA

QLoRA
```

At this point a critical question appears:

> How do we know whether our AI system is actually good?

This is one of the most overlooked topics in AI engineering.

Many teams spend months building:

```text
RAG Systems

Agents

Fine-Tuned Models
```

but cannot answer:

```text
Is It Better?
```

Evaluation solves this problem.

---

# Why Evaluation Matters

Without evaluation:

```text
No Benchmark

No Comparison

No Improvement Measurement
```

---

Example:

```text
Model A

Model B
```

---

Question:

```text
Which One Is Better?
```

---

Need:

```text
Metrics
```

---

## Interview Takeaway

Evaluation provides objective measurements of model quality.

---

# Types Of Evaluation

Modern AI systems typically evaluate:

```text
Model Quality

Retrieval Quality

Answer Quality

Agent Quality
```

---

Different metrics are needed for each.

---

## Interview Takeaway

Evaluation depends on the component being measured.

---

# Traditional NLP Evaluation

Before LLMs:

```text
Machine Translation

Summarization

Classification
```

used fixed metrics.

---

Examples:

```text
BLEU

ROUGE

Accuracy
```

---

LLMs introduced new challenges.

---

## Interview Takeaway

Traditional NLP metrics are useful but often insufficient for LLMs.

---

# Perplexity

One of the most common interview topics.

---

# Definition

Measures:

```text
How Surprised

The Model Is
```

by the data.

---

Lower:

```text
Better
```

---

Higher:

```text
Worse
```

---

## Intuition

Suppose model predicts:

```text
The Cat Sat On The
```

---

Expected:

```text
Mat
```

---

If model strongly predicts:

```text
Mat
```

↓

```text
Low Perplexity
```

---

## Interview Takeaway

Perplexity measures language modeling quality.

---

# Why Perplexity Matters

Good language models:

```text
Predict Next Tokens Well
```

---

Therefore:

```text
Lower Perplexity
```

---

Generally indicates:

```text
Better Language Modeling
```

---

## Interview Takeaway

Perplexity is commonly used during pretraining evaluation.

---

# Limitations Of Perplexity

Important interview topic.

---

Low perplexity does not guarantee:

```text
Good Reasoning

Good Retrieval

Good Conversations
```

---

Example:

```text
Predict Text Well

Still Hallucinate
```

---

## Interview Takeaway

Perplexity measures prediction quality, not usefulness.

---

# BLEU Score

Traditional machine translation metric.

---

Measures:

```text
N-Gram Overlap
```

between:

```text
Generated Text

Reference Text
```

---

Example:

```text
Reference:
The cat sat on the mat.

Generated:
The cat sat on the mat.
```

---

High BLEU score.

---

## Interview Takeaway

BLEU evaluates overlap with reference answers.

---

# BLEU Limitations

Problem:

```text
Many Correct Answers Exist
```

---

Example:

```text
Reference:
The cat sat on the mat.

Generated:
A cat rested on the mat.
```

---

Meaning:

```text
Same
```

---

BLEU:

```text
Lower Score
```

---

## Interview Takeaway

BLEU struggles with semantic equivalence.

---

# ROUGE

Common summarization metric.

---

Measures:

```text
Recall-Oriented Overlap
```

---

Popular variants:

```text
ROUGE-1

ROUGE-2

ROUGE-L
```

---

Used heavily in:

```text
Summarization
```

---

## Interview Takeaway

ROUGE measures similarity between generated and reference summaries.

---

# ROUGE Limitations

Like BLEU:

```text
Surface Matching
```

---

Cannot fully capture:

```text
Meaning

Reasoning

Correctness
```

---

## Interview Takeaway

ROUGE is useful but insufficient for LLM evaluation.

---

# BERTScore

Modern semantic evaluation metric.

---

Instead of:

```text
Word Matching
```

---

Uses:

```text
Embeddings
```

---

Measures:

```text
Semantic Similarity
```

---

Example:

```text
Car

Automobile
```

---

High similarity.

---

## Interview Takeaway

BERTScore evaluates semantic similarity rather than exact wording.

---

# Why BERTScore Is Better

Captures:

```text
Meaning

Context

Semantics
```

---

Rather than:

```text
Exact Tokens
```

---

## Interview Takeaway

BERTScore correlates better with human judgments.

---

# Human Evaluation

Still considered:

```text
Gold Standard
```

---

Humans judge:

```text
Correctness

Helpfulness

Relevance

Safety
```

---

## Interview Takeaway

Human evaluation remains the most reliable assessment method.

---

# Problems With Human Evaluation

---

Expensive

---

Slow

---

Hard To Scale

---

Inconsistent

---

## Interview Takeaway

Human evaluation is accurate but costly.

---

# LLM-As-A-Judge

One of the hottest interview topics.

---

Idea:

```text
Use LLM

To Evaluate LLM
```

---

Example:

```text
Question

Reference Answer

Generated Answer
```

---

Judge Model scores:

```text
Correctness

Relevance

Quality
```

---

## Interview Takeaway

LLM-as-a-Judge automates large-scale evaluation.

---

# Why LLM-As-A-Judge Works

Strong models understand:

```text
Facts

Reasoning

Language Quality
```

---

Can approximate:

```text
Human Preferences
```

---

## Interview Takeaway

LLM judges often correlate surprisingly well with humans.

---

# Pairwise Evaluation

Common in model comparisons.

---

Instead of:

```text
Score One Answer
```

---

Compare:

```text
Answer A

vs

Answer B
```

---

Judge selects:

```text
Better Answer
```

---

## Interview Takeaway

Pairwise comparisons reduce evaluation bias.

---

# Hallucination Evaluation

Critical production topic.

---

Question:

```text
Did Model Invent Information?
```

---

Need:

```text
Fact Verification
```

---

Examples:

```text
Fake Citations

Wrong Numbers

Invented Facts
```

---

## Interview Takeaway

Hallucination evaluation measures factual reliability.

---

# Faithfulness

One of the most important RAG metrics.

---

Definition:

```text
Answer Supported

By Retrieved Context
```

---

Question:

```text
Did Answer Stay Grounded?
```

---

## Interview Takeaway

Faithfulness measures adherence to retrieved evidence.

---

# Groundedness

Closely related.

---

Measures:

```text
How Much Of Answer

Comes From Context
```

---

Higher:

```text
Better
```

---

## Interview Takeaway

Groundedness measures evidence-based generation.

---

# Correctness

Different from faithfulness.

---

Question:

```text
Is The Answer Actually Correct?
```

---

Possible:

```text
Faithful

But Wrong
```

---

if source documents contain errors.

---

## Interview Takeaway

Correctness and faithfulness are not identical.

---

# RAG Evaluation Framework

Evaluate:

```text
Retriever

Reranker

Generator
```

separately.

---

# Retrieval Evaluation

Metrics:

```text
Recall@K

MRR

NDCG
```

---

Measures:

```text
Retriever Quality
```

---

## Interview Takeaway

Retrieval quality must be evaluated independently.

---

# Recall@K

Question:

```text
Did Relevant Document

Appear In Top K?
```

---

Example:

```text
Recall@5
```

---

## Interview Takeaway

Recall@K measures retrieval coverage.

---

# MRR

Measures:

```text
How Early

Relevant Result Appears
```

---

Higher:

```text
Better
```

---

## Interview Takeaway

MRR rewards high-ranking relevant results.

---

# NDCG

Measures:

```text
Ranking Quality
```

---

Considers:

```text
Position

And Relevance
```

---

## Interview Takeaway

NDCG evaluates retrieval ranking effectiveness.

---

# End-To-End RAG Evaluation

Evaluate:

```text
Question
      ↓

Retrieval
      ↓

Generation
      ↓

Final Answer
```

---

Need to identify:

```text
Retriever Failure

Generator Failure
```

---

## Interview Takeaway

End-to-end evaluation isolates system bottlenecks.

---

# Offline Evaluation

Performed before deployment.

---

Uses:

```text
Benchmark Datasets

Test Questions

Ground Truth Answers
```

---

Advantages:

```text
Cheap

Repeatable
```

---

## Interview Takeaway

Offline evaluation is used during development.

---

# Online Evaluation

Performed in production.

---

Uses:

```text
Real Users
```

---

Measures:

```text
Engagement

Success Rate

User Satisfaction
```

---

## Interview Takeaway

Online evaluation measures real-world effectiveness.

---

# A/B Testing

Very common production topic.

---

Compare:

```text
System A

System B
```

---

Example:

```text
Old Retriever

New Retriever
```

---

Measure:

```text
User Outcomes
```

---

## Interview Takeaway

A/B testing validates improvements in production.

---

# Enterprise Evaluation Pipeline

```text
Offline Benchmarks
         ↓

Human Review
         ↓

LLM Judge
         ↓

A/B Testing
         ↓

Production Monitoring
```

---

This is how modern AI teams evaluate systems.

---

## Interview Takeaway

Evaluation should occur throughout the AI lifecycle.

# Common Interview Questions

---

## Question

What is perplexity?

### Answer

```text
A metric measuring how well a language model predicts text.
```

---

## Question

Why is lower perplexity better?

### Answer

```text
Because the model is less surprised by the data.
```

---

## Question

Difference between BLEU and ROUGE?

### Answer

```text
BLEU is commonly used for translation.

ROUGE is commonly used for summarization.
```

---

## Question

What is BERTScore?

### Answer

```text
An embedding-based metric measuring semantic similarity.
```

---

## Question

What is faithfulness?

### Answer

```text
The degree to which an answer is supported by retrieved context.
```

---

## Question

What is groundedness?

### Answer

```text
How strongly an answer relies on provided evidence.
```

---

## Question

What is LLM-as-a-Judge?

### Answer

```text
Using a language model to evaluate generated outputs.
```

---

## Question

How do you evaluate a RAG system?

### Answer

```text
Evaluate retrieval quality and answer quality separately.
```

---

## Question

What is Recall@K?

### Answer

```text
The percentage of queries where relevant documents appear in the top K results.
```
