```text
RNN

LSTM

GRU
```

```text
One Token At A Time
```

---

```text
I Love Machine Learning
```

Processing:

```text
I
 ↓

Love
 ↓

Machine
 ↓

Learning
```

---

Problems:

```text
Slow Training

Limited Parallelism

Long Range Dependencies

Vanishing Gradients
```

---

Researchers wanted:

```text
Faster Training

Better Context Understanding

Massive Scalability
```

---

Result:

```text
Transformer
```

---

Transformers were designed to eliminate sequential processing limitations.

# High-Level Transformer Architecture

At a very high level:

```text
Input Text
       ↓

Tokenization
       ↓

Token IDs
       ↓

Embeddings
       ↓

Positional Encoding
       ↓

Transformer Layers
       ↓

Output
```

# Embedding Layer Review

Input:

```text
I Love AI
```

---

Token IDs:

```text
105

289

901
```

---

Embedding Layer converts IDs into:

```text
Dense Vectors
```

---

Example:

```text
[0.23, -0.44, 0.78]
```

---

Now words have:

```text
Meaning
```

---

Problem:

```text
No Position Information
```

---

## Interview Takeaway

Embeddings capture meaning but not word order.

---

# Why Positional Encoding Exists

Consider:

```text
Dog Bites Man
```

---

and

```text
Man Bites Dog
```

---

Same words.

---

Different meaning.

---

Embeddings alone cannot distinguish them.

---

Transformer attention is:

```text
Permutation Invariant
```

---

Meaning:

```text
A B C

C B A
```

appear similar.

---

Need:

```text
Position Information
```

---

## Interview Takeaway

Positional encoding injects word order into transformers.

---

# Positional Encoding

## Definition

A technique that adds position information to embeddings.

---

Instead of:

```text
Embedding
```

use:

```text
Embedding

+

Position Information
```

---

Result:

```text
Meaning

+

Order
```

---

## Interview Takeaway

Positional encoding combines semantics with sequence order.

---

# Original Transformer Positional Encoding

The original paper used:

```text
Sinusoidal Encoding
```

---

Based on:

```text
Sine

Cosine
```

functions.

---

Each position receives:

```text
Unique Pattern
```

---

Example:

```text
Position 1

Position 2

Position 3
```

all receive different vectors.

---

## Why Use Sinusoids?

Allows model to learn:

```text
Relative Distances

Between Words
```

---

Even for unseen sequence lengths.

---

## Interview Takeaway

Sinusoidal encoding provides deterministic position information.

---

# Learned Positional Encoding

Instead of fixed values:

```text
Learn Position Embeddings
```

during training.

---

Advantages:

```text
Flexible

Task Specific
```

---

Disadvantages:

```text
Limited Generalization
```

---

## Interview Takeaway

Learned positional embeddings are trainable alternatives to sinusoidal encoding.

---

# RoPE (Rotary Positional Encoding)

Very important modern interview topic.

---

Used by:

```text
Llama

Mistral

Qwen

DeepSeek
```

---

Instead of:

```text
Adding Position
```

---

RoPE:

```text
Rotates Embeddings
```

in vector space.

---

Benefits:

```text
Better Long Context

Better Generalization
```

---

## Interview Takeaway

Most modern LLMs use RoPE instead of traditional positional encoding.

---

# ALiBi

Another modern approach.

---

Idea:

```text
Bias Attention Scores
```

based on distance.

---

Benefits:

```text
Long Context Efficiency
```

---

Less common than RoPE.

---

## Interview Takeaway

ALiBi improves long-context attention through positional bias.

---

# Self-Attention Review

Every token creates:

```text
Query

Key

Value
```

---

Attention score:

```text
Query

×

Key
```

---

Produces:

```text
Importance Score
```

---

Softmax converts scores into:

```text
Probabilities
```

---

Weighted values produce:

```text
Context-Aware Representation
```

---

## Interview Takeaway

Self-attention enables global context understanding.

---

# Why One Attention Head Is Not Enough

Consider sentence:

```text
The dog chased the cat.
```

---

One attention head may learn:

```text
Subject-Object Relation
```

---

But language contains many relationships.

---

Examples:

```text
Grammar

Meaning

Pronouns

Long Context
```

---

Need:

```text
Multiple Perspectives
```

---

## Interview Takeaway

One attention head cannot capture all language relationships.

---

# Multi-Head Attention

The solution.

---

Instead of:

```text
One Attention Mechanism
```

---

Use:

```text
Many Attention Mechanisms
```

in parallel.

---

Example:

```text
Head 1 → Grammar

Head 2 → Meaning

Head 3 → Coreference

Head 4 → Long Context
```

---

All heads learn different patterns.

---

## Interview Takeaway

Multi-head attention learns multiple relationships simultaneously.

---

# Multi-Head Attention Flow

```text
Input
      ↓

Multiple QKV Projections
      ↓

Attention Head 1

Attention Head 2

Attention Head 3
      ↓

Concatenate Outputs
      ↓

Final Projection
```

---

Result:

```text
Richer Representation
```

---

## Interview Takeaway

Multiple heads provide diverse contextual understanding.

---

# Feed Forward Network (FFN)

Often ignored in interviews.

---

Every transformer block contains:

```text
Attention
      ↓

Feed Forward Network
```

---

Typical structure:

```text
Linear
     ↓

Activation
     ↓

Linear
```

---

Purpose:

```text
Feature Transformation
```

---

## Interview Takeaway

FFNs increase representational capacity after attention.

---

# Residual Connections

Deep networks suffer from:

```text
Vanishing Gradients
```

---

Solution:

```text
Skip Connections
```

---

Formula:

```text
Output = Layer(x) + x
```

---

Information can bypass layers.

---

Training becomes easier.

---

## Interview Takeaway

Residual connections improve gradient flow and training stability.

---

# Layer Normalization

Training deep networks is difficult.

---

Activations may become:

```text
Too Large

Too Small
```

---

LayerNorm normalizes activations.

---

Benefits:

```text
Stable Training

Faster Convergence
```

---

## Interview Takeaway

Layer normalization stabilizes transformer training.

---

# Transformer Encoder

Used by:

```text
BERT

RoBERTa

DeBERTa
```

---

Structure:

```text
Multi-Head Attention
         ↓

Add & Norm
         ↓

Feed Forward
         ↓

Add & Norm
```

---

Repeated many times.

---

Purpose:

```text
Language Understanding
```

---

## Interview Takeaway

Encoders learn rich bidirectional representations.

---

# Bidirectional Context

```text
Left Context

And

Right Context
```

simultaneously.

---

Example:

```text
The bank approved my loan.
```

---

Word:

```text
Bank
```

uses surrounding words to infer meaning.

---

## Interview Takeaway

BERT understands context from both directions.

---

```text
Text Generation
```

---

---

Sentence:

```text
I Love AI
```

---

When predicting:

```text
Love
```

---

Model cannot see:

```text
AI
```

---

---

Called:

```text
Causal Masking
```

---

## Interview Takeaway

Causal masking prevents information leakage during generation.

---

# Common Interview Questions

---

## Question

Why did Transformers replace RNNs?

### Answer

```text
Because they support parallel training and handle long-range dependencies more effectively.
```

---

## Question

Why is positional encoding necessary?

### Answer

```text
Because attention alone has no notion of word order.
```

---

## Question

What is multi-head attention?

### Answer

```text
Multiple attention mechanisms operating in parallel to learn different relationships.
```

---

## Question

Why use residual connections?

### Answer

```text
To improve gradient flow and stabilize deep networks.
```

---

## Question

What is LayerNorm?

### Answer

```text
A normalization technique that stabilizes transformer training.
```

---

## Question

What is causal masking?

### Answer

```text
A mechanism preventing a decoder from seeing future tokens.
```

---

## Question

Difference between BERT and GPT?

### Answer

```text
BERT is encoder-based and bidirectional.

GPT is decoder-based and autoregressive.
```

---

## Question

Why is attention O(n²)?

### Answer

```text
Because every token attends to every other token.
```
