# What Is An LLM?

## Definition

LLM stands for:

```text
Large Language Model
```

---

A language model trained on:

```text
Massive Data

Massive Compute

Massive Parameters
```

---

Goal:

```text
Predict Tokens

Understand Language

Generate Text

Reason About Information
```

---

Examples:

```text
GPT

Claude

Gemini

Llama

DeepSeek

Qwen
```

---

## Interview Takeaway

LLMs are large transformer-based models trained on massive text corpora.

---

# Why Are They Called "Large"?

Three dimensions.

---

# Large Dataset

Training data:

```text
Books

Wikipedia

Web Pages

Code

Scientific Papers
```

---

Measured in:

```text
Billions

Trillions

Of Tokens
```

---

# Large Parameters

Parameters:

```text
Millions

Billions

Trillions
```

---

Parameters store:

```text
Learned Knowledge
```

---

# Large Compute

Training requires:

```text
GPUs

TPUs

Distributed Clusters
```

---

## Interview Takeaway

"Large" refers to data, parameters, and compute.

---

# What Does An LLM Learn?

Initially:

```text
Nothing
```

---

During training it learns:

```text
Grammar

Language Structure

Facts

Patterns

Reasoning Behaviors
```

---

Example:

```text
Paris → France

Delhi → India
```

---

Not memorized as a database.

---

Learned through:

```text
Statistical Patterns
```

---

## Interview Takeaway

LLMs learn language patterns rather than storing explicit rules.

---

# The LLM Training Pipeline

High-Level Flow:

```text
Raw Data
      ↓

Tokenization
      ↓

Pretraining
      ↓

Instruction Tuning
      ↓

Alignment
      ↓

Deployment
```

---

This pipeline powers nearly every modern LLM.

---

# Stage 1: Pretraining

The most expensive stage.

---

# Definition

Train the model using:

```text
Next Token Prediction
```

on massive datasets.

---

Example:

```text
The Cat Sat On The
```

Predict:

```text
Mat
```

---

Repeated:

```text
Billions Of Times
```

---

## Interview Takeaway

Pretraining teaches general language understanding.

---

# Why Pretraining Matters

Without pretraining:

```text
Model Knows Nothing
```

---

After pretraining:

```text
Language

Facts

Patterns

Syntax

Semantics
```

---

The model develops:

```text
World Knowledge
```

---

## Interview Takeaway

Pretraining creates the foundation of LLM intelligence.

---

# Self-Supervised Learning

Common interview topic.

---

Traditional ML:

```text
Input

Label
```

---

Example:

```text
Cat → Animal
```

---

LLMs use:

```text
Self-Supervision
```

---

Labels generated automatically.

---

Example:

```text
The Cat Sat On The
```

Label:

```text
Mat
```

already exists.

---

## Interview Takeaway

LLMs learn using self-generated labels.

---

# Foundation Models

After pretraining:

```text
GPT

Llama

Claude

Gemini
```

become:

```text
Foundation Models
```

---

Capable of many tasks.

---

Examples:

```text
Summarization

Translation

Question Answering

Coding
```

---

## Interview Takeaway

A foundation model is a broadly capable pretrained model.

---

# Why Pretraining Alone Is Not Enough

Pretrained models complete text.

---

Prompt:

```text
How do I bake a cake?
```

---

Model may continue unpredictably.

---

Need:

```text
Instruction Following
```

---

## Interview Takeaway

Pretraining teaches language, not obedience.

---

# Stage 2: Instruction Tuning

One of the most important interview topics.

---

Goal:

```text
Teach Model

How To Follow Instructions
```

---

Training examples:

```text
Instruction

↓

Desired Response
```

---

Example:

```text
Summarize This Article
```

↓

```text
Short Summary
```

---

## Interview Takeaway

Instruction tuning teaches task execution.

---

# Why Instruction Tuning Works

Pretraining teaches:

```text
Language
```

---

Instruction tuning teaches:

```text
Behavior
```

---

After tuning:

```text
ChatGPT-like Interactions
```

become possible.

---

## Interview Takeaway

Instruction tuning converts foundation models into assistants.

---

# Supervised Fine-Tuning (SFT)

Most common industry term.

---

Pipeline:

```text
Pretrained Model
        ↓

Human Demonstrations
        ↓

Fine-Tuned Model
```

---

Examples:

```text
Customer Support

Medical QA

Legal QA

Code Generation
```

---

## Interview Takeaway

SFT teaches desired behaviors using labeled examples.

---

# Stage 3: RLHF

One of the most asked LLM interview topics.

---

RLHF:

```text
Reinforcement Learning
From Human Feedback
```

---

Goal:

```text
Align Model Behavior
```

with human preferences.

---

## Interview Takeaway

RLHF improves helpfulness, safety, and alignment.

---

# Why RLHF Exists

Suppose model produces:

```text
Technically Correct

But Poor Response
```

---

Humans rank outputs.

---

Example:

```text
Response A

Response B
```

---

Humans choose:

```text
Better Answer
```

---

Model learns preferences.

---

## Interview Takeaway

RLHF optimizes for preferred responses rather than only correctness.

---

# RLHF Pipeline

```text
Pretrained Model
       ↓

Instruction Tuning
       ↓

Human Rankings
       ↓

Reward Model
       ↓

Policy Optimization
       ↓

Aligned Assistant
```

---

## Interview Takeaway

RLHF aligns model behavior with human expectations.

---

# In-Context Learning

Very common interview topic.

---

Example:

```text
English → French

Dog → Chien

Cat → Chat

Bird →
```

---

Model predicts:

```text
Oiseau
```

---

Without retraining.

---

Learns from:

```text
Prompt Context
```

---

## Interview Takeaway

In-context learning allows models to learn from examples inside prompts.

---

# Zero-Shot Learning

No examples provided.

---

Prompt:

```text
Translate To French:
Hello
```

---

Model performs task immediately.

---

## Interview Takeaway

Zero-shot learning requires no demonstrations.

---

# Few-Shot Learning

Provide examples.

---

Prompt:

```text
Dog → Chien

Cat → Chat

Bird →
```

---

Model infers pattern.

---

## Interview Takeaway

Few-shot learning uses demonstrations inside the prompt.

---

# Hallucination

One of the most important production topics.

---

# Definition

Generating information that sounds correct but is false.

---

Example:

```text
Invented Facts

Fake Citations

Incorrect Numbers
```

---

## Interview Takeaway

Hallucination occurs when a model generates unsupported content.

---

# Why Hallucinations Happen

LLMs optimize for:

```text
Likely Text
```

---

Not:

```text
Truth
```

---

Prediction objective:

```text
Next Token
```

---

not:

```text
Fact Verification
```

---

## Interview Takeaway

Hallucinations arise because language modeling is not truth modeling.

---

# Reducing Hallucinations

Common approaches:

```text
RAG

Tool Calling

Grounding

Fine-Tuning

Verification Systems
```

---

RAG is currently the most common.

---

## Interview Takeaway

Grounding models with external knowledge reduces hallucinations.

---

# Common Interview Questions

---

## Question

What is an LLM?

### Answer

```text
A large transformer-based language model trained on massive text corpora.
```

---

## Question

What is pretraining?

### Answer

```text
Training a model using self-supervised next-token prediction.
```

---

## Question

What is instruction tuning?

### Answer

```text
Training a model to follow user instructions.
```

---

## Question

What is RLHF?

### Answer

```text
A technique that aligns model behavior using human feedback.
```

---

## Question

What is a foundation model?

### Answer

```text
A broadly capable pretrained model that can be adapted to many tasks.
```

---

## Question

What is hallucination?

### Answer

```text
Generation of plausible but incorrect information.
```

---

## Question

What is in-context learning?

### Answer

```text
Learning patterns from examples provided inside the prompt.
```

---

## Question

Difference between zero-shot and few-shot learning?

### Answer

```text
Zero-shot uses no examples.

Few-shot uses examples in the prompt.
```
