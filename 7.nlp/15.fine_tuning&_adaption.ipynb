# Chapter 15: Fine-Tuning & Adaptation

In the previous chapters we studied:

At this point, one of the biggest questions in AI engineering appears:

> If I already have GPT, Claude, Gemini or Llama, why would I fine-tune a model?

---

# Why Fine-Tuning Exists

Pretrained models are:

```text
General Purpose Models
```

---

They understand:

```text
Language

Reasoning

Coding

Knowledge
```

---

But enterprises need:

```text
Company Knowledge

Custom Behaviors

Domain Expertise

Specific Output Formats
```

---

Examples:

```text
Medical Assistant

Legal Assistant

Insurance Assistant

Financial Assistant
```

---

## Interview Takeaway

Fine-tuning adapts a general-purpose model to a specific domain or behavior.

---

# The Evolution Of LLM Training

```text
Pretraining
       ↓

Foundation Model
       ↓

Instruction Tuning
       ↓

Aligned Assistant
       ↓

Domain Adaptation
```

---

Fine-tuning happens after:

```text
Pretraining
```

---

## Interview Takeaway

Fine-tuning builds on a pretrained foundation model.

---

# What Is Fine-Tuning?

## Definition

Continue training a pretrained model using:

```text
Task-Specific Data
```

---

Example:

```text
Base Llama Model
```

↓

```text
Medical Dataset
```

↓

```text
Medical Assistant
```

---

## Interview Takeaway

Fine-tuning modifies model weights using specialized data.

---

# Why Not Train From Scratch?

Training from scratch requires:

```text
Trillions Of Tokens

Thousands Of GPUs

Months Of Compute
```

---

Cost:

```text
Millions Of Dollars
```

---

Fine-tuning requires:

```text
Much Less Data

Much Less Compute
```

---

## Interview Takeaway

Fine-tuning is dramatically cheaper than pretraining.

---

# Types Of Fine-Tuning

Modern AI systems use multiple adaptation methods.

---

# Full Fine-Tuning

Traditional approach.

---

Update:

```text
Every Parameter
```

---

Example:

```text
7 Billion Parameter Model
```

↓

```text
Update All 7 Billion Parameters
```

---

## Advantages

```text
Maximum Flexibility

Highest Potential Accuracy
```

---

## Disadvantages

```text
Very Expensive

High GPU Requirements

Storage Cost
```

---

## Interview Takeaway

Full fine-tuning updates all model parameters.

---

# Why Full Fine-Tuning Is Expensive

Need to store:

```text
Weights

Gradients

Optimizer States
```

---

Memory requirement grows rapidly.

---

Example:

```text
70B Model
```

can require:

```text
Hundreds Of GBs
```

during training.

---

## Interview Takeaway

Memory consumption is the primary challenge of full fine-tuning.

---

# Supervised Fine-Tuning (SFT)

One of the most common interview topics.

---

# Definition

Train using:

```text
Instruction

↓

Desired Response
```

pairs.

---

Example:

```text
Question:
What Is Machine Learning?

Answer:
Machine learning is...
```

---

Model learns:

```text
Desired Behavior
```

---

## Interview Takeaway

SFT teaches models how to respond to instructions.

---

# Instruction Tuning

A specialized form of SFT.

---

Goal:

```text
Teach Models

To Follow Instructions
```

---

Example:

```text
Summarize

Translate

Classify

Explain
```

---

Without instruction tuning:

```text
Model Completes Text
```

---

With instruction tuning:

```text
Model Acts Like Assistant
```

---

## Interview Takeaway

Instruction tuning transforms foundation models into assistants.

---

# Domain Adaptation

Very common enterprise topic.

---

Goal:

```text
Adapt To Specialized Knowledge
```

---

Examples:

```text
Healthcare

Finance

Legal

Insurance
```

---

Dataset:

```text
Domain Documents

Domain QA

Domain Conversations
```

---

## Interview Takeaway

Domain adaptation improves performance in specialized fields.

---

# Problems With Full Fine-Tuning

Organizations discovered:

```text
Too Expensive
```

---

Need:

```text
Cheaper Alternatives
```

---

This led to:

```text
PEFT
```

---

# Parameter Efficient Fine-Tuning (PEFT)

One of the most important modern concepts.

---

# Definition

Update only a small subset of parameters.

---

Instead of:

```text
7 Billion Parameters
```

update:

```text
Few Million Parameters
```

---

Result:

```text
Huge Cost Reduction
```

---

## Interview Takeaway

PEFT reduces training cost by updating a small number of parameters.

---

# Why PEFT Became Popular

Benefits:

```text
Lower GPU Usage

Lower Storage

Faster Training

Multiple Fine-Tuned Variants
```

---

Enterprise adoption became widespread.

---

## Interview Takeaway

PEFT enables practical fine-tuning of large models.

---

# LoRA

The most important fine-tuning technique.

---

LoRA:

```text
Low Rank Adaptation
```

---

Introduced:

```text
2021
```

---

Now industry standard.

---

## Interview Takeaway

LoRA is the most widely used PEFT technique.

---

# Core Idea Behind LoRA

Instead of modifying:

```text
Original Weights
```

---

Freeze them.

---

Add:

```text
Small Trainable Matrices
```

---

Training updates:

```text
LoRA Layers Only
```

---

Base model remains unchanged.

---

## Interview Takeaway

LoRA freezes the original model and trains lightweight adapters.

---

# Why Is It Called Low Rank?

Suppose weight matrix:

```text
W
```

---

Instead of learning:

```text
Full Matrix
```

---

Learn:

```text
A × B
```

where:

```text
Rank << Dimension
```

---

Fewer parameters.

---

## Interview Takeaway

LoRA approximates weight updates using low-rank matrices.

---

# Benefits Of LoRA

---

# Smaller Memory Footprint

---

# Faster Training

---

# Multiple Adapters

---

# Easy Deployment

---

## Interview Takeaway

LoRA achieves near full fine-tuning performance with significantly lower cost.

---

# LoRA Architecture

```text
Base Model
     ↓

Frozen Weights
     ↓

LoRA Adapters
     ↓

Training
```

---

Only adapters learn.

---

## Interview Takeaway

The base model remains fixed while adapters are updated.

---

# LoRA Hyperparameters

Frequently asked.

---

# Rank (r)

Controls:

```text
Adapter Capacity
```

---

Small rank:

```text
Lower Cost
```

---

Large rank:

```text
Higher Capacity
```

---

# Alpha

Controls:

```text
Update Scaling
```

---

## Interview Takeaway

Rank determines adaptation capacity.

---

# QLoRA

One of the hottest interview topics.

---

QLoRA:

```text
Quantized LoRA
```

---

Combines:

```text
Quantization

+

LoRA
```

---

Purpose:

```text
Train Large Models

Using Small GPUs
```

---

## Interview Takeaway

QLoRA enables efficient fine-tuning through quantization.

---

# How QLoRA Works

Step 1:

```text
Quantize Base Model
```

typically:

```text
4-bit
```

---

Step 2:

```text
Freeze Model
```

---

Step 3:

```text
Train LoRA Adapters
```

---

Result:

```text
Massive Memory Reduction
```

---

## Interview Takeaway

QLoRA trains adapters on top of quantized models.

---

# Why QLoRA Was A Breakthrough

Before:

```text
Need Expensive GPUs
```

---

After:

```text
Consumer GPUs
```

can fine-tune large models.

---

## Interview Takeaway

QLoRA democratized LLM fine-tuning.

---

# Adapters

LoRA belongs to a broader family.

---

Idea:

```text
Insert Small Modules

Between Layers
```

---

Train:

```text
Adapter Layers Only
```

---

## Interview Takeaway

Adapters provide modular model customization.

---

# Prefix Tuning

Another PEFT technique.

---

Idea:

```text
Learn Virtual Tokens
```

---

Inserted before prompts.

---

Model learns:

```text
Task Specific Behavior
```

---

## Interview Takeaway

Prefix tuning learns trainable prompt representations.

---

# Prompt Tuning

Even lighter.

---

Train:

```text
Soft Prompts
```

instead of model weights.

---

Advantages:

```text
Extremely Cheap
```

---

## Interview Takeaway

Prompt tuning updates prompts rather than model parameters.

---

# PEFT Comparison

| Method | Train Parameters |
|----------|----------|
| Full Fine-Tuning | All |
| LoRA | Few |
| QLoRA | Few |
| Adapter | Few |
| Prefix Tuning | Very Few |
| Prompt Tuning | Minimal |

---

## Interview Takeaway

PEFT methods reduce cost by limiting trainable parameters.

---

# Catastrophic Forgetting

Important interview topic.

---

Problem:

After fine-tuning:

```text
Model Forgets
```

previous capabilities.

---

Example:

```text
Great At Medical Tasks

Worse At General Tasks
```

---

## Interview Takeaway

Fine-tuning can overwrite previously learned knowledge.

---

# Alignment

Goal:

```text
Improve Behavior
```

---

Examples:

```text
Helpfulness

Safety

Formatting

Professional Tone
```

---

Often combined with:

```text
SFT

RLHF
```

---

## Interview Takeaway

Alignment focuses on behavior rather than knowledge.

---

# Fine-Tuning vs RAG

One of the most common interview questions.

---

# Fine-Tuning

Changes:

```text
Model Weights
```

---

Good for:

```text
Behavior

Style

Domain Skills
```

---

# RAG

Changes:

```text
Knowledge Source
```

---

Good for:

```text
Documents

Policies

Frequently Changing Data
```

---

## Interview Takeaway

Fine-tuning changes behavior; RAG provides knowledge.

---

# When To Use RAG

Examples:

```text
Company Policies

Product Documentation

Knowledge Bases

Frequently Updated Data
```

---

## Interview Takeaway

Use RAG when knowledge changes frequently.

---

# When To Use Fine-Tuning

Examples:

```text
Output Format

Writing Style

Specialized Domain Behavior

Classification Tasks
```

---

## Interview Takeaway

Use fine-tuning when behavior must change.

---

# When To Use Both

Most enterprise systems use:

```text
Fine-Tuning

+

RAG
```

---

Example:

```text
Fine-Tuned Customer Support Model

+

Company Knowledge Base
```

---

Best of both worlds.

---

## Interview Takeaway

Fine-tuning and RAG are complementary technologies.

---

# Enterprise Design Decision

Question:

```text
Need New Knowledge?
```

↓

```text
Use RAG
```

---

Question:

```text
Need New Behavior?
```

↓

```text
Fine-Tune
```

---

Question:

```text
Need Both?
```

↓

```text
Combine Both
```

---

## Interview Takeaway

The RAG vs Fine-Tuning decision depends on whether the problem is knowledge or behavior.

---
