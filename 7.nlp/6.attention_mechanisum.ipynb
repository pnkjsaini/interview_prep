In the previous chapters we studied:

```text
Tokenization

Embeddings

Sentence Embeddings

Semantic Search
```

```text
Words → Vectors

Sentences → Vectors
```

# The Problem Before Attention

Before Transformers, NLP was dominated by:

```text
RNN

LSTM

GRU
```

These models processed text:

```text
One Word At A Time
```

---

Example:

```text
I Love Machine Learning
```

Processing order:

```text
I
 ↓

Love
 ↓

Machine
 ↓

Learning
```

---

This sequential nature created serious limitations.

---

## Interview Takeaway

RNNs process information sequentially, creating bottlenecks.

---

# The Long-Range Dependency Problem

One of the most important concepts in NLP.

---

Consider:

```text
The animal didn't cross the street because it was too tired.
```

Question:

```text
What does "it" refer to?
```

---

Answer:

```text
Animal
```

---

Notice:

```text
Animal
```

and

```text
It
```

are far apart.

---

The model must remember information over long distances.

---

RNNs struggle with this.

---

## Interview Takeaway

Long-range dependencies are difficult for sequential models.

---

# Example Of Failure

Sentence:

```text
The movie that I watched last month during my vacation in Goa was fantastic.
```

---

To understand:

```text
Fantastic
```

the model must remember:

```text
Movie
```

from much earlier.

---

Information gets diluted over time.

---

## Interview Takeaway

Important information can be forgotten in long sequences.

---

# Why LSTMs Were Not Enough

LSTMs improved RNNs.

---

They introduced:

```text
Memory Cells

Gates
```

---

But still suffered from:

```text
Sequential Processing

Limited Parallelization

Long Context Challenges
```

---

Training remained slow.

---

## Interview Takeaway

LSTMs reduced the problem but did not eliminate it.

---

# The Key Question

Researchers asked:

> Why force a model to remember everything?

---

Instead:

```text
Let The Model

Look Directly

At Relevant Words
```

---

This idea became:

```text
Attention
```

---

# What Is Attention?

## Definition

Attention allows a model to focus on the most relevant parts of the input when processing information.

---

Instead of:

```text
Remember Everything
```

---

The model learns:

```text
What To Focus On
```

---

## Interview Takeaway

Attention helps models focus on relevant information.

---

# Human Intuition

Suppose someone asks:

```text
Who won the Cricket World Cup in 2011?
```

---

Your brain focuses on:

```text
Cricket

World Cup

2011
```

---

and ignores:

```text
Who

The

In
```

---

Humans naturally use attention.

---

Neural networks learned to do the same.

---

## Interview Takeaway

Attention mimics selective focus used by humans.

---

# Core Idea Of Attention

Instead of treating every word equally:

```text
Assign Importance Scores
```

---

Example:

```text
The cat sat on the mat.
```

---

When processing:

```text
Cat
```

the model may focus strongly on:

```text
Sat

Mat
```

---

and weakly on:

```text
The
```

---

## Interview Takeaway

Attention assigns different importance to different words.

---

# Attention Visualization

Sentence:

```text
The dog chased the cat.
```

---

When processing:

```text
Cat
```

the model may attend strongly to:

```text
Dog

Chased
```

---

because they provide useful context.

---

## Interview Takeaway

Attention dynamically determines relevant context.

---

# Query, Key, And Value

The most important interview topic in attention.

---

Every token produces:

```text
Query (Q)

Key (K)

Value (V)
```

---

These are learned vectors.

---

# Why Three Vectors?

Attention answers:

```text
What Am I Looking For?
```

↓

```text
Query
```

---

```text
What Information Do Others Have?
```

↓

```text
Key
```

---

```text
What Information Should Be Returned?
```

↓

```text
Value
```

---

## Interview Takeaway

Queries search, Keys identify, Values provide information.

---

# Library Analogy

Imagine a library.

---

# Query

What book are you searching for?

---

# Key

Catalog entries.

---

# Value

Actual book content.

---

The system matches:

```text
Query

With

Keys
```

---

Then retrieves:

```text
Values
```

---

## Interview Takeaway

Attention retrieves relevant information using Q, K, and V.

---

# Step 1: Create Queries, Keys, Values

Input:

```text
Dog
```

---

Embedding:

```text
Vector
```

---

Generate:

```text
Q

K

V
```

using learned matrices.

---

## Interview Takeaway

Queries, Keys, and Values are derived from embeddings.

---

# Step 2: Calculate Similarity

Compare:

```text
Query

With

All Keys
```

---

This determines:

```text
Relevance
```

---

Higher similarity:

```text
Higher Attention
```

---

## Interview Takeaway

Attention scores are based on Query-Key similarity.

---

# Dot Product Attention

Most common approach.

---

Formula:

```text
Q · K
```

---

Large value:

```text
Highly Relevant
```

---

Small value:

```text
Less Relevant
```

---

## Interview Takeaway

Dot product measures token relevance.

---

# Attention Scores

After computing similarities:

```text
Token A → 0.9

Token B → 0.7

Token C → 0.1
```

---

These become:

```text
Attention Scores
```

---

Scores indicate:

```text
Importance
```

---

## Interview Takeaway

Attention scores represent token importance.

---

# Why Softmax Is Used

Raw scores can be:

```text
Negative

Positive

Large

Small
```

---

Need probabilities.

---

Softmax converts:

```text
Scores
```

↓

```text
Probabilities
```

---

Example:

```text
0.9

0.7

0.1
```

↓

```text
0.52

0.35

0.13
```

---

Total:

```text
1.0
```

---

## Interview Takeaway

Softmax converts attention scores into probabilities.

---

# Weighted Sum Of Values

Final step.

---

Use probabilities to weight:

```text
Values
```

---

Example:

```text
52% Value A

35% Value B

13% Value C
```

---

Combine them.

---

Result:

```text
Context-Aware Representation
```

---

## Interview Takeaway

Attention creates context-aware representations using weighted values.

---

# Scaled Dot Product Attention

Transformer attention uses:

```text
Scaled Dot Product Attention
```

---

Formula:

```text
(QKᵀ) / √d
```

---

Why divide?

---

Large dimensions produce:

```text
Huge Scores
```

---

Softmax becomes unstable.

---

Scaling fixes this.

---

## Interview Takeaway

Scaling improves numerical stability.

---

# Self-Attention

The breakthrough that changed NLP.

---

# Definition

Each word attends to:

```text
Every Other Word

Including Itself
```

---

Hence:

```text
Self-Attention
```

---

## Interview Takeaway

Self-attention allows tokens to interact directly.

---

# Example

Sentence:

```text
The dog chased the cat.
```

---

When processing:

```text
Cat
```

---

Model may attend to:

```text
Dog

Chased

Cat
```

---

simultaneously.

---

No sequential dependency.

---

## Interview Takeaway

Self-attention captures relationships across entire sequences.

---

# Why Self-Attention Is Powerful

RNN:

```text
Sequential
```

---

Self-Attention:

```text
Global
```

---

Every word can immediately access:

```text
Every Other Word
```

---

## Interview Takeaway

Self-attention eliminates long-distance communication barriers.

---

# Limitations Of Attention

Important interview topic.

---

# Memory Cost

Attention compares:

```text
Every Token

With

Every Other Token
```

---

Complexity:

```text
O(n²)
```

---

Long documents become expensive.

---

## Interview Takeaway

Attention scales quadratically with sequence length.

---

# Common Interview Questions

---

## Question

Why was attention introduced?

### Answer

```text
To solve long-range dependency problems in sequence models.
```

---

## Question

What is attention?

### Answer

```text
A mechanism that allows models to focus on relevant parts of the input.
```

---

## Question

What are Queries, Keys, and Values?

### Answer

```text
Query searches,
Key identifies,
Value provides information.
```

---

## Question

How are attention scores calculated?

### Answer

```text
Using Query-Key similarity, typically through dot products.
```

---

## Question

Why is Softmax used?

### Answer

```text
To convert attention scores into probabilities.
```

---

## Question

What is self-attention?

### Answer

```text
A mechanism where each token attends to all tokens in the sequence, including itself.
```

---

## Question

What problem does self-attention solve?

### Answer

```text
Long-range dependency and sequential processing limitations.
```

---

## Question

Why is attention important?

### Answer

```text
Because it forms the foundation of Transformer architectures and modern LLMs.
```
