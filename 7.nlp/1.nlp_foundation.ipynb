# Chapter 1: NLP Foundations

Before studying:

```text
Tokenization

Embeddings

Attention

Transformers

LLMs
```

> Why does NLP exist?

Understanding this chapter is important because every modern AI system, from ChatGPT to enterprise agents, is ultimately solving an NLP problem.

---

# What Is NLP?

## Definition

Natural Language Processing (NLP) is the field of AI that enables computers to understand, process, and generate human language.

---

## Layman Explanation

Humans communicate using:

```text
English

Hindi

French

Spanish
```

---

Computers communicate using:

```text
Numbers

Vectors

Matrices
```

---

NLP acts as the bridge between:

```text
Human Language

↓

Machine Understanding
```

---

## Interview Takeaway

NLP enables machines to understand and generate human language.

---

# Why NLP Exists

Computers are naturally good at:

```text
Mathematics

Logic

Calculations
```

---

Computers are naturally bad at:

```text
Language

Context

Meaning

Intent
```

---

Example:

```text
Apple
```

Could mean:

```text
Fruit

Company
```

---

Humans easily understand the meaning from context.

Computers struggle.

---

NLP was created to solve this problem.

---

## Interview Takeaway

NLP exists because human language is ambiguous and difficult for machines to understand.

---

# The Language Understanding Problem

Consider the sentence:

```text
I saw a man with a telescope.
```

---

Who has the telescope?

```text
You?

The Man?
```

---

The sentence is grammatically correct but ambiguous.

---

Humans use:

```text
Experience

Context

World Knowledge
```

to infer meaning.

---

Machines need algorithms to do the same.

---

## Interview Takeaway

Language contains ambiguity that machines must resolve.

---

# Why Language Is Hard

One of the most important interview discussions.

---

# Ambiguity

Same word.

Different meanings.

---

Example:

```text
Bank
```

Could mean:

```text
River Bank

Financial Bank
```

---

# Synonyms

Different words.

Same meaning.

---

Example:

```text
Car

Automobile
```

---

# Sarcasm

Example:

```text
Great Job!
```

Could mean:

```text
Praise

Or

Criticism
```

---

# Context Dependency

Example:

```text
He is cold.
```

Could mean:

```text
Temperature

Emotionally Distant
```

---

## Interview Takeaway

Language meaning depends heavily on context.

---

# The NLP Pipeline

Traditional NLP follows a pipeline.

---

```text
Raw Text
      ↓

Text Cleaning
      ↓

Tokenization
      ↓

Feature Extraction
      ↓

Model
      ↓

Prediction
```

---

Example:

```text
"The movie was amazing."
```

---

Convert text into:

```text
Tokens

Features

Vectors
```

---

Then allow the model to learn.

---

## Interview Takeaway

NLP transforms text into machine-understandable representations.

---

# Structured vs Unstructured Data

A concept frequently asked in interviews.

---

# Structured Data

Examples:

```text
Age

Salary

Country
```

Stored as:

```text
Rows

Columns
```

---

Easy for ML algorithms.

---

# Unstructured Data

Examples:

```text
Emails

Documents

Chats

Articles

PDFs
```

---

No fixed schema.

---

Much harder to process.

---

## Interview Takeaway

Language is unstructured data, making it significantly harder to analyze.

---

# Evolution Of NLP

NLP has evolved through three major eras.

---

# Era 1: Rule-Based NLP

Early systems relied on:

```text
Grammar Rules

Dictionaries

Handcrafted Logic
```

---

Example:

```text
If Word = "Good"

Then Positive Sentiment
```

---

Problem:

```text
Rules Do Not Scale
```

---

## Interview Takeaway

Rule-based systems were simple but difficult to maintain.

---

# Era 2: Statistical NLP

Machine learning entered NLP.

---

Popular techniques:

```text
Naive Bayes

SVM

Logistic Regression
```

---

Features were manually engineered.

---

Examples:

```text
Bag Of Words

TF-IDF

N-Grams
```

---

Problem:

```text
Heavy Feature Engineering
```

---

## Interview Takeaway

Statistical NLP improved performance but still depended on handcrafted features.

---

# Era 3: Deep Learning NLP

Deep Learning changed everything.

---

Instead of:

```text
Handcrafted Features
```

---

Neural networks learned:

```text
Representations Automatically
```

---

Examples:

```text
Word2Vec

LSTM

Transformer

GPT
```

---

## Interview Takeaway

Deep Learning introduced automatic representation learning.

---

# Why Traditional NLP Was Not Enough

Suppose we have:

```text
"The bank approved my loan."
```

---

Traditional NLP often relied on:

```text
Word Frequency
```

---

It struggled to understand:

```text
Meaning

Intent

Context
```

---

Modern NLP focuses on:

```text
Semantics
```

rather than:

```text
Keyword Matching
```

---

## Interview Takeaway

Modern NLP aims to understand meaning, not just words.

---

# Syntax vs Semantics

One of the most common interview topics.

---

# Syntax

Structure of language.

---

Example:

```text
Grammar

Sentence Formation
```

---

# Semantics

Meaning of language.

---

Example:

```text
What The Sentence Means
```

---

Sentence:

```text
The cat chased the dog.

The dog chased the cat.
```

---

Syntax is similar.

Meaning is very different.

---

## Interview Takeaway

Syntax focuses on structure, semantics focuses on meaning.

---

# Common Interview Questions

---

## Question

What is NLP?

### Answer

```text
NLP enables machines to understand, process, and generate human language.
```

---

## Question

Why is NLP difficult?

### Answer

```text
Because language contains ambiguity, context, and multiple meanings.
```

---

## Question

What is the difference between syntax and semantics?

### Answer

```text
Syntax is sentence structure.
Semantics is sentence meaning.
```

---

## Question

What is unstructured data?

### Answer

```text
Data without a fixed tabular format, such as text, documents, and emails.
```

---

## Question

What are common NLP tasks?

### Answer

```text
Classification

NER

Translation

Question Answering

Summarization

Information Retrieval
```

---

## Question

Why did NLP move from rule-based systems to Deep Learning?

### Answer

```text
Because handcrafted rules and features do not scale to complex language understanding tasks.
```
