# What Makes A Good Embedding?

A good embedding should place:

```text
Semantically Similar Text

Close Together
```

---

Example:

```text
How do I reset my password?
```

and

```text
How can I change my account credentials?
```

should be close.

---

While:

```text
How do I reset my password?
```

and

```text
Best pizza recipe
```

should be far apart.

---

## Interview Takeaway

Embedding quality is determined by semantic separation.

---

```text
Better Representation

Higher Accuracy
```

## Interview Takeaway

# Cosine Similarity

Most important retrieval metric.

---

Measures:

```text
Angle Between Vectors
```

---

Range:

```text
-1 To 1
```

---

Higher value:

```text
More Similar
```

---

## Interview Takeaway

Cosine similarity is the dominant metric in vector search.

---

# Dot Product

Measures:

```text
Direction

+

Magnitude
```

---

Often used in:

```text
Transformer Retrieval

ANN Search
```

---

## Interview Takeaway

Dot product is computationally efficient and widely used.

---

# Euclidean Distance

Measures:

```text
Physical Distance
```

between vectors.

---

Less common in retrieval systems.

---

## Interview Takeaway

Cosine similarity generally performs better for semantic search.

---

# Dense Retrieval

One of the most important modern NLP concepts.

---

Traditional retrieval:

```text
Keyword Matching
```

---

Dense retrieval:

```text
Embedding Matching
```

---

Pipeline:

```text
Query
      ↓

Embedding
      ↓

Similarity Search
      ↓

Top Results
```

---

## Interview Takeaway

Dense retrieval searches using embeddings rather than keywords.

---

# Sparse Retrieval vs Dense Retrieval

| Sparse Retrieval | Dense Retrieval |
|----------|----------|
| BM25 | Embeddings |
| Keyword Matching | Meaning Matching |
| Exact Terms | Semantic Similarity |
| Fast | More Accurate |

---

## Interview Takeaway

Dense retrieval focuses on semantic relevance.

---

# Bi-Encoder Architecture

Foundation of modern embedding models.

---

Architecture:

```text
Query Encoder

Document Encoder
```

---

Generate embeddings independently.

---

Example:

```text
Query
      ↓

Embedding Q
```

---

```text
Document
      ↓

Embedding D
```

---

Compare:

```text
Q

vs

D
```

---

## Interview Takeaway

Bi-encoders generate embeddings independently for efficient retrieval.

---

# Advantages Of Bi-Encoders

---

Fast Retrieval

---

Precomputed Embeddings

---

Scalable To Millions Of Documents

---

Vector Database Friendly

---

## Interview Takeaway

Bi-encoders enable scalable semantic search.

---

# Cross Encoder

Very common interview topic.

---

Instead of:

```text
Separate Encoders
```

---

Use:

```text
Query + Document
```

together.

---

Input:

```text
Query [SEP] Document
```

---

Output:

```text
Relevance Score
```

---

## Interview Takeaway

Cross encoders directly score query-document relevance.

---

# Advantages Of Cross Encoders

Higher accuracy.

---

Better ranking quality.

---

Better contextual understanding.

---

## Interview Takeaway

Cross encoders typically outperform bi-encoders in ranking quality.

---

# Disadvantages Of Cross Encoders

Need inference for:

```text
Every Query-Document Pair
```

---

Expensive.

---

Not scalable to:

```text
Millions Of Documents
```

---

## Interview Takeaway

Cross encoders are accurate but computationally expensive.

---

# Bi-Encoder vs Cross Encoder

| Topic | Bi Encoder | Cross Encoder |
|----------|----------|----------|
| Retrieval Speed | Fast | Slow |
| Scalability | Excellent | Poor |
| Ranking Quality | Good | Excellent |
| Vector DB Compatible | Yes | No |
| RAG Retrieval | Yes | No |

---

## Interview Takeaway

Bi-encoders retrieve, cross encoders rerank.

---

# Why Rerankers Exist

A critical RAG concept.

---

Vector search retrieves:

```text
Top 50 Chunks
```

---

Not all are equally relevant.

---

Need:

```text
Reranking
```

---

## Interview Takeaway

Rerankers improve retrieval precision.

---

# Reranker Pipeline

```text
Query
      ↓

Embedding Search
      ↓

Top 50 Chunks
      ↓

Cross Encoder
      ↓

Top 5 Chunks
```

---

This is the standard enterprise RAG pipeline.

---

## Interview Takeaway

Modern RAG systems combine retrieval and reranking.

---

# Recall@K

Measures:

```text
Relevant Result

Present In Top K?
```

---

Example:

```text
Recall@10
```

---

## Interview Takeaway

Recall@K measures retrieval coverage.

---

# MRR

Mean Reciprocal Rank.

---

Measures:

```text
Position Of First Relevant Result
```

---

Higher:

```text
Better
```

---

## Interview Takeaway

MRR rewards early retrieval of relevant documents.

---

# NDCG

Measures:

```text
Ranking Quality
```

---

Considers:

```text
Position

And

Relevance
```

---

Widely used in search evaluation.

---

## Interview Takeaway

NDCG evaluates ranking effectiveness.

# Common Interview Questions

---

## Question

What makes a good embedding?

### Answer

```text
One that captures semantic similarity while supporting accurate retrieval.
```

---

## Question

Why are OpenAI embedding models separate from GPT models?

### Answer

```text
Because embeddings optimize representation while GPT optimizes generation.
```

---

## Question

What is dense retrieval?

### Answer

```text
Retrieval using semantic embeddings instead of keyword matching.
```

---

## Question

What is a bi-encoder?

### Answer

```text
A model that independently embeds queries and documents.
```

---

## Question

What is a cross encoder?

### Answer

```text
A model that jointly processes query-document pairs to produce relevance scores.
```

---

## Question

Why use rerankers?

### Answer

```text
To improve retrieval precision after vector search.
```

---

## Question

Difference between retrieval and reranking?

### Answer

```text
Retrieval finds candidates.

Reranking orders candidates by relevance.
```

---

## Question

What metrics evaluate retrieval quality?

### Answer

```text
Recall@K

MRR

NDCG
```
