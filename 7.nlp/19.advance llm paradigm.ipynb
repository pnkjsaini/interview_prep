# RLHF

One of the most famous LLM techniques.

---

RLHF:

```text
Reinforcement Learning
From Human Feedback
```

---

Pipeline:

```text
Pretrained Model
       ↓

Instruction Tuning
       ↓

Human Rankings
       ↓

Reward Model
       ↓

PPO Optimization
       ↓

Aligned Model
```

---

## Interview Takeaway

RLHF aligns model outputs with human preferences.

---

# Why RLHF Works

Instead of optimizing:

```text
Next Token Prediction
```

---

Optimize:

```text
Preferred Answers
```

---

Humans rank:

```text
Response A

vs

Response B
```

---

Model learns preference patterns.

---

## Interview Takeaway

RLHF teaches models what humans prefer.

---

# Problems With RLHF

---

Expensive

---

Human Labeling Required

---

Complex Training Pipeline

---

Reward Hacking

---

## Interview Takeaway

RLHF produces strong results but is costly.

---

# RLAIF

Alternative to RLHF.

---

RLAIF:

```text
Reinforcement Learning
From AI Feedback
```

---

Instead of:

```text
Human Rankings
```

---

Use:

```text
Strong LLM
```

as evaluator.

---

## Interview Takeaway

RLAIF replaces human feedback with AI-generated feedback.

---

---

Idea:

```text
Provide Principles

To The Model
```

---

Example:

```text
Be Helpful

Avoid Harm

Be Honest
```

---

Model critiques itself.

---

Pipeline:

```text
Generate
      ↓

Critique
      ↓

Revise
```

---

## Interview Takeaway

Constitutional AI uses self-critique guided by principles.

---

# DPO

One of the hottest interview topics.

---

DPO:

```text
Direct Preference Optimization
```

---

Goal:

```text
Replace RLHF PPO Training
```

---

Traditional RLHF:

```text
Complex
```

---

DPO:

```text
Simpler

More Stable
```

---

## Interview Takeaway

DPO directly learns human preferences without reinforcement learning.

---

# Why DPO Became Popular

Advantages:

```text
Simpler Training

No Reward Model

Lower Cost

More Stable
```

---

Many modern pipelines now use:

```text
DPO
```

instead of:

```text
PPO
```

---

## Interview Takeaway

DPO is becoming the industry standard for preference alignment.

---

# ORPO

Extension of preference optimization.

---

ORPO:

```text
Odds Ratio Preference Optimization
```

---

Combines:

```text
Supervised Learning

+

Preference Learning
```

---

Single training stage.

---

## Interview Takeaway

ORPO simplifies preference optimization further.

---

# GRPO

Very important modern topic.

---

GRPO:

```text
Group Relative Policy Optimization
```

---

Used heavily in:

```text
Reasoning Models

DeepSeek R1
```

---

Instead of:

```text
Single Reward
```

---

Compare:

```text
Multiple Generated Solutions
```

---

Learn relative quality.

---

## Interview Takeaway

GRPO enables efficient reasoning-focused training.

---

# RLHF vs DPO vs GRPO

| Method | Complexity | Reward Model | Reasoning Usage |
|----------|----------|----------|----------|
| RLHF | High | Yes | Moderate |
| DPO | Medium | No | High |
| GRPO | Medium | No | Very High |

---

## Interview Takeaway

Modern reasoning models increasingly favor DPO and GRPO over PPO-based RLHF.

# Section 4: Mixture Of Experts (MoE)

One of the most important modern architecture concepts.

---

# Why MoE Exists

Scaling challenge:

```text
Larger Models
      ↓

Higher Cost
```

---

Need:

```text
Higher Capacity

Without Using All Parameters
```

---

## Interview Takeaway

MoE increases model capacity without proportional compute growth.

---

# Dense Models

Traditional architecture.

---

Example:

```text
70B Parameters
```

---

Every token uses:

```text
All 70B Parameters
```

---

## Interview Takeaway

Dense models activate all parameters for every token.

---

# Sparse Models

MoE architecture.

---

Example:

```text
400B Total Parameters
```

---

Token may use:

```text
20B Active Parameters
```

---

Only selected experts execute.

---

## Interview Takeaway

MoE activates only a subset of experts.

---

# Expert

An expert is:

```text
Specialized Feed Forward Network
```

---

Each expert learns:

```text
Different Patterns
```

---

## Interview Takeaway

Experts specialize in different regions of the problem space.

---

# Router

Most important MoE component.

---

Pipeline:

```text
Token
     ↓

Router
     ↓

Choose Experts
     ↓

Expert Outputs
```

---

Router decides:

```text
Which Experts Execute
```

---

## Interview Takeaway

The router controls expert selection.

---

# Top-K Routing

Common MoE design.

---

Example:

```text
Top 2 Experts
```

---

Router selects:

```text
Most Relevant Experts
```

---

Others remain inactive.

---

## Interview Takeaway

Top-K routing enables sparse computation.

---

# Dense vs MoE

| Topic | Dense | MoE |
|----------|----------|----------|
| Active Parameters | All | Subset |
| Capacity | Lower | Higher |
| Compute Cost | Higher | Lower |
| Routing | None | Required |

---

## Interview Takeaway

MoE separates total capacity from active computation.

---

# Common Interview Questions

---

## Question

Why is DPO replacing RLHF?

### Answer

```text
Because DPO is simpler, more stable, and avoids reward model training.
```

---

## Question

What is GRPO?

### Answer

```text
A preference optimization technique widely used in reasoning model training.
```

---

## Question

What is knowledge distillation?

### Answer

```text
Transferring knowledge from a larger teacher model to a smaller student model.
```

---

## Question

What is Chain Of Thought?

### Answer

```text
Generating intermediate reasoning steps before producing an answer.
```

---

## Question

What is Mixture Of Experts?

### Answer

```text
An architecture where only a subset of experts is activated for each input.
```

---

## Question

Why is MoE popular?

### Answer

```text
Because it increases capacity while controlling compute cost.
```

---

## Question

What is CLIP?

### Answer

```text
A multimodal model that learns aligned image and text embeddings.
```

---

## Question

How does a Vision Transformer work?

### Answer

```text
By converting image patches into tokens processed by a transformer.
```

---

## Question

What is Graph RAG?

### Answer

```text
A retrieval architecture that combines knowledge graphs and LLMs.
```
