# What Is Context Window?

## Definition

Maximum number of tokens a model can process at once.

---

Example:

```text
Prompt

Conversation History

Retrieved Documents

Generated Response
```

must fit inside:

```text
Context Window
```

---

Examples:

```text
4K

8K

32K

128K

1M+
```

---

## Interview Takeaway

Context window determines how much information a model can access simultaneously.

---

# Why Long Context Matters

Enterprise use cases:

```text
Legal Contracts

Research Papers

Code Repositories

Knowledge Bases
```

---

Need:

```text
Large Context Windows
```

---

Without long context:

```text
Information Loss
```

---

## Interview Takeaway

Long-context models reduce the need to aggressively truncate information.

---

# The Transformer Context Problem

Recall self-attention:

```text
Every Token

Attends To

Every Other Token
```

---

For N tokens:

```text
Attention Complexity

O(N²)
```

---

Examples:

```text
1K Tokens
```

↓

```text
1 Million Attention Scores
```

---

```text
100K Tokens
```

↓

```text
10 Billion Attention Scores
```

---

Explodes rapidly.

---

## Interview Takeaway

Quadratic attention complexity limits context length.

---

# Why O(N²) Is A Problem

As context grows:

```text
Memory Usage ↑

Latency ↑

GPU Cost ↑
```

---

Long context becomes expensive.

---

## Interview Takeaway

Attention computation becomes the bottleneck for long sequences.

---

# Positional Encoding Review

Transformers must know:

```text
Token Position
```

---

Otherwise:

```text
Dog Bites Man

Man Bites Dog
```

appear identical.

---

Need:

```text
Positional Information
```

---

## Interview Takeaway

Position information enables sequence understanding.

---

# RoPE (Rotary Position Embeddings)

One of the most important modern interview topics.

---

Most modern LLMs use:

```text
RoPE
```

---

Instead of traditional positional encodings.

---

Examples:

```text
Llama

Qwen

DeepSeek

Mistral
```

---

## Interview Takeaway

RoPE is the dominant positional encoding mechanism in modern LLMs.

---

# Intuition Behind RoPE

Instead of:

```text
Adding Position Information
```

---

RoPE:

```text
Rotates Embedding Vectors
```

based on token position.

---

Benefits:

```text
Relative Position Awareness
```

---

## Interview Takeaway

RoPE encodes position through rotational transformations.

---

# Why RoPE Became Popular

Advantages:

```text
Better Extrapolation

Relative Position Encoding

Efficient Computation
```

---

Compared to:

```text
Learned Position Embeddings
```

---

## Interview Takeaway

RoPE generalizes better to unseen sequence lengths.

---

# RoPE Limitation

Models trained at:

```text
8K Context
```

may struggle at:

```text
128K Context
```

---

Problem:

```text
Position Distribution Shift
```

---

Need:

```text
RoPE Scaling
```

---

## Interview Takeaway

RoPE does not automatically generalize to arbitrarily long contexts.

---

# Position Interpolation

Early long-context technique.

---

Idea:

```text
Compress Positions
```

---

Example:

```text
128K Tokens

↓

Fit Into

32K Positional Space
```

---

Benefits:

```text
Longer Context

Without Retraining
```

---

## Interview Takeaway

Position interpolation extends context by rescaling positions.

---

# Sliding Window Attention

Popular in:

```text
Mistral

Long Context Models
```

---

Idea:

```text
Attend To Nearby Tokens
```

instead of:

```text
Entire Context
```

---

Example:

```text
Window = 4096
```

---

Each token attends only locally.

---

## Interview Takeaway

Sliding window attention reduces attention complexity.

---

# Benefits Of Sliding Windows

```text
Lower Memory

Lower Compute

Longer Context Support
```

---

Tradeoff:

```text
Reduced Global Awareness
```

---

## Interview Takeaway

Sliding windows improve scalability by limiting attention scope.

---

# Mixture Of Experts (MoE)

One of the most important modern architecture concepts.

---

Traditional transformer:

```text
All Parameters

Used Every Token
```

---

MoE:

```text
Only Some Experts

Activated
```

---

## Interview Takeaway

MoE increases model capacity without proportional compute cost.

---

# Intuition Behind MoE

Think:

```text
Hospital
```

---

Instead of:

```text
One Doctor
```

---

Use:

```text
Many Specialists
```

---

Router chooses:

```text
Relevant Expert
```

---

## Interview Takeaway

MoE routes inputs to specialized subnetworks.

---

# MoE Architecture

```text
Input
     ↓

Router
     ↓

Expert Selection
     ↓

Selected Experts
     ↓

Output
```

---

Not all experts execute.

---

## Interview Takeaway

MoE uses sparse activation to improve efficiency.

---

# Dense Model vs MoE

Example:

```text
70B Dense Model
```

---

Every token uses:

```text
70B Parameters
```

---

MoE:

```text
200B Total Parameters
```

---

Token may use:

```text
20B Active Parameters
```

---

Benefits:

```text
Large Capacity

Lower Compute
```

---

## Interview Takeaway

MoE separates total parameters from active parameters.

---

# Common Interview Questions

---

## Question

What is a context window?

### Answer

```text
The maximum number of tokens a model can process simultaneously.
```

---

## Question

Why is transformer attention expensive?

### Answer

```text
Because attention complexity grows quadratically with sequence length.
```

---

## Question

What is RoPE?

### Answer

```text
Rotary Position Embeddings, a positional encoding method used in modern LLMs.
```

---

## Question

Why do we need RoPE scaling?

### Answer

```text
Because models trained on shorter contexts often struggle with much longer sequences.
```

---

## Question

What is Flash Attention?

### Answer

```text
A memory-efficient attention algorithm that accelerates transformer inference and training.
```

---

## Question

What is Sliding Window Attention?

### Answer

```text
A technique where tokens attend only to nearby tokens instead of the full sequence.
```

---

## Question

What is Mamba?

### Answer

```text
A state-space architecture designed for efficient long-sequence processing.
```

---

## Question

What is Mixture of Experts?

### Answer

```text
An architecture that activates only a subset of experts for each input.
```

---

## Question

Why is MoE popular?

### Answer

```text
Because it increases model capacity while controlling compute costs.
```

---

## Question

How does DeepSeek use MoE?

### Answer

```text
Through sparse expert routing that activates only a fraction of total parameters.
```
