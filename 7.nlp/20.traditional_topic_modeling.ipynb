# Part 1: Information Retrieval Foundations

---

# What Is Information Retrieval?

Goal:

```text
Given A Query

Retrieve Relevant Documents
```

---

Example:

```text
Query:
Machine Learning Interview Questions
```

↓

```text
Relevant Documents
```

---

## Database vs Information Retrieval

Database:

```sql
SELECT * FROM employee
WHERE id = 10
```

---

Exact matching.

---

Information Retrieval:

```text
Approximate Matching

Relevance Ranking
```

---

## Interview Takeaway

Information Retrieval focuses on relevance rather than exact matches.

---

# Inverted Index

The foundation of search engines.

---

Instead of:

```text
Document
    ↓
Words
```

Store:

```text
Word
    ↓
Documents
```

---

Example:

```text
Machine
    ↓
Doc1
Doc3
Doc5
```

---

```text
Learning
    ↓
Doc1
Doc2
Doc5
```

---

## Interview Takeaway

An inverted index maps terms to documents containing them.

---

# TF-IDF

TF-IDF:

```text
Term Frequency

Inverse Document Frequency
```

---

Goal:

```text
Measure Word Importance
```

---

# Term Frequency (TF)

Measures:

```text
How Frequently
A Word Appears
```

---

# Inverse Document Frequency (IDF)

Measures:

```text
How Rare
A Word Is
```

---

High TF-IDF:

```text
Frequent In Document

Rare Across Corpus
```

---

## Interview Takeaway

TF-IDF identifies important document-specific words.

---

# Limitations Of TF-IDF

Cannot understand:

```text
Meaning

Context

Synonyms
```

---

Example:

```text
Car

Automobile
```

treated differently.

---

## Interview Takeaway

TF-IDF captures frequency, not semantics.

---

# BM25

Most important traditional retrieval algorithm.

---

Improves TF-IDF using:

```text
Term Saturation

Length Normalization
```

---

# Term Saturation

Problem:

```text
Machine

Machine

Machine

Machine
```

Repeated terms shouldn't increase importance forever.

---

BM25 models:

```text
Diminishing Returns
```

---

## Interview Takeaway

BM25 prevents repeated words from dominating rankings.

---

# Document Length Normalization

Long documents naturally contain more words.

---

BM25 compensates for:

```text
Document Length
```

---

## Interview Takeaway

BM25 prevents long documents from receiving unfairly high scores.

---

# BM25 vs Embeddings

BM25:

```text
Keyword Matching
```

---

Embeddings:

```text
Semantic Matching
```

---

Example:

```text
Car

Automobile
```

---

BM25:

```text
May Miss Match
```

---

Embeddings:

```text
Likely Match
```

---

## Interview Takeaway

BM25 is lexical while embeddings are semantic.

---

# Hybrid Search

Modern systems combine:

```text
BM25

+

Embeddings
```

---

Benefits:

```text
Exact Match

+

Semantic Understanding
```

---

## Interview Takeaway

Hybrid search often outperforms pure vector search.

---

# Elasticsearch & Lucene

Built on:

```text
Inverted Index

BM25
```

---

Components:

```text
Analyzer

Tokenizer

Index

Retriever
```

---

## Interview Takeaway

Elasticsearch uses BM25 and inverted indexes for retrieval.

---

# Part 2: Traditional NLP Pipeline

---

Before transformers:

```text
Text
   ↓

Cleaning
   ↓

Tokenization
   ↓

Stopword Removal
   ↓

Stemming
   ↓

Lemmatization
   ↓

Feature Extraction
   ↓

Classifier
```

---

# Stemming

Reduce words to root forms.

---

Example:

```text
Playing

Played

Player
```

↓

```text
Play
```

---

Algorithms:

```text
Porter Stemmer

Snowball Stemmer
```

---

## Interview Takeaway

Stemming removes suffixes using heuristics.

---

# Lemmatization

Uses:

```text
Dictionary

Grammar Rules
```

---

Example:

```text
Better
```

↓

```text
Good
```

---

## Interview Takeaway

Lemmatization produces linguistically valid root words.

---

# Stemming vs Lemmatization

| Topic | Stemming | Lemmatization |
|----------|----------|----------|
| Speed | Faster | Slower |
| Accuracy | Lower | Higher |
| Dictionary | No | Yes |

---

## Interview Takeaway

Lemmatization is more accurate but computationally heavier.

---

# Part 3: Sequence Labeling

---

# What Is Sequence Labeling?

Assign labels to tokens.

---

Examples:

```text
POS Tagging

NER

Chunking
```

---

## Interview Takeaway

Sequence labeling predicts labels for each token.

---

# Part Of Speech (POS) Tagging

Assign grammatical categories.

---

Example:

```text
Dog
```

↓

```text
Noun
```

---

Example:

```text
Runs
```

↓

```text
Verb
```

---

## Interview Takeaway

POS tagging identifies grammatical roles.

---

# Named Entity Recognition (NER)

Identify entities.

---

Example:

```text
Satya Nadella
```

↓

```text
PERSON
```

---

Example:

```text
Microsoft
```

↓

```text
ORGANIZATION
```

---

## Interview Takeaway

NER extracts structured information from text.

---

# BIO Tagging

Example:

```text
Satya → B-PER

Nadella → I-PER

works → O
```

---

## Interview Takeaway

BIO tagging identifies entity boundaries.

---

# Hidden Markov Models (HMM)

Classical sequence model.

---

Components:

```text
States

Transitions

Emissions
```

---

Used historically for:

```text
POS Tagging
```

---

## Interview Takeaway

HMMs model sequential dependencies probabilistically.

---

# Conditional Random Fields (CRF)

Improved sequence modeling.

---

Advantages:

```text
Label Dependencies

Higher Accuracy
```

---

Popular for:

```text
NER
```

---

## Interview Takeaway

CRFs model relationships between neighboring labels.

---

# BiLSTM-CRF

State-of-the-art before transformers.

---

Pipeline:

```text
Sentence
      ↓

BiLSTM
      ↓

CRF
      ↓

Labels
```

---

## Interview Takeaway

BiLSTM-CRF dominated sequence labeling before BERT.

---

# Part 4: Topic Modeling

---

# What Is Topic Modeling?

Goal:

```text
Automatically Discover Topics
```

---

Without labels.

---

## Interview Takeaway

Topic modeling discovers hidden themes in documents.

---

# Latent Semantic Analysis (LSA)

Pipeline:

```text
Term Document Matrix
      ↓

SVD
      ↓

Latent Topics
```

---

Connection:

```text
Module 3

SVD
```

---

## Interview Takeaway

LSA uses SVD to discover latent semantic structure.

---

# Probabilistic Latent Semantic Analysis (PLSA)

Represents:

```text
Documents

As

Topic Mixtures
```

---

## Interview Takeaway

PLSA introduced probabilistic topic modeling.

---

# Latent Dirichlet Allocation (LDA)

Most important topic modeling algorithm.

---

Assumption:

```text
Document
     ↓

Mixture Of Topics
```

---

Topic:

```text
Probability Distribution
Over Words
```

---

## Interview Takeaway

LDA models documents as mixtures of latent topics.

---

# Dirichlet Distribution

Used to generate:

```text
Topic Distributions
```

---

Properties:

```text
Probabilities Sum To One
```

---

## Interview Takeaway

Dirichlet distributions are priors over probability distributions.

---

# LDA Generative Process

```text
Choose Topic Distribution
          ↓

Choose Topic
          ↓

Choose Word
```

---

Repeat.

---

## Interview Takeaway

LDA assumes documents are generated from topic-word distributions.

---

# LDA vs Clustering

Clustering:

```text
One Document
      ↓

One Cluster
```

---

LDA:

```text
One Document
      ↓

Multiple Topics
```

---

## Interview Takeaway

LDA provides soft topic assignments.

---

# Part 4.5: Document Clustering & Modern Topic Discovery

One of the most important additions for modern NLP.

---

# Why Clustering Appears In NLP

Examples:

```text
Customer Reviews

Support Tickets

Research Papers

News Articles
```

Need:

```text
Automatic Topic Discovery
```

---

# K-Means For Document Clustering

Pipeline:

```text
Documents
      ↓

Embeddings
      ↓

K-Means
      ↓

Clusters
```

---

Problems:

```text
Need K

Sensitive To Outliers

Assumes Spherical Clusters
```

---

## Interview Takeaway

K-Means is simple but often struggles with real embedding spaces.

---

# DBSCAN

Density Based Spatial Clustering Of Applications With Noise.

---

Core Concepts:

```text
Core Point

Border Point

Noise Point
```

---

Hyperparameters:

```text
eps

min_samples
```

---

Advantages:

```text
No Need For K

Detects Outliers

Arbitrary Cluster Shapes
```

---

## NLP Use Cases

```text
Review Clustering

Topic Discovery

Outlier Detection
```

---

## Interview Takeaway

DBSCAN discovers clusters based on density rather than distance to centroids.

---

# DBSCAN vs K-Means

| Topic | K-Means | DBSCAN |
|----------|----------|----------|
| Need K | Yes | No |
| Outlier Detection | No | Yes |
| Arbitrary Shapes | No | Yes |
| Density Based | No | Yes |

---

# HDBSCAN

Hierarchical DBSCAN.

---

Why?

DBSCAN assumes:

```text
Single Density Threshold
```

---

Real embedding spaces contain:

```text
Dense Topics

Sparse Topics
```

simultaneously.

---

HDBSCAN handles:

```text
Variable Density Clusters
```

---

Advantages:

```text
Better Topic Discovery

Less Parameter Tuning

Handles Embeddings Well
```

---

## Interview Takeaway

HDBSCAN is often preferred over DBSCAN for embedding clustering.

---

# BERTopic

Very important modern topic.

---

Pipeline:

```text
Documents
      ↓

Sentence Transformer
      ↓

Embeddings
      ↓

UMAP
      ↓

HDBSCAN
      ↓

Topic Extraction
```

---

Notice:

```text
No LDA
```

---

Modern topic modeling increasingly uses:

```text
Embeddings

+

HDBSCAN
```

---

## Interview Takeaway

BERTopic combines embeddings and HDBSCAN for modern topic discovery.

---

# LDA vs BERTopic

| Topic | LDA | BERTopic |
|----------|----------|----------|
| Input | Word Counts | Embeddings |
| Semantics | Weak | Strong |
| Modern Usage | Lower | Higher |
| Transformer Compatible | No | Yes |

---

## Interview Takeaway

BERTopic largely outperforms LDA on semantic topic discovery.

---

Represent documents as:

```text
Word Count Vectors
```

---

Ignores:

```text
Order

Meaning

Context
```

---

## Interview Takeaway

BoW creates sparse frequency-based representations.

---

# Naive Bayes

Popular for:

```text
Spam Detection

Sentiment Analysis
```

---

Assumption:

```text
Feature Independence
```

---

## Interview Takeaway

Naive Bayes was a strong traditional NLP baseline.

---

# Logistic Regression

Pipeline:

```text
TF-IDF
      ↓

Logistic Regression
```

---

## Interview Takeaway

TF-IDF plus Logistic Regression was a common production NLP pipeline.

---

# SVM For Text

Before BERT:

```text
SVM

Was Often State Of The Art
```

---

Works well with:

```text
Sparse Features
```

---

## Interview Takeaway

SVMs were highly effective for traditional text classification.

---

# Why Transformers Won

Traditional NLP suffered from:

```text
Feature Engineering

Sparse Features

Poor Context Understanding

Weak Semantics
```

---

Transformers introduced:

```text
Contextual Embeddings

Transfer Learning

Semantic Understanding
```

---

## Interview Takeaway

Transformers replaced handcrafted features with learned representations.
