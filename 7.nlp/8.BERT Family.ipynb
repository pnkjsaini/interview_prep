# Chapter 8: BERT Family

In the previous chapter we studied:

```text
Transformer Architecture

Encoder

Decoder

Multi-Head Attention

Positional Encoding

Causal Masking
```

Now we answer a very important question:

> What happens if we use only the Encoder portion of a Transformer?

```text
BERT
```

```text
Embeddings

Semantic Search

Sentence Transformers

RAG Retrieval
```

# Why BERT Exists

Before BERT, most NLP systems relied on:

```text
Word2Vec

FastText

LSTM
```

---

Problems:

```text
Limited Context

Static Embeddings

Weak Language Understanding
```

---

Example:

```text
Bank
```

Word2Vec always assigns:

```text
One Vector
```

---

But:

```text
River Bank

Financial Bank
```

have different meanings.

---

Need:

```text
Contextual Understanding
```

---

## Interview Takeaway

BERT was designed to learn context-dependent language representations.

---

# What Is BERT?

BERT stands for:

```text
Bidirectional Encoder Representations from Transformers
```

---

Released by:

```text
Google

2018
```

---

Built using:

```text
Transformer Encoder
```

only.

---

No decoder.

---

## Interview Takeaway

BERT is an encoder-only transformer model.

---

# Why "Bidirectional"?

One of the most frequently asked interview questions.

---

Traditional models read:

```text
Left → Right
```

---

Example:

```text
The bank approved my loan.
```

---

When processing:

```text
Bank
```

BERT sees:

```text
The

Approved

My

Loan
```

simultaneously.

---

Both:

```text
Left Context

Right Context
```

---

This is called:

```text
Bidirectional Understanding
```

---

## Interview Takeaway

BERT understands context from both directions simultaneously.

---

# BERT Architecture

High-level flow:

```text
Text
     ↓

Tokenization
     ↓

Embeddings
     ↓

Positional Encoding
     ↓

Transformer Encoder Layers
     ↓

Contextual Representations
```

---

Core building block:

```text
Encoder
```

---

No:

```text
Text Generation
```

capability.

---

## Interview Takeaway

BERT focuses on understanding rather than generation.

---

# Input Representation In BERT

BERT uses:

```text
Token Embeddings

+

Position Embeddings

+

Segment Embeddings
```

---

Combined representation:

```text
Final Input Embedding
```

---

## Interview Takeaway

BERT combines token, position, and segment information.

---

# Masked Language Modeling (MLM)

The most important BERT interview topic.

---

# Why MLM Exists

GPT learns:

```text
Next Token Prediction
```

---

BERT learns differently.

---

Example:

```text
The cat sat on the [MASK].
```

---

Goal:

```text
Predict Missing Word
```

---

Expected output:

```text
Mat
```

---

## Interview Takeaway

BERT is trained using Masked Language Modeling.

---

# Why Masking Works

To predict:

```text
[MASK]
```

the model must understand:

```text
Entire Sentence
```

---

Not just previous words.

---

This encourages:

```text
Deep Contextual Understanding
```

---

## Interview Takeaway

Masking forces BERT to learn language understanding.

---

# Next Sentence Prediction (NSP)

Original BERT training task.

---

Example:

Sentence A:

```text
I love cricket.
```

---

Sentence B:

```text
Virat Kohli scored a century.
```

---

Question:

```text
Does Sentence B Follow Sentence A?
```

---

Model predicts:

```text
Yes / No
```

---

Purpose:

```text
Learn Relationships Between Sentences
```

---

## Interview Takeaway

NSP helps BERT learn inter-sentence relationships.

---

# Contextual Embeddings

One of BERT's biggest innovations.

---

Word:

```text
Bank
```

---

Sentence 1:

```text
I deposited money in the bank.
```

---

Sentence 2:

```text
The boat reached the river bank.
```

---

Word2Vec:

```text
Same Embedding
```

---

BERT:

```text
Different Embeddings
```

---

because context differs.

---

## Interview Takeaway

BERT generates context-dependent embeddings.

---

# Why Contextual Embeddings Matter

Enable:

```text
Better Search

Better Retrieval

Better Classification

Better Understanding
```

---

This directly influenced:

```text
Modern Embedding Models
```

---

## Interview Takeaway

Contextual embeddings dramatically improved NLP performance.

---

# Common Interview Questions

---

## Question

What does BERT stand for?

### Answer

```text
Bidirectional Encoder Representations from Transformers.
```

---

## Question

Why is BERT bidirectional?

### Answer

```text
Because it attends to both left and right context simultaneously.
```

---

## Question

How is BERT trained?

### Answer

```text
Using Masked Language Modeling.
```

---

## Question

What is the purpose of the CLS token?

### Answer

```text
To represent the entire sentence for classification tasks.
```

---

## Question

What is the purpose of the SEP token?

### Answer

```text
To separate sentences or segments.
```

---

## Question

Why are contextual embeddings important?

### Answer

```text
Because word meaning depends on context.
```

---

## Question

Why was SBERT created?

### Answer

```text
To make semantic similarity search efficient.
```

---

## Question

Difference between BERT and GPT?

### Answer

```text
BERT is encoder-based and optimized for understanding.

GPT is decoder-based and optimized for generation.
```
