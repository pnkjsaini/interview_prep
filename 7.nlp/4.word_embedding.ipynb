# Module 7: NLP For LLM Engineers
# Chapter 4: Word Embeddings
In the previous chapter we studied:
```text
Tokenization
BPE
WordPiece
SentencePiece
Token IDs
Context Windows
```
At the end of tokenization, we have:
```text
Text
    ↓
Tokens
    ↓
Token IDs
```
Example:
```text
Cat → 102
Dog → 215
Car → 387
```
But a major question remains:
> How does a neural network understand that "Cat" and "Dog" are more similar than "Cat" and "Car"?
This is the problem that embeddings solve.
---
# Why Embeddings Exist
Computers understand:
```text
Numbers
```
---
Humans understand:
```text
Meaning
```
---
Token IDs are just numbers.
---
Example:
```text
Cat → 102
Dog → 215
Car → 387
```
---
Does this mean:
```text
Dog Is Twice Cat?
Car Is Bigger Than Dog?
```
---
No.
These numbers contain:
```text
No Semantic Meaning
```
---
## Interview Takeaway
Token IDs identify words but do not represent meaning.
---
# The Meaning Problem
Consider:
```text
King
Queen
Prince
Princess
```
Humans instantly recognize:
```text
Related Concepts
```
---
Token IDs cannot capture this relationship.
---
Neural networks need:
```text
Meaningful Numerical Representations
```
---
These are called:
```text
Embeddings
```
---
## Interview Takeaway
Embeddings convert words into vectors that capture meaning.
---
# One-Hot Encoding
Before embeddings existed, NLP relied heavily on:
```text
One-Hot Encoding
```
---
# What Is One-Hot Encoding?
Suppose vocabulary contains:
```text
Cat
Dog
Car
```
---
Representations:
```text
Cat → [1,0,0]
Dog → [0,1,0]
Car → [0,0,1]
```
---
Only one position contains:
```text
1
```
All others contain:
```text
0
```
---
## Interview Takeaway
One-hot encoding represents words using sparse binary vectors.
---
# Advantages Of One-Hot Encoding
## Simple
Easy to understand.
---
## Unique Representation
Every word gets a unique vector.
---
## Deterministic
No training required.
---
# Problems With One-Hot Encoding
One of the most important interview topics.
---
# Problem 1: No Similarity Information
Consider:
```text
King
Queen
```
---
Vectors:
```text
[1,0,0]
[0,1,0]
```
---
Distance:
```text
Exactly Same As
King ↔ Banana
```
---
Model sees:
```text
No Relationship
```
---
## Interview Takeaway
One-hot vectors do not capture semantic similarity.
---
# Problem 2: Sparse Representation
Suppose vocabulary:
```text
100,000 Words
```
---
Each vector contains:
```text
99,999 Zeros
```
---
Wasteful.
---
## Interview Takeaway
One-hot encoding is memory inefficient.
---
# Problem 3: Curse Of Dimensionality
Vocabulary:
```text
1 Million Words
```
---
Vector Dimension:
```text
1 Million
```
---
Extremely large.
---
Training becomes expensive.
---
## Interview Takeaway
Large vocabularies create massive sparse vectors.
---
# What Is An Embedding?
## Definition
An embedding is a dense numerical representation of a word.
---
Instead of:
```text
[0,0,0,0,0,1,0,0,0]
```
---
Use:
```text
[0.32, -0.17, 0.89, 0.12]
```
---
These numbers are:
```text
Learned
```
during training.
---
## Interview Takeaway
Embeddings are dense learned vector representations.
---
# Core Idea Of Embeddings
Words with similar meanings should have:
```text
Similar Vectors
```
---
Example:
```text
King
Queen
```
should be closer than:
```text
King
Banana
```
---
This is the foundation of:
```text
Semantic Understanding
```
---
## Interview Takeaway
Embeddings place semantically similar words near each other.
---
# Embedding Space
One of the most important concepts in NLP.
---
# Definition
A geometric space where words are represented as vectors.
---
Example:
```text
Dog
Cat
Tiger
Lion
```
may appear near each other.
---
While:
```text
Car
Airplane
Bicycle
```
form another region.
---
## Interview Takeaway
Embedding space organizes words according to meaning.
---
# Visual Intuition
Imagine a map.
---
Animals cluster together:
```text
Dog
Cat
Lion
Tiger
```
---
Countries cluster together:
```text
India
China
Japan
```
---
Vehicles cluster together:
```text
Car
Bus
Train
```
---
Embeddings learn these relationships automatically.
---
## Interview Takeaway
Embedding spaces naturally form semantic neighborhoods.
---
# Semantic Similarity
A concept used everywhere in:
```text
RAG
Vector Search
Recommendation Systems
```
---
# Definition
The degree to which two pieces of text have similar meaning.
---
Example:
```text
Dog
Puppy
```
---
High similarity.
---
Example:
```text
Dog
Airplane
```
---
Low similarity.
---
## Interview Takeaway
Semantic similarity measures meaning rather than exact words.
---
# Measuring Similarity
Common techniques:
```text
Cosine Similarity
Dot Product
Euclidean Distance
```
---
Most common:
```text
Cosine Similarity
```
---
# Why Cosine Similarity?
Focuses on:
```text
Direction
```
rather than magnitude.
---
Widely used in:
```text
Embeddings
RAG
Vector Databases
```
---
## Interview Takeaway
Cosine similarity is the most common embedding similarity metric.
---
# Word2Vec
One of the most important NLP breakthroughs.
---
Developed by:
```text
Google
2013
```
---
Goal:
```text
Learn Embeddings Automatically
```
---
Instead of manually designing vectors.
---
## Interview Takeaway
Word2Vec learns semantic embeddings from text.
---
# Core Idea Of Word2Vec
Words appearing in similar contexts tend to have similar meanings.
---
Example:
```text
The dog barked.
The puppy barked.
```
---
Model learns:
```text
Dog ≈ Puppy
```
---
without human supervision.
---
## Interview Takeaway
Word2Vec learns meaning from surrounding context.
---
# Word2Vec Architectures
Two major variants:
```text
CBOW
Skip-Gram
```
---
# CBOW
Continuous Bag Of Words.
---
# Goal
Predict:
```text
Target Word
```
using surrounding words.
---
Example:
```text
The ____ barked loudly.
```
---
Predict:
```text
Dog
```
---
## Interview Takeaway
CBOW predicts a word from its context.
---
# Skip-Gram
The opposite approach.
---
# Goal
Predict:
```text
Context Words
```
from a target word.
---
Example:
```text
Dog
```
---
Predict:
```text
Pet
Animal
Barked
```
---
## Interview Takeaway
Skip-Gram predicts surrounding words from a target word.
---
# CBOW vs Skip-Gram
| Topic | CBOW | Skip-Gram |
|----------|----------|----------|
| Predicts | Word | Context |
| Speed | Faster | Slower |
| Rare Words | Weaker | Better |
| Data Requirement | Lower | Higher |
---
## Interview Takeaway
Skip-Gram generally performs better on rare words.
---
# Negative Sampling
Frequently asked in interviews.
---
# Problem
Vocabulary may contain:
```text
100,000+
Words
```
---
Training becomes expensive.
---
# Solution
Instead of learning from:
```text
All Words
```
learn from:
```text
Small Number Of
Positive Examples
Negative Examples
```
---
Result:
```text
Much Faster Training
```
---
## Interview Takeaway
Negative sampling makes Word2Vec scalable.
---
# Famous Word2Vec Property
One of the most famous AI demonstrations.
---
Vector arithmetic:
```text
King
-
Man
+
Woman
=
Queen
```
---
Model learns:
```text
Gender Relationships
Country Relationships
Semantic Structure
```
---
without explicit rules.
---
## Interview Takeaway
Word2Vec embeddings capture meaningful semantic relationships.
---
# Limitations Of Word2Vec
Critical interview topic.
---
# Static Embeddings
Each word receives:
```text
One Vector
```
---
Example:
```text
Bank
```
always receives:
```text
Same Embedding
```
---
Problem:
```text
River Bank
Financial Bank
```
have different meanings.
---
Word2Vec cannot distinguish them.
---
## Interview Takeaway
Word2Vec generates context-independent embeddings.
---
# FastText
Developed by:
```text
Meta (Facebook)
```
---
# Why FastText Exists
Word2Vec struggles with:
```text
Rare Words
Misspellings
OOV Words
```
---
FastText solves this using:
```text
Subwords
```
---
Example:
```text
ChatGPT
```
---
Can be decomposed into smaller pieces.
---
Model can still create useful embeddings.
---
## Interview Takeaway
FastText improves handling of unseen and rare words.
---
# Word2Vec vs FastText
| Topic | Word2Vec | FastText |
|----------|----------|----------|
| OOV Handling | Weak | Strong |
| Subwords | No | Yes |
| Rare Words | Weak | Strong |
| Speed | Faster | Slightly Slower |
---
## Interview Takeaway
FastText extends Word2Vec using character-level information.
---

# Semantic Search

---
# RAG

# Recommendation Systems

represented in shared embedding spaces.

# Agent Memory
Modern agents store memories as embeddings.

Embeddings are the foundation of modern retrieval systems.
