# Chapter 14: Advanced RAG Architectures

In the previous chapter we studied:

```text
Basic RAG

Chunking

Retrieval

Reranking

Prompt Augmentation

Grounding
```

Basic RAG works well.

However, enterprise AI systems face new challenges:

```text
Complex Questions

Multi-Hop Reasoning

Poor Retrieval

Knowledge Graphs

Agentic Workflows
```

To solve these problems, advanced RAG architectures emerged.

---

# Why Basic RAG Is Not Enough

Basic RAG assumes:

```text
One Query

↓

One Retrieval

↓

One Answer
```

---

Real-world questions are often more complex.

---

Example:

```text
Which Databricks feature can help reduce retrieval latency
for documents stored in Unity Catalog?
```

To answer:

```text
Need Multiple Pieces Of Information
```

---

Basic retrieval may fail.

---

## Interview Takeaway

Complex questions often require more than a single retrieval step.

---

# Query Transformation

One of the most common advanced RAG techniques.

---

# Problem

Users ask poor questions.

---

Example:

```text
Reset Password
```

---

Retriever may struggle.

---

Transform into:

```text
How Can I Change My Account Credentials?
```

---

Retrieval improves.

---

## Interview Takeaway

Query transformation improves retrieval quality.

---

# Query Expansion

Expand the query with related concepts.

---

Example:

```text
Reset Password
```

↓

```text
Reset Password

Change Credentials

Account Recovery

Authentication
```

---

Benefits:

```text
Higher Recall
```

---

## Interview Takeaway

Query expansion improves retrieval coverage.

---

# Multi-Query Retrieval

A very common interview topic.

---

Instead of:

```text
One Query
```

Generate:

```text
Multiple Queries
```

---

Example:

```text
How To Reset Password?

How To Recover Account?

How To Change Credentials?
```

---

Retrieve from all.

---

Merge results.

---

Benefits:

```text
Improved Recall

Reduced Retrieval Misses
```

---

## Interview Takeaway

Multi-query retrieval improves document coverage.

---

# HyDE (Hypothetical Document Embeddings)

One of the most important modern RAG techniques.

---

# Problem

Short queries often have weak embeddings.

---

Example:

```text
Databricks Governance
```

---

Embedding may lack context.

---

# Solution

Generate:

```text
Hypothetical Answer
```

first.

---

Example:

```text
Databricks governance involves Unity Catalog...
```

---

Embed the generated text.

---

Then retrieve.

---

## Interview Takeaway

HyDE improves retrieval by embedding hypothetical answers instead of raw queries.

---

# Why HyDE Works

Generated text contains:

```text
Context

Keywords

Semantic Signals
```

---

Retriever performs better.

---

## Interview Takeaway

HyDE enriches sparse queries before retrieval.

---

# Multi-Hop Retrieval

Extremely common senior interview topic.

---

# Definition

Answering a question requiring multiple retrieval steps.

---

Example:

```text
Who is the CEO of the company that acquired GitHub?
```

---

Need:

```text
Step 1:
Who Acquired GitHub?

↓

Microsoft

Step 2:
Who Is Microsoft's CEO?

↓

Satya Nadella
```

---

Basic RAG often struggles.

---

## Interview Takeaway

Multi-hop retrieval requires retrieving multiple related facts.

---

# Multi-Hop RAG Pipeline

```text
Question
      ↓

Retrieve Fact A
      ↓

Generate Follow-Up Query
      ↓

Retrieve Fact B
      ↓

Combine Information
      ↓

Answer
```

---

## Interview Takeaway

Multi-hop RAG decomposes complex reasoning into retrieval steps.

---

# Parent-Child Retrieval

A highly practical enterprise technique.

---

Problem:

```text
Small Chunks

Good Precision

Poor Context
```

---

Large Chunks:

```text
Good Context

Poor Precision
```

---

Solution:

```text
Retrieve Small Child Chunk

Return Large Parent Chunk
```

---

Best of both worlds.

---

## Interview Takeaway

Parent-child retrieval balances precision and context.

---

# Contextual Compression

Very important production topic.

---

Problem:

```text
Retrieved Chunks

Too Large
```

---

Waste context window.

---

Solution:

```text
Compress Retrieved Content
```

before generation.

---

Methods:

```text
Summarization

Extraction

Filtering
```

---

## Interview Takeaway

Contextual compression reduces prompt size while preserving relevance.

---

# Corrective RAG (CRAG)

Modern enterprise pattern.

---

Problem:

```text
Retriever May Fail
```

---

Solution:

```text
Evaluate Retrieval Quality
```

before generation.

---

If retrieval is weak:

```text
Retry

Reformulate Query

Search Again
```

---

## Interview Takeaway

CRAG improves reliability through retrieval validation.

---

# Self-RAG

One of the most important modern topics.

---

Traditional RAG:

```text
Always Retrieve
```

---

Self-RAG:

```text
Model Decides

Whether Retrieval Is Needed
```

---

Examples:

```text
What Is Python?
```

No retrieval.

---

Example:

```text
What Is Our Company's Leave Policy?
```

Retrieve.

---

## Interview Takeaway

Self-RAG dynamically determines when retrieval is necessary.

---

# Adaptive RAG

Extension of Self-RAG.

---

System decides:

```text
Simple Question
↓

Basic Retrieval

Complex Question
↓

Advanced Retrieval
```

---

Benefits:

```text
Lower Cost

Lower Latency
```

---

## Interview Takeaway

Adaptive RAG matches retrieval complexity to query complexity.

---

# Graph RAG

One of the hottest interview topics.

---

# Why Graph RAG Exists

Vector search is excellent for:

```text
Semantic Similarity
```

---

But weak for:

```text
Relationships

Dependencies

Connected Facts
```

---

Need:

```text
Knowledge Graphs
```

---

## Interview Takeaway

Graph RAG introduces explicit relationships between entities.

---

# Knowledge Graph Basics

Represent information as:

```text
Nodes

Edges
```

---

Example:

```text
Satya Nadella
       │
       │ CEO
       │
Microsoft
       │
       │ Acquired
       │
GitHub
```

---

Relationships become searchable.

---

## Interview Takeaway

Knowledge graphs model relationships explicitly.

---

# Graph RAG Pipeline

```text
Documents
      ↓

Entity Extraction
      ↓

Knowledge Graph
      ↓

Graph Retrieval
      ↓

LLM
```

---

Benefits:

```text
Better Reasoning

Better Multi-Hop Retrieval
```

---

## Interview Takeaway

Graph RAG improves relationship-aware retrieval.

---

# Graph RAG vs Vector RAG

| Topic | Vector RAG | Graph RAG |
|----------|----------|----------|
| Similarity Search | Strong | Moderate |
| Relationship Search | Weak | Strong |
| Multi-Hop Reasoning | Moderate | Strong |
| Simplicity | High | Lower |

---

## Interview Takeaway

Graph RAG excels at relationship-centric questions.

---

# Agentic RAG

One of the most important modern AI topics.

---

Traditional RAG:

```text
Retrieve Once
```

---

Agentic RAG:

```text
Plan

Retrieve

Reason

Retrieve Again

Validate

Answer
```

---

The system behaves like an agent.

---

## Interview Takeaway

Agentic RAG performs iterative retrieval and reasoning.

---

# Agentic RAG Workflow

```text
Question
      ↓

Planner
      ↓

Retriever
      ↓

Reasoner
      ↓

Additional Retrieval
      ↓

Answer
```

---

Unlike traditional RAG:

```text
Multiple Retrieval Loops
```

---

## Interview Takeaway

Agentic RAG supports complex reasoning tasks.

---

# Reflection-Based Retrieval

Emerging pattern.

---

Model evaluates:

```text
Do I Have Enough Information?
```

---

If not:

```text
Retrieve Again
```

---

Benefits:

```text
Higher Accuracy
```

---

## Interview Takeaway

Reflection improves retrieval completeness.

---

# RAG Evaluation In Production

A critical interview topic.

---

Need to evaluate:

```text
Retriever

Reranker

Generator
```

separately.

---

# Retrieval Metrics

```text
Recall@K

MRR

NDCG

Hit Rate
```

---

# Generation Metrics

```text
Faithfulness

Groundedness

Correctness
```

---

## Interview Takeaway

RAG evaluation requires both retrieval and generation metrics.

---

---

Advanced version:

```text
Query Rewriter
      ↓

Multi Query Retrieval
      ↓

Hybrid Search
      ↓

Reranker
      ↓

Agentic Reasoning
      ↓

LLM
```

# Choosing The Right RAG Strategy

| Problem | Recommended Approach |
|----------|----------|
| Basic FAQ | Standard RAG |
| Poor Retrieval | Query Expansion / HyDE |
| Complex Questions | Multi-Hop RAG |
| Relationship Queries | Graph RAG |
| Enterprise Search | Hybrid RAG |
| Autonomous Systems | Agentic RAG |

# Common Interview Questions

---

## Question

What is HyDE?

### Answer

```text
A retrieval technique that generates a hypothetical answer and uses its embedding for retrieval.
```

---

## Question

Why use multi-query retrieval?

### Answer

```text
To improve recall by searching from multiple perspectives.
```

---

## Question

What is multi-hop retrieval?

### Answer

```text
A retrieval strategy requiring multiple retrieval steps to answer a question.
```

---

## Question

What is Graph RAG?

### Answer

```text
A RAG architecture that combines knowledge graphs with LLMs for relationship-aware retrieval.
```

---

## Question

What is Agentic RAG?

### Answer

```text
A RAG system that performs iterative retrieval, reasoning, and planning.
```

---

## Question

What is Self-RAG?

### Answer

```text
A RAG architecture where the model decides whether retrieval is necessary.
```

---

## Question

What is Corrective RAG?

### Answer

```text
A retrieval pipeline that evaluates retrieval quality and corrects failures before generation.
```
