Word embeddings were a major breakthrough.

```text
Sentence Embeddings

Semantic Search

Vector Databases

RAG

Agent Memory
```

which form the foundation of modern AI systems.

---

# Why Word Embeddings Are Not Enough

Word embeddings capture:

```text
Word Meaning
```

---

Example:

```text
King

Queen

Dog

Cat
```

---

But real-world AI systems process:

```text
Questions

Documents

Chunks

Emails

Conversations
```

---

Not individual words.

---

## Interview Takeaway

Word embeddings capture word-level meaning, not sentence-level meaning.

---

# The Sentence Meaning Problem

Consider:

```text
The food was good.
```

---

and

```text
The movie was good.
```

---

The word:

```text
Good
```

appears in both.

---

However:

```text
Meaning Of Sentence
```

is different.

---

A system must understand:

```text
Entire Context
```

not just individual words.

---

## Interview Takeaway

Understanding sentences requires more than understanding individual words.

---

# What Is A Sentence Embedding?

## Definition

A sentence embedding is a dense vector representation of an entire sentence.

---

Example:

```text
I love machine learning.
```

↓

```text
[0.21, -0.44, 0.78, ...]
```

---

One vector represents:

```text
Entire Sentence Meaning
```

---

## Interview Takeaway

Sentence embeddings compress sentence meaning into a numerical vector.

---

# Why Sentence Embeddings Exist

Modern AI systems frequently need:

```text
Meaning Matching
```

instead of:

```text
Keyword Matching
```

---

Examples:

```text
Search

Recommendation Systems

RAG

Question Answering

Chatbots
```

---

Sentence embeddings make this possible.

---

## Interview Takeaway

Sentence embeddings enable semantic understanding.

---

# Keyword Search vs Semantic Search

One of the most important interview topics.

---

# Traditional Keyword Search

Searches for:

```text
Exact Words
```

---

Query:

```text
How do I reset my password?
```

---

Document:

```text
Steps to change your account credentials.
```

---

Keyword overlap:

```text
Almost None
```

---

Traditional search may fail.

---

# Semantic Search

Sentence embeddings capture:

```text
Meaning
```

---

System understands:

```text
Password

Credentials
```

are related concepts.

---

Search succeeds.

---

## Interview Takeaway

Semantic search retrieves meaning, not exact words.

---

# What Is Semantic Similarity?

## Definition

A measure of how similar two pieces of text are in meaning.

---

Example:

```text
How do I reset my password?
```

---

and

```text
How can I change my account credentials?
```

---

Semantically:

```text
Very Similar
```

---

Even though words differ.

---

## Interview Takeaway

Semantic similarity focuses on meaning rather than wording.

---

# Measuring Similarity

After generating embeddings, we compare vectors.

---

Most common methods:

```text
Cosine Similarity

Dot Product

Euclidean Distance
```

---

# Cosine Similarity

Most important interview topic.

---

## Idea

Measure:

```text
Angle Between Vectors
```

---

Not magnitude.

---

Example:

```text
Similar Meaning
```

↓

```text
Small Angle
```

↓

```text
High Similarity
```

---

## Range

```text
-1 To 1
```

---

Practical embedding systems usually operate between:

```text
0 To 1
```

---

## Interview Takeaway

Cosine similarity is the most common metric used in vector search.

---

# Dot Product

Measures:

```text
Direction

+

Magnitude
```

---

Frequently used in:

```text
Neural Retrieval Systems

Transformer Attention
```

---

## Interview Takeaway

Dot product is computationally efficient and widely used.

---

# Euclidean Distance

Measures:

```text
Physical Distance
```

between vectors.

---

Less common in modern retrieval systems.

---

## Interview Takeaway

Cosine similarity is generally preferred over Euclidean distance for embeddings.

---

# Evolution Of Sentence Embeddings

Sentence embeddings evolved through several stages.

---

# Stage 1: Averaging Word Embeddings

Simple approach:

```text
Sentence

↓

Word Embeddings

↓

Average
```

---

Example:

```text
Dog

Barks

Loudly
```

---

Average all vectors.

---

Advantages:

```text
Simple
```

---

Problems:

```text
Loses Word Order

Loses Context
```

---

## Interview Takeaway

Averaging embeddings is simple but ignores sentence structure.

---

# Stage 2: Neural Encoders

Researchers used:

```text
RNN

LSTM

GRU
```

---

to create sentence representations.

---

Advantages:

```text
Context Awareness
```

---

Problems:

```text
Slow

Limited Scalability
```

---

## Interview Takeaway

RNN-based sentence embeddings improved context understanding but were inefficient.

---

# Stage 3: Transformer-Based Embeddings

Modern approach.

---

Uses:

```text
Transformers
```

to encode meaning.

---

Examples:

```text
BERT

SBERT

E5

BGE

OpenAI Embeddings
```

---

## Interview Takeaway

Most modern embedding systems are transformer-based.

---

# Why BERT Alone Is Not Enough

One of the most common interview questions.

---

BERT produces:

```text
Contextual Embeddings
```

---

However:

```text
Similarity Search
```

requires comparing millions of documents.

---

BERT is computationally expensive for this.

---

## Interview Takeaway

BERT is powerful but inefficient for large-scale similarity search.

---

# Sentence-BERT (SBERT)

One of the most important interview topics.

---

Developed to solve:

```text
Efficient Semantic Search
```

---

# Core Idea

Precompute:

```text
Sentence Embeddings
```

---

Store them.

---

Compare embeddings directly.

---

Instead of:

```text
Running BERT

For Every Pair
```

---

## Interview Takeaway

SBERT enables scalable semantic similarity search.

---

# Why SBERT Became Popular

Before SBERT:

```text
Pairwise Comparison
```

required expensive BERT inference.

---

After SBERT:

```text
Embedding Once

Search Many Times
```

---

Huge performance improvement.

---

## Interview Takeaway

SBERT dramatically improves retrieval efficiency.

---

# How SBERT Works

Pipeline:

```text
Sentence
      ↓

Transformer Encoder
      ↓

Sentence Embedding
      ↓

Similarity Search
```

---

Simple but powerful.

---

## Interview Takeaway

SBERT converts entire sentences into comparable vectors.

---

# Chunk Embeddings

Critical RAG concept.

---

Documents are rarely embedded as:

```text
One Huge Vector
```

---

Instead:

```text
Document
     ↓

Chunks
     ↓

Embeddings
```

---

Each chunk receives its own vector.

---

## Interview Takeaway

RAG systems typically embed chunks rather than entire documents.

---

# Why Chunking Matters

Suppose:

```text
100 Page PDF
```

---

Embedding entire document:

```text
Poor Retrieval
```

---

Chunking:

```text
Improves Precision
```

---

because retrieval occurs at:

```text
Relevant Section
```

---

## Interview Takeaway

Chunking improves retrieval granularity.

---

# Document Embeddings

Sometimes we embed:

```text
Entire Documents
```

---

Useful for:

```text
Document Classification

Document Clustering
```

---

Less useful for:

```text
Precise Retrieval
```

---

## Interview Takeaway

Document embeddings trade retrieval precision for simplicity.

---

# Sentence vs Chunk vs Document Embeddings

| Type | Typical Length | Use Case |
|----------|----------|----------|
| Sentence | Small | Semantic Similarity |
| Chunk | Medium | RAG |
| Document | Large | Classification |

---

## Interview Takeaway

Chunk embeddings dominate modern RAG systems.

---

## Interview Takeaway

User Query:

```text
How do I reset my password?
```

---

Query Embedding:

```text
Vector
```

---

Search:

```text
Nearest Neighbor Retrieval
```

---

Return:

```text
Most Similar Chunks
```

---

## Interview Takeaway

Vector search retrieves semantically similar content.

---

# Embedding Evaluation

Important senior-level topic.

---

Question:

```text
Are My Embeddings Good?
```

---

Need evaluation metrics.

---

# Recall@K

Most common retrieval metric.

---

Measures:

```text
Did Relevant Result

Appear In Top K?
```

---

Example:

```text
Top 10 Results
```

---

Relevant document appears:

```text
Yes
```

---

Success.

---

## Interview Takeaway

Recall@K measures retrieval coverage.

---

# MRR (Mean Reciprocal Rank)

Measures:

```text
How Early

Correct Result Appears
```

---

Higher rank:

```text
Better Experience
```

---

## Interview Takeaway

MRR rewards systems that return relevant documents early.

---

# NDCG

Measures:

```text
Ranking Quality
```

---

Considers:

```text
Position

Relevance
```

---

Widely used in search evaluation.

---

## Interview Takeaway

NDCG evaluates overall ranking effectiveness.

---

# Limitations Of Sentence Embeddings

No model is perfect.

---

# Limitation 1

Domain Shift

---

Medical embeddings may not work well for:

```text
Legal Documents
```

---

# Limitation 2

Long Documents

---

Meaning may be compressed too aggressively.

---

# Limitation 3

False Similarity

---

Different concepts may appear:

```text
Artificially Similar
```

---

## Interview Takeaway

Embedding quality depends on domain and use case.

---
