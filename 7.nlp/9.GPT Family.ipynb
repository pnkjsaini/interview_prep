# GPT Architecture

Unlike BERT:

```text
BERT
↓
Encoder Only
```

---

GPT uses:

```text
Decoder Only
```

---

Architecture:

```text
Tokenization
       ↓

Embeddings
       ↓

Positional Encoding
       ↓

Transformer Decoder Blocks
       ↓

Next Token Prediction
```

---

## Interview Takeaway

GPT uses stacked decoder blocks with causal masking.

---

# Why Decoder Only?

Remember:

```text
Decoder
↓
Generation
```

---

GPT's goal:

```text
Predict The Next Token
```

---

Example:

```text
I Love
```

Predict:

```text
Machine
```

---

Then:

```text
I Love Machine
```

Predict:

```text
Learning
```

---

Continue repeatedly.

---

## Interview Takeaway

GPT generates text by repeatedly predicting the next token.

# Autoregressive Generation

One of the most important interview concepts.

---

# Definition

Generate one token at a time using previously generated tokens.

---

Example:

```text
Input:
I Love
```

---

Step 1:

```text
Predict:
Machine
```

---

Step 2:

```text
I Love Machine
```

Predict:

```text
Learning
```

---

Step 3:

```text
I Love Machine Learning
```

Predict:

```text
.
```

---

## Interview Takeaway

GPT generates text autoregressively.

---

# Causal Masking

A critical GPT concept.

---

Sentence:

```text
I Love Machine Learning
```

---

When predicting:

```text
Machine
```

---

GPT can only see:

```text
I

Love
```

---

Cannot see:

```text
Learning
```

---

Future tokens are hidden.

---

Called:

```text
Causal Masking
```

---

## Interview Takeaway

Causal masking prevents information leakage during training.

---

# Logits

A common interview topic.

---

Before prediction:

```text
Model Outputs Scores
```

---

Called:

```text
Logits
```

---

Example:

```text
Cat → 8.3

Dog → 6.1

Car → 1.2
```

---

Not probabilities yet.

---

## Interview Takeaway

Logits are raw prediction scores before Softmax.

---

# Softmax

Converts:

```text
Logits
```

↓

```text
Probabilities
```

---

Example:

```text
Cat → 0.82

Dog → 0.16

Car → 0.02
```

---

Total:

```text
1.0
```

---

## Interview Takeaway

Softmax converts logits into probability distributions.

---

# Greedy Decoding

Simplest generation strategy.

---

Rule:

```text
Always Pick

Highest Probability Token
```

---

Example:

```text
Cat → 0.82

Dog → 0.16
```

---

Pick:

```text
Cat
```

---

Problem:

```text
Repetitive Outputs
```

---

## Interview Takeaway

Greedy decoding is simple but often produces less diverse text.

---

# High Temperature

# Top-K Sampling

Instead of considering:

```text
Entire Vocabulary
```

---

Keep only:

```text
Top K Tokens
```

---

Example:

```text
Top 50
```

---

Sample from those.

---

## Interview Takeaway

Top-K limits generation to the most likely tokens.

---

# Top-P Sampling

Also called:

```text
Nucleus Sampling
```

---

Keep tokens whose cumulative probability reaches:

```text
P
```

---

Example:

```text
95%
```

---

Dynamic alternative to Top-K.

---

## Interview Takeaway

Top-P adapts dynamically based on probability distribution.

---

# Top-K vs Top-P

| Topic | Top-K | Top-P |
|----------|----------|----------|
| Fixed Candidates | Yes | No |
| Dynamic | No | Yes |
| Flexibility | Lower | Higher |
| Common Usage | Moderate | Very Common |

---

## Interview Takeaway

Top-P is generally preferred in modern LLM systems.

---

# Context Window

Critical LLM topic.

---

Definition:

```text
Maximum Tokens
Model Can Process
```

---

Includes:

```text
Prompt

Retrieved Documents

Conversation History

Response
```

---

## Interview Takeaway

Context window determines how much information an LLM can use.

---

# KV Cache

Very common senior-level topic.

---

Problem:

Without optimization:

```text
Recompute Attention

For Every Token
```

---

Expensive.

---

Solution:

```text
Store Previous Keys And Values
```

---

Reuse them.

---

Called:

```text
KV Cache
```

---

Benefits:

```text
Lower Latency

Faster Inference
```

---

## Interview Takeaway

KV Cache accelerates autoregressive generation.

# Common Interview Questions

---

## Question

What does GPT stand for?

### Answer

```text
Generative Pretrained Transformer.
```

---

## Question

Why is GPT decoder-only?

### Answer

```text
Because decoders are naturally suited for text generation.
```

---

## Question

What is autoregressive generation?

### Answer

```text
Generating text one token at a time using previous tokens.
```

---

## Question

What is next-token prediction?

### Answer

```text
The training objective where GPT predicts the next token in a sequence.
```

---

## Question

What is causal masking?

### Answer

```text
Preventing the model from seeing future tokens.
```

---

## Question

What is temperature?

### Answer

```text
A parameter controlling randomness during generation.
```

---

## Question

Difference between Top-K and Top-P?

### Answer

```text
Top-K keeps a fixed number of tokens.

Top-P keeps a dynamic probability mass.
```

---

## Question

What is KV Cache?

### Answer

```text
A mechanism that stores previous attention computations to speed up inference.
```
