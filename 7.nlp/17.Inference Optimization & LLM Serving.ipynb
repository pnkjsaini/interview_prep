# Training vs Inference

One of the most common interview questions.

---

# Training

Goal:

```text
Learn Model Weights
```

---

Operations:

```text
Forward Pass

Backward Pass

Gradient Updates
```

---

Very expensive.

---

# Inference

Goal:

```text
Generate Responses
```

---

Operations:

```text
Forward Pass Only
```

---

Much cheaper than training.

---

## Interview Takeaway

Inference performs prediction using fixed model weights.

---

# Why Inference Optimization Matters

Suppose:

```text
10,000 Users
```

query a chatbot simultaneously.

---

Without optimization:

```text
High GPU Cost

Slow Responses

Resource Exhaustion
```

---

Need:

```text
Efficient Serving
```

---

## Interview Takeaway

Inference optimization improves latency, throughput, and cost.

---

# LLM Inference Pipeline

```text
User Prompt
      ↓

Tokenization
      ↓

Prefill Phase
      ↓

Transformer
      ↓

Decode Phase
      ↓

Generated Tokens
```

---

Most optimization techniques target:

```text
Prefill

Decode
```

---

# Prefill Phase

One of the most important concepts.

---

Input:

```text
Entire Prompt
```

---

Example:

```text
2000 Token Prompt
```

---

Model processes:

```text
All Tokens Together
```

---

Characteristics:

```text
Compute Heavy

Highly Parallel
```

---

## Interview Takeaway

Prefill processes the entire input context.

---

# Decode Phase

After first token generation.

---

Example:

```text
Token 1 Generated
```

---

Need:

```text
Token 2
```

---

Then:

```text
Token 3
```

---

And so on.

---

Characteristics:

```text
Sequential

Latency Sensitive
```

---

## Interview Takeaway

Decode generates tokens one at a time.

---

# Why Decode Is Expensive

Remember:

```text
Autoregressive Generation
```

---

Each new token requires:

```text
Another Forward Pass
```

---

Without optimization:

```text
Very Slow
```

---

## Interview Takeaway

Autoregressive generation creates inference bottlenecks.

---

# KV Cache

One of the most important LLM interview topics.

---

# Problem

Suppose prompt:

```text
I Love Machine Learning
```

---

When generating:

```text
Token 10
```

---

Attention requires:

```text
Previous Tokens
```

again.

---

Without optimization:

```text
Recompute Everything
```

---

## Interview Takeaway

Repeated attention computation is expensive.

---

# Solution: KV Cache

Store:

```text
Keys

Values
```

from previous tokens.

---

Reuse them.

---

Instead of:

```text
Recompute
```

---

Result:

```text
Much Faster Generation
```

---

## Interview Takeaway

KV Cache stores previous attention states to accelerate decoding.

---

# KV Cache Visualization

Without cache:

```text
Token 1

Token 2

Token 3

Recompute All
```

---

With cache:

```text
Reuse Previous KV

Compute New Token Only
```

---

Huge latency reduction.

---

## Interview Takeaway

KV Cache transforms inference efficiency.

---

# Batching

Common production technique.

---

Instead of:

```text
One Request
```

---

Process:

```text
Many Requests Together
```

---

Example:

```text
Request 1

Request 2

Request 3
```

---

Single GPU operation.

---

## Interview Takeaway

Batching increases GPU utilization.

---

# Static Batching

Predefined batch size.

---

Example:

```text
Batch = 32
```

---

Wait until:

```text
32 Requests
```

arrive.

---

Problems:

```text
Latency
```

---

## Interview Takeaway

Static batching improves throughput but can increase latency.

---

# Dynamic Batching

Create batches dynamically.

---

Example:

```text
Incoming Requests
```

grouped automatically.

---

Benefits:

```text
Better Resource Usage

Lower Latency
```

---

## Interview Takeaway

Dynamic batching balances latency and throughput.

---

# Continuous Batching

One of the hottest interview topics.

---

Used by:

```text
vLLM

Modern Serving Engines
```

---

Idea:

```text
Add New Requests

While Existing Requests Run
```

---

No need to wait for batch completion.

---

## Interview Takeaway

Continuous batching maximizes GPU utilization.

---

# Why Continuous Batching Matters

Traditional batching:

```text
Idle GPU Time
```

---

Continuous batching:

```text
Near Constant GPU Usage
```

---

Huge throughput improvement.

---

## Interview Takeaway

Continuous batching is critical for large-scale LLM serving.

---

# Quantization

Extremely common interview topic.

---

# Definition

Reduce numerical precision of weights.

---

Example:

```text
FP32
```

↓

```text
INT8
```

---

Smaller memory footprint.

---

## Interview Takeaway

Quantization reduces model size and inference cost.

---

# FP32

Traditional training format.

---

Precision:

```text
32 Bits
```

---

Advantages:

```text
Very Accurate
```

---

Disadvantages:

```text
High Memory
```

---

# FP16

Widely used.

---

Precision:

```text
16 Bits
```

---

Benefits:

```text
Lower Memory

Faster Computation
```

---

## Interview Takeaway

FP16 is commonly used for efficient inference.

---

# BF16

Very important modern topic.

---

BF16:

```text
Brain Floating Point
```

---

Benefits:

```text
FP32 Range

FP16 Efficiency
```

---

Widely used on:

```text
NVIDIA

TPUs
```

---

## Interview Takeaway

BF16 provides stability with reduced memory consumption.

---

# INT8 Quantization

Convert weights to:

```text
8 Bit Integers
```

---

Benefits:

```text
Smaller Model

Faster Inference
```

---

Tradeoff:

```text
Small Accuracy Loss
```

---

## Interview Takeaway

INT8 is a popular production quantization format.

---

# INT4 Quantization

Even more aggressive.

---

Benefits:

```text
Very Small Models

Lower GPU Requirements
```

---

Tradeoff:

```text
Potential Quality Loss
```

---

Used heavily in:

```text
Local LLM Deployment
```

---

## Interview Takeaway

INT4 dramatically reduces memory requirements.

---

# Quantization Comparison

| Format | Memory | Accuracy |
|----------|----------|----------|
| FP32 | Highest | Highest |
| BF16 | High | Very High |
| FP16 | Medium | High |
| INT8 | Low | Good |
| INT4 | Very Low | Moderate |

---

## Interview Takeaway

Quantization trades accuracy for efficiency.

---

# Speculative Decoding

Modern inference optimization.

---

Problem:

```text
Large Models

Generate Slowly
```

---

Solution:

```text
Small Draft Model
```

---

---

Benefits:

```text
Higher Throughput

Lower Latency
```

---

## Interview Takeaway

Speculative decoding accelerates generation using a draft model.

---

## Interview Takeaway

## Interview Takeaway

Pipeline parallelism distributes layers across GPUs.

# Common Interview Questions

---

## Question

What is KV Cache?

### Answer

```text
A mechanism that stores previously computed keys and values to accelerate decoding.
```

---

## Question

What is the difference between prefill and decode?

### Answer

```text
Prefill processes the input prompt.

Decode generates output tokens sequentially.
```

---

## Question

Why is decoding slower than prefill?

### Answer

```text
Because decoding is autoregressive and sequential.
```

---

## Question

What is continuous batching?

### Answer

```text
A technique that dynamically adds requests to active batches.
```

---

## Question

What is quantization?

### Answer

```text
Reducing numerical precision to lower memory and compute requirements.
```

---

## Question

Why use BF16?

### Answer

```text
It combines FP32 numerical range with FP16 efficiency.
```

---

## Question

What is speculative decoding?

### Answer

```text
Using a smaller draft model to accelerate token generation.
```

---

## Question

Why is vLLM popular?

### Answer

```text
Because it provides efficient KV cache management and continuous batching.
```

---

## Question

Tensor Parallelism vs Pipeline Parallelism?

### Answer

```text
Tensor parallelism splits computations.

Pipeline parallelism splits layers.
```
