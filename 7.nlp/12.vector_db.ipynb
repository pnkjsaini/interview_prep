# Chapter 12: Vector Databases & Approximate Nearest Neighbor Search

In the previous chapter we studied:

```text
Embeddings

Dense Retrieval

Bi-Encoders

Cross-Encoders

Rerankers
```

At this point we know:

```text
Document
      ↓

Embedding
      ↓

Vector
```

A new problem appears:

> How do we search through millions or billions of vectors efficiently?

This problem led to:

```text
Vector Databases

Approximate Nearest Neighbor Search
```

which form the foundation of:

```text
RAG

Semantic Search

Agent Memory

Recommendation Systems
```

---

# Why Vector Databases Exist

Traditional databases store:

```text
Numbers

Strings

Dates
```

---

Example:

```sql
SELECT * FROM users
WHERE age > 30;
```

---

Works perfectly for:

```text
Structured Data
```

---

Embeddings are different.

---

Example:

```text
[0.32, -0.71, 0.11, ...]
```

---

Question:

```text
Find Similar Vectors
```

---

Traditional databases are not optimized for this.

---

## Interview Takeaway

Vector databases are designed for similarity search rather than exact matching.

---

# The Similarity Search Problem

Suppose:

```text
1 Million Documents
```

---

Each document:

```text
1536-Dimensional Embedding
```

---

User Query:

```text
Reset My Password
```

---

Need:

```text
Find Most Similar Embeddings
```

---

Brute force approach:

```text
Compare Query

With Every Vector
```

---

Complexity:

```text
O(N)
```

---

Too expensive.

---

## Interview Takeaway

Brute-force similarity search does not scale.

---

# Nearest Neighbor Search

Core idea:

```text
Find Closest Vectors
```

---

Example:

```text
Query Embedding
```

↓

```text
Nearest Documents
```

---

Most retrieval systems rely on:

```text
Nearest Neighbor Search
```

---

## Interview Takeaway

Vector retrieval is fundamentally a nearest-neighbor problem.

---

# Exact Nearest Neighbor Search

Simple approach:

```text
Compare Against Every Vector
```

---

Advantages:

```text
Perfect Accuracy
```

---

Disadvantages:

```text
Slow

Expensive

Poor Scalability
```

---

## Interview Takeaway

Exact search is accurate but inefficient at scale.

---

# Approximate Nearest Neighbor (ANN)

One of the most important interview topics.

---

# Definition

Find vectors that are:

```text
Almost Correct
```

instead of:

```text
Perfectly Correct
```

---

Trade:

```text
Tiny Accuracy Loss

Huge Speed Gain
```

---

## Interview Takeaway

ANN dramatically improves retrieval speed.

---

# Why ANN Works

Suppose:

```text
10 Million Vectors
```

---

Searching all vectors:

```text
Slow
```

---

ANN searches:

```text
Small Candidate Set
```

---

Result:

```text
Much Faster Retrieval
```

---

## Interview Takeaway

ANN reduces search space before similarity computation.

---

# Retrieval Pipeline

```text
Query
      ↓

Embedding
      ↓

ANN Search
      ↓

Top K Chunks
      ↓

Reranker
      ↓

Final Results
```

---

This is the standard modern RAG architecture.

---

# Indexes In Vector Search

A very important concept.

---

Instead of:

```text
Store Raw Vectors
```

---

Build:

```text
Index
```

---

Purpose:

```text
Faster Search
```

---

Analogy:

```text
Book Index
```

instead of reading every page.

---

## Interview Takeaway

Indexes accelerate vector retrieval.

---

# Flat Index

Simplest FAISS index.

---

Method:

```text
Compare Every Vector
```

---

Advantages:

```text
Perfect Accuracy
```

---

Disadvantages:

```text
Slow
```

---

Equivalent to:

```text
Brute Force Search
```

---

## Interview Takeaway

Flat indexes provide exact retrieval.

---

# IVF (Inverted File Index)

Very common interview topic.

---

Core idea:

```text
Cluster Vectors
```

---

Search:

```text
Relevant Clusters Only
```

---

Instead of:

```text
Entire Dataset
```

---

Advantages:

```text
Much Faster
```

---

## Interview Takeaway

IVF reduces search space using clustering.

---

# IVF Workflow

```text
Vectors
      ↓

Clustering
      ↓

Centroids
      ↓

Relevant Cluster
      ↓

Similarity Search
```

---

## Interview Takeaway

IVF searches local regions rather than the entire vector space.

---

# HNSW

One of the most important ANN algorithms.

---

HNSW:

```text
Hierarchical Navigable Small World
```

---

Used by:

```text
Milvus

Pinecone

Weaviate

Azure AI Search

Databricks Vector Search
```

---

## Interview Takeaway

HNSW is the dominant ANN algorithm in production systems.

---

# Intuition Behind HNSW

Think:

```text
Road Network
```

---

Instead of checking:

```text
Every House
```

---

Use:

```text
Highways

Roads

Local Streets
```

---

to quickly navigate.

---

## Interview Takeaway

HNSW creates graph structures for efficient search.

---

# Why HNSW Is Powerful

Advantages:

```text
High Recall

Fast Search

Scalable
```

---

Often provides:

```text
Near Exact Accuracy
```

with ANN speed.

---

## Interview Takeaway

HNSW balances accuracy and performance extremely well.

---

# IVF vs HNSW

| Topic | IVF | HNSW |
|----------|----------|----------|
| Structure | Clusters | Graph |
| Search Speed | Fast | Very Fast |
| Recall | Good | Excellent |
| Memory Usage | Lower | Higher |
| Industry Adoption | High | Very High |

---

## Interview Takeaway

HNSW is generally preferred for high-quality retrieval.

---

# Common Interview Questions

---

## Question

Why can't relational databases perform vector search efficiently?

### Answer

```text
Because they are optimized for exact matching rather than similarity search.
```

---

## Question

What is ANN?

### Answer

```text
Approximate Nearest Neighbor search, which trades small accuracy loss for major speed improvements.
```

---

## Question

What is FAISS?

### Answer

```text
An open-source vector similarity search library developed by Meta.
```

---

## Question

What is IVF?

### Answer

```text
An indexing technique that clusters vectors and searches only relevant clusters.
```

---

## Question

What is HNSW?

### Answer

```text
A graph-based ANN algorithm widely used in production vector databases.
```

---

## Question

Why is HNSW popular?

### Answer

```text
Because it provides excellent retrieval quality with low latency.
```

---

## Question

What is hybrid search?

### Answer

```text
Combining keyword search and vector search.
```

---

## Question

Why use metadata filtering?

### Answer

```text
To restrict retrieval to relevant subsets of documents.
```
