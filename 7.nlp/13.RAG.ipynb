# Module 7: NLP For LLM Engineers

# Chapter 13: Retrieval-Augmented Generation (RAG)

In the previous chapters we studied:

```text
Embeddings

Dense Retrieval

Rerankers

Vector Databases

ANN Search
```

At this point we know:

```text
Documents
      ↓

Embeddings
      ↓

Vector Database
      ↓

Retrieval
```

A new question appears:

> If LLMs already know language, why do we need retrieval?

The answer leads to one of the most important concepts in modern AI:

```text
Retrieval-Augmented Generation (RAG)
```

---

# Why RAG Exists

Large Language Models have limitations.

---

# Limitation 1: Knowledge Cutoff

Example:

```text
What happened yesterday?
```

---

LLM may not know.

---

Because:

```text
Training Data Is Static
```

---

## Interview Takeaway

LLMs cannot automatically know information created after training.

---

# Limitation 2: Hallucinations

Example:

```text
Invented Facts

Fake Policies

Wrong Numbers
```

---

LLMs generate:

```text
Probable Text

Not Verified Truth
```

---

## Interview Takeaway

LLMs can confidently generate incorrect information.

---

# Limitation 3: Private Data

Example:

```text
Company Documents

Internal Policies

Contracts

Support Tickets
```

---

These are not present in:

```text
Pretraining Data
```

---

## Interview Takeaway

LLMs do not know organization-specific knowledge by default.

---

# The Core Idea Of RAG

Instead of relying only on:

```text
Model Memory
```

---

Give the model:

```text
Relevant Documents
```

at runtime.

---

Pipeline:

```text
Question
      ↓

Retrieve Knowledge
      ↓

Inject Context
      ↓

Generate Answer
```

---

## Interview Takeaway

RAG combines retrieval with generation.

---

# What Does RAG Mean?

Retrieval:

```text
Find Relevant Information
```

---

Augmented:

```text
Provide Retrieved Information
```

---

Generation:

```text
Generate Grounded Answer
```

---

## Interview Takeaway

RAG grounds LLM responses using external knowledge.

---

# High Level RAG Architecture

```text
Documents
      ↓

Chunking
      ↓

Embeddings
      ↓

Vector Database

====================

User Query
      ↓

Embedding
      ↓

Retrieval
      ↓

Relevant Chunks
      ↓

Prompt Construction
      ↓

LLM
      ↓

Answer
```

---

This architecture powers most enterprise AI systems.

---

# RAG In One Example

Knowledge Base:

```text
Password Reset Policy
```

---

User asks:

```text
How do I reset my password?
```

---

System:

```text
Retrieve Policy
```

---

Injects:

```text
Policy Text
```

---

LLM generates:

```text
Grounded Response
```

---

## Interview Takeaway

RAG enables answering questions using external documents.

---

# RAG Pipeline Overview

The standard RAG workflow contains two phases:

```text
Indexing

Retrieval
```

---

# Phase 1: Indexing

Performed offline.

---

Documents:

```text
PDF

DOCX

HTML

Emails

Knowledge Base
```

---

Converted into:

```text
Chunks
```

---

Then:

```text
Embeddings
```

---

Stored in:

```text
Vector Database
```

---

## Interview Takeaway

Indexing prepares knowledge for retrieval.

---

# Phase 2: Retrieval

Performed at runtime.

---

User Question:

```text
How do I reset my password?
```

---

Generate:

```text
Query Embedding
```

---

Search:

```text
Vector Database
```

---

Retrieve:

```text
Top K Chunks
```

---

## Interview Takeaway

Retrieval finds relevant knowledge dynamically.

---

# Why Chunking Exists

One of the most asked RAG interview questions.

---

Suppose document:

```text
300 Pages
```

---

Embedding entire document:

```text
Poor Retrieval
```

---

Instead:

```text
Split Into Chunks
```

---

Example:

```text
Chunk 1

Chunk 2

Chunk 3
```

---

Retrieve only relevant sections.

---

## Interview Takeaway

Chunking improves retrieval precision.

---

# Chunk Size Tradeoff

Small Chunks:

```text
Better Precision
```

---

Problems:

```text
Loss Of Context
```

---

Large Chunks:

```text
More Context
```

---

Problems:

```text
Lower Precision
```

---

## Interview Takeaway

Chunk size is a balance between context and precision.

---

# Chunk Overlap

Very common interview topic.

---

Example:

```text
Chunk 1

Sentence A
Sentence B
Sentence C
```

---

```text
Chunk 2

Sentence C
Sentence D
Sentence E
```

---

Shared text:

```text
Sentence C
```

---

Called:

```text
Overlap
```

---

Purpose:

```text
Prevent Context Loss
```

---

## Interview Takeaway

Overlap preserves information across chunk boundaries.

---

# Retrieval

After chunking:

```text
Chunks
      ↓

Embeddings
      ↓

Vector Search
```

---

Retrieve:

```text
Top K Similar Chunks
```

---

Common values:

```text
Top 3

Top 5

Top 10
```

---

## Interview Takeaway

Retrieval identifies the most relevant chunks.

---

# Similarity Search

Usually uses:

```text
Cosine Similarity
```

---

Query:

```text
Password Reset
```

---

Find chunks with:

```text
Highest Similarity Score
```

---

## Interview Takeaway

Similarity search drives retrieval quality.

---

# Prompt Augmentation

After retrieval:

```text
Relevant Chunks
```

must be added to the prompt.

---

Example:

```text
Context:
[Retrieved Documents]

Question:
How do I reset my password?
```

---

This step is called:

```text
Prompt Augmentation
```

---

## Interview Takeaway

Retrieved knowledge is injected into the LLM prompt.

---

# Grounding

One of the most important RAG concepts.

---

Definition:

```text
Force Answers

To Depend On Retrieved Evidence
```

---

Instead of:

```text
Model Memory
```

---

Use:

```text
Retrieved Facts
```

---

## Interview Takeaway

Grounding reduces hallucinations.

---

# Why RAG Reduces Hallucinations

Without RAG:

```text
Answer From Memory
```

---

With RAG:

```text
Answer From Retrieved Evidence
```

---

More reliable.

---

## Interview Takeaway

RAG improves factual accuracy.

---

# Retriever

Core RAG component.

---

Responsibilities:

```text
Find Relevant Chunks
```

---

Examples:

```text
BGE

E5

OpenAI Embeddings

SBERT
```

---

## Interview Takeaway

Retriever quality strongly impacts RAG performance.

---

# Generator

Second major component.

---

Typically:

```text
GPT

Claude

Llama

Gemini
```

---

Responsibilities:

```text
Generate Final Response
```

---

## Interview Takeaway

The generator synthesizes retrieved information.

---

# Retriever vs Generator

| Component | Responsibility |
|------------|------------|
| Retriever | Find Knowledge |
| Generator | Generate Answer |

---

## Interview Takeaway

RAG separates knowledge retrieval from language generation.

---

# Why Rerankers Exist

Retrieval often returns:

```text
Top 20 Chunks
```

---

Not all equally relevant.

---

Need:

```text
Cross Encoder
```

---

to reorder results.

---

Pipeline:

```text
Retriever
      ↓

Top 20 Chunks
      ↓

Reranker
      ↓

Top 5 Chunks
```

---

## Interview Takeaway

Rerankers improve retrieval precision.

---

# Hybrid Search

Very common production approach.

---

Combine:

```text
Keyword Search

+

Vector Search
```

---

Example:

```text
BM25

+

Embeddings
```

---

Benefits:

```text
Exact Matching

+

Semantic Matching
```

---

## Interview Takeaway

Hybrid retrieval often outperforms pure vector retrieval.

---

# Metadata Filtering

Example:

```text
Department = HR

Region = India

Year = 2025
```

---

Retrieve:

```text
Only Relevant Documents
```

---

before vector search.

---

## Interview Takeaway

Metadata filtering improves relevance and latency.

---

# Common RAG Failures

Important interview topic.

---

# Retrieval Failure

Wrong chunks retrieved.

---

# Chunking Failure

Information split incorrectly.

---

# Context Window Overflow

Too much retrieved text.

---

# Hallucinated Synthesis

Correct retrieval.

Wrong answer generation.

---

## Interview Takeaway

Many RAG failures originate in retrieval rather than generation.

---

---

# Multi-Query Retrieval

# Parent Child Retrieval

```text
Small Chunks
```

---

Balances:

```text
Precision

Context
```

## Interview Takeaway

Advanced retrieval strategies improve coverage and relevance.

# RAG vs Fine-Tuning

Extremely common interview question.

| RAG | Fine-Tuning |
|----------|----------|
| External Knowledge | Model Weights |
| Dynamic Updates | Retraining Needed |
| Lower Cost | Higher Cost |
| Better For Documents | Better For Behavior |

---

## Interview Takeaway

RAG updates knowledge, fine-tuning changes behavior.

---
