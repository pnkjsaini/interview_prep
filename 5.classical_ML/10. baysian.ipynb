# Module 5: Classical Machine Learning

# Chapter 10: Naive Bayes

In the previous chapter we studied:

```text
KNN
```

which makes predictions using:

```text
Similarity

Distance

Nearest Neighbors
```

Naive Bayes takes a completely different approach.

Instead of asking:

```text
Which Observations Look Similar?
```

it asks:

```text
Given The Evidence,

What Is The Probability Of Each Class?
```

Naive Bayes is one of the simplest and fastest probabilistic machine learning algorithms.

---

# Why Naive Bayes Exists

Suppose we receive an email.

---

Words present:

```text
Free

Winner

Prize

Offer
```

---

Question:

```text
Spam

Or

Not Spam?
```

---

Instead of looking for similar emails,

Naive Bayes calculates:

```text
Probability(Spam)

Probability(Not Spam)
```

---

Whichever probability is larger becomes the prediction.

---

## Interview Takeaway

Naive Bayes predicts the class with the highest probability.

---

# Core Idea

Naive Bayes is based on:

```text
Bayes Theorem
```

which we studied in Probability.

---

Bayes helps us answer:

> Given the evidence, what is the probability of a class?

---

## Example

Given:

```text
Customer Age

Income

Purchase History
```

---

Calculate:

```text
Probability(Buy)

Probability(Not Buy)
```

---

Choose the larger probability.

---

## Interview Takeaway

Naive Bayes uses probability to perform classification.

---

# Bayes Theorem Recap

## Formula

```text
P(A|B)

=

P(B|A) × P(A)

----------------
      P(B)
```

---

## Meaning

Bayes updates probability when new evidence becomes available.

---

## Example

Question:

```text
What Is The Probability

Of Disease

Given Positive Test?
```

---

Bayes combines:

```text
Prior Knowledge

+

Observed Evidence
```

---

to produce:

```text
Updated Probability
```

---

## Interview Takeaway

Bayes Theorem updates beliefs using evidence.

---

# Understanding The Formula

---

# Prior Probability

## Definition

Probability before seeing evidence.

---

Example

```text
P(Spam)
```

---

Percentage of emails that are spam.

---

## Interview Takeaway

Prior represents existing belief.

---

# Likelihood

## Definition

Probability of observing evidence given a class.

---

Example

```text
P(Free | Spam)
```

---

Probability of the word:

```text
Free
```

appearing in spam emails.

---

## Interview Takeaway

Likelihood measures how strongly evidence supports a class.

---

# Posterior Probability

## Definition

Final probability after considering evidence.

---

Example

```text
P(Spam | Free)
```

---

Probability email is spam after seeing:

```text
Free
```

---

## Interview Takeaway

Posterior is the updated probability.

---

# Why Is It Called "Naive"?

One of the most common interview questions.

---

Naive Bayes assumes:

```text
All Features Are Independent
```

given the class.

---

## Example

Suppose we predict:

```text
Loan Approval
```

---

Features:

```text
Income

Savings

Net Worth
```

---

Reality:

```text
These Features Are Related
```

---

Naive Bayes assumes:

```text
They Are Independent
```

---

This assumption is:

```text
Naive
```

---

Hence the name:

```text
Naive Bayes
```

---

## Interview Takeaway

Naive Bayes assumes conditional independence among features.

---

# Conditional Independence

## Definition

Knowing one feature provides no additional information about another feature after the class is known.

---

## Layman Explanation

Suppose class is:

```text
Spam
```

---

After knowing:

```text
Spam
```

Naive Bayes treats:

```text
Free

Prize

Offer
```

as independent signals.

---

## Why It Matters

This assumption makes probability calculations simple.

---

## Interview Takeaway

Conditional independence dramatically simplifies computation.

---

# How Naive Bayes Makes Predictions

Suppose we want to classify:

```text
Spam

Not Spam
```

---

# Step 1

Calculate:

```text
P(Spam)
```

---

# Step 2

Calculate:

```text
P(Words | Spam)
```

---

# Step 3

Multiply probabilities.

---

# Step 4

Repeat for:

```text
Not Spam
```

---

# Step 5

Choose larger probability.

---

## Interview Takeaway

Naive Bayes predicts the class with the highest posterior probability.

---

# Gaussian Naive Bayes

## When To Use

Continuous numerical features.

---

Examples:

```text
Age

Income

Salary

Temperature
```

---

# Core Assumption

Features follow:

```text
Normal Distribution
```

---

## Example

Medical Diagnosis

Using:

```text
Age

Blood Pressure

Heart Rate
```

---

## Interview Takeaway

Gaussian Naive Bayes is used for continuous features.

---

# Multinomial Naive Bayes

One of the most important variants.

---

## When To Use

Count data.

---

Examples:

```text
Word Counts

Term Frequencies

Document Analysis
```

---

## Common Applications

```text
Spam Detection

Document Classification

Sentiment Analysis
```

---

## Example

Email

```text
Free → 5 Times

Winner → 3 Times

Offer → 2 Times
```

---

Word frequency matters.

---

## Interview Takeaway

Multinomial Naive Bayes is widely used in NLP.

---

# Bernoulli Naive Bayes

## When To Use

Binary features.

---

Features become:

```text
Present

Or

Absent
```

---

Example

Word:

```text
Free
```

---

Feature:

```text
1 = Present

0 = Not Present
```

---

Frequency does not matter.

---

Only existence matters.

---

## Interview Takeaway

Bernoulli Naive Bayes focuses on feature presence or absence.

---

# Laplace Smoothing

One of the most common interview questions.

---

# Problem

Suppose a word never appeared in training.

---

Example

```text
Bitcoin
```

---

Probability becomes:

```text
0
```

---

Multiplication then becomes:

```text
0
```

---

Entire prediction collapses.

---

# Solution

Add a small value.

---

Typically:

```text
+1
```

---

Result:

```text
No Probability Becomes Zero
```

---

## Interview Takeaway

Laplace Smoothing prevents zero probabilities.

---

# Why Naive Bayes Works Surprisingly Well

The independence assumption is often wrong.

---

Yet Naive Bayes performs well because:

```text
Probability Ranking

Is Often More Important

Than Exact Probabilities
```

---

Even imperfect assumptions can produce good classifications.

---

## Interview Takeaway

Naive Bayes works well despite unrealistic assumptions.

---

# Advantages

## Extremely Fast

Training is very efficient.

---

## Works Well On Small Datasets

Requires little data.

---

## Handles High Dimensions

Works well for text data.

---

## Easy To Implement

Simple probabilistic framework.

---

# Limitations

## Independence Assumption

Often unrealistic.

---

## Limited Expressiveness

Cannot model complex feature interactions.

---

## Probability Estimates

May not always be perfectly calibrated.

---

## Performance Ceiling

Often outperformed by boosting algorithms.

---

# Naive Bayes vs Logistic Regression

| Topic | Naive Bayes | Logistic Regression |
|----------|----------|----------|
| Approach | Probabilistic | Discriminative |
| Training Speed | Faster | Slower |
| Independence Assumption | Yes | No |
| Feature Interaction | Limited | Better |
| Interpretability | High | High |

---

# Naive Bayes vs KNN

| Topic | Naive Bayes | KNN |
|----------|----------|----------|
| Training | Fast | Almost None |
| Prediction | Fast | Slow |
| Memory Usage | Low | High |
| Scalability | Better | Worse |
