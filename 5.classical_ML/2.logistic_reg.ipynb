# Module 5: Classical Machine Learning

# Chapter 2: Logistic Regression

In the previous chapter we learned:

```text
Linear Regression
```

which predicts:

```text
Continuous Values
```

Examples:

```text
House Price

Salary

Revenue

Sales
```

However many business problems do not require a number.

Instead they require a decision.

Examples:

```text
Spam Or Not Spam

Fraud Or Not Fraud

Customer Churn Or Not

Disease Or No Disease
```

These are:

```text
Classification Problems
```

and Linear Regression is not suitable for them.

This leads to:

```text
Logistic Regression
```

---

# Why Linear Regression Fails For Classification

Suppose we want to predict:

```text
Fraud

Or

Not Fraud
```

---

Target Values:

```text
0 = Not Fraud

1 = Fraud
```

---

Linear Regression can produce:

```text
-2

3.5

7.8
```

---

These outputs do not make sense because probabilities must lie between:

```text
0 And 1
```

---

We need a model that always produces:

```text
Valid Probabilities
```

---

## Interview Takeaway

Linear Regression can generate invalid probabilities, making it unsuitable for classification.

---

# Core Idea Of Logistic Regression

Instead of predicting:

```text
Class Directly
```

Logistic Regression predicts:

```text
Probability Of Belonging To A Class
```

---

Example

```text
Fraud Probability

=

0.95
```

---

Interpretation:

```text
95% Chance Of Fraud
```

---

## Interview Takeaway

Logistic Regression predicts probabilities rather than raw classes.

---

# Sigmoid Function

## Definition

The sigmoid function converts any number into a value between:

```text
0 And 1
```

---

## Formula

```text
σ(x)

=

1 / (1 + e⁻ˣ)
```

---

## Layman Explanation

Think of sigmoid as a probability converter.

---

Input:

```text
-100
```

Output:

```text
≈ 0
```

---

Input:

```text
0
```

Output:

```text
0.5
```

---

Input:

```text
100
```

Output:

```text
≈ 1
```

---

## Why It Matters

It ensures predictions remain valid probabilities.

---

## Interview Takeaway

Sigmoid converts any real number into a probability.

---

# Logistic Regression Equation

Linear Regression:

```text
y

=

β₀ + β₁x₁ + β₂x₂
```

---

Logistic Regression:

```text
z

=

β₀ + β₁x₁ + β₂x₂
```

---

Then:

```text
Probability

=

Sigmoid(z)
```

---

## Flow

```text
Features
      ↓
Linear Combination
      ↓
Sigmoid
      ↓
Probability
```

---

## Interview Takeaway

Logistic Regression is Linear Regression followed by a Sigmoid function.

---

# Probability Interpretation

Suppose model output:

```text
0.90
```

---

Interpretation:

```text
90% Probability
```

of belonging to the positive class.

---

Suppose output:

```text
0.20
```

---

Interpretation:

```text
20% Probability
```

of belonging to the positive class.

---

## Interview Takeaway

Logistic Regression outputs class probabilities.

---

# Decision Boundary

## Definition

A decision boundary converts probabilities into class labels.

---

## Layman Explanation

Eventually we need:

```text
Fraud

Or

Not Fraud
```

---

not probabilities.

---

A threshold is chosen.

Typically:

```text
0.5
```

---

Rule:

```text
Probability ≥ 0.5

→ Class 1
```

---

```text
Probability < 0.5

→ Class 0
```

---

## Example

```text
0.90 → Fraud

0.70 → Fraud

0.20 → Not Fraud
```

---

## Why It Matters

Thresholds control classification behavior.

---

## Interview Takeaway

Decision boundaries convert probabilities into classes.

---

# Odds

## Definition

Odds compare the probability of success against failure.

---

## Formula

```text
Odds

=

p / (1-p)
```

---

## Example

Probability:

```text
0.80
```

---

Odds:

```text
0.80 / 0.20

=

4
```

---

Interpretation:

```text
Success Is 4 Times More Likely Than Failure
```

---

## Why It Matters

Logistic Regression models odds rather than probabilities directly.

---

## Interview Takeaway

Odds compare the likelihood of success versus failure.

---

# Log Odds (Logit)

## Definition

Log Odds are the logarithm of odds.

---

## Formula

```text
log(p/(1-p))
```

---

## Layman Explanation

Probabilities are difficult to model linearly.

---

Log Odds can take values from:

```text
-∞

To

+∞
```

---

making them suitable for linear modeling.

---

## Example

Probability:

```text
0.5
```

---

Log Odds:

```text
0
```

---

Probability:

```text
0.9
```

---

Positive Log Odds.

---

Probability:

```text
0.1
```

---

Negative Log Odds.

---

## Interview Takeaway

Logistic Regression models log odds as a linear function of features.

---

# Why It Is Called Logistic Regression

One of the most common interview questions.

---

The model performs:

```text
Classification
```

---

but internally it models:

```text
Regression On Log Odds
```

---

Hence the name:

```text
Logistic Regression
```

---

## Interview Takeaway

It is called Logistic Regression because it performs regression on log odds.

---

# How Logistic Regression Learns

Suppose:

```text
Random Coefficients
```

---

Predictions:

```text
Poor
```

---

Calculate:

```text
Prediction Error
```

---

Update coefficients.

---

Repeat:

```text
Prediction
      ↓
Error
      ↓
Update
      ↓
Better Prediction
```

---

The optimization process is similar to Linear Regression.

---

## Interview Takeaway

Logistic Regression learns coefficients that maximize prediction quality.

---

# Why MSE Is Not Used

One of the most common interview questions.

---

Linear Regression uses:

```text
MSE
```

---

Logistic Regression uses:

```text
Cross Entropy Loss
```

---

## Why?

The sigmoid function creates a nonlinear output.

---

MSE produces:

```text
Poor Optimization Behavior
```

---

Cross Entropy provides:

```text
Better Gradients

Faster Learning

Probabilistic Interpretation
```

---

## Interview Takeaway

Logistic Regression uses Cross Entropy instead of MSE.

---

# Binary Cross Entropy

## Definition

Binary Cross Entropy measures error between actual labels and predicted probabilities.

---

## Layman Explanation

Correct and confident predictions:

```text
Small Loss
```

---

Wrong and confident predictions:

```text
Huge Loss
```

---

## Example

Actual:

```text
Fraud = 1
```

---

Prediction:

```text
0.95
```

Small loss.

---

Prediction:

```text
0.05
```

Large loss.

---

## Interview Takeaway

Binary Cross Entropy is the standard loss function for Logistic Regression.

---

# Maximum Likelihood Estimation (MLE)

## Definition

MLE chooses coefficients that make observed data most likely.

---

## Layman Explanation

Suppose actual outcomes already occurred.

---

We want coefficients that make those outcomes:

```text
Most Probable
```

---

## Why It Matters

Logistic Regression is mathematically derived using:

```text
Maximum Likelihood Estimation
```

---

## Interview Takeaway

Logistic Regression learns parameters by maximizing likelihood.

---

# Multiclass Logistic Regression

## Problem

What if classes are:

```text
Dog

Cat

Horse
```

instead of:

```text
0

1
```

---

## Solution

Use:

```text
Softmax
```

instead of sigmoid.

---

Softmax produces:

```text
Probabilities Across Multiple Classes
```

---

Example

```text
Dog = 0.80

Cat = 0.15

Horse = 0.05
```

---

## Interview Takeaway

Multiclass Logistic Regression uses Softmax instead of Sigmoid.

---

# Advantages

## Easy To Interpret

Coefficients have clear meaning.

---

## Fast Training

Efficient optimization.

---

## Probabilistic Output

Produces confidence scores.

---

## Strong Baseline

Works surprisingly well on many problems.

---

# Limitations

## Linear Decision Boundary

May fail on complex nonlinear relationships.

---

## Sensitive To Outliers

Can be influenced by extreme observations.

---

## Feature Engineering Often Needed

Complex patterns may require transformations.

---

# Logistic Regression In Machine Learning

Applications:

```text
Fraud Detection

Spam Detection

Medical Diagnosis

Customer Churn

Credit Risk

Marketing Response Prediction
```

# Linear Regression vs Logistic Regression

| Topic | Linear Regression | Logistic Regression |
|----------|----------|----------|
| Problem Type | Regression | Classification |
| Output | Continuous Value | Probability |
| Activation | None | Sigmoid |
| Loss Function | MSE | Cross Entropy |
| Target | Numerical | Class Label |
