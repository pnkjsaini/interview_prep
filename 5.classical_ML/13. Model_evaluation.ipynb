# Module 5: Classical Machine Learning

# Chapter 13: Model Evaluation

So far we have studied many machine learning algorithms:

```text
Linear Regression

Logistic Regression

Decision Trees

Random Forest

Gradient Boosting

XGBoost

SVM

KNN

Naive Bayes
```

A critical question remains:

> "How do we know if a model is good?"

Many beginners look only at:

```text
Accuracy
```

This is one of the biggest mistakes in Machine Learning.

Different business problems require different evaluation metrics.

Understanding model evaluation is often more important than understanding the algorithm itself.

---

# Why Model Evaluation Exists

Suppose we build a fraud detection model.

---

Results:

```text
Accuracy = 99%
```

---

Looks excellent.

---

However:

```text
All Transactions

Predicted As

Not Fraud
```

---

Accuracy remains:

```text
99%
```

because fraud is rare.

---

But the model is useless.

---

This is why we need:

```text
Evaluation Metrics
```

---

## Interview Takeaway

Accuracy alone can be misleading.

---

Machine Learning evaluation is divided into:

```text
Classification Metrics

Regression Metrics
```

---

# Classification

Target:

```text
Fraud

Spam

Disease

Churn
```

---

Metrics:

```text
Accuracy

Precision

Recall

F1 Score

ROC-AUC
```

---

# Regression

Target:

```text
House Price

Revenue

Sales
```

---

Metrics:

```text
MAE

MSE

RMSE

R²
```

---

## Interview Takeaway

Different prediction problems require different evaluation metrics.

---

# Confusion Matrix

The foundation of classification evaluation.

---

# What Is A Confusion Matrix?

A table showing:

```text
Actual Values

Vs

Predicted Values
```

---

| | Predicted Positive | Predicted Negative |
|----------|----------|----------|
| Actual Positive | TP | FN |
| Actual Negative | FP | TN |

---

Everything in classification metrics comes from this table.

---

## Interview Takeaway

The confusion matrix is the foundation of classification evaluation.

---

# True Positive (TP)

## Definition

Model predicts:

```text
Positive
```

and prediction is correct.

---

## Example

Fraud Transaction

↓

Predicted Fraud

---

Correct.

---

## Interview Takeaway

TP means correctly identified positives.

---

# True Negative (TN)

## Definition

Model predicts:

```text
Negative
```

and prediction is correct.

---

## Example

Legitimate Transaction

↓

Predicted Legitimate

---

Correct.

---

## Interview Takeaway

TN means correctly identified negatives.

---

# False Positive (FP)

## Definition

Model predicts:

```text
Positive
```

but prediction is wrong.

---

## Example

Legitimate Transaction

↓

Predicted Fraud

---

Wrong.

---

## Business Impact

Customer may be blocked unnecessarily.

---

## Interview Takeaway

FP means false alarm.

---

# False Negative (FN)

## Definition

Model predicts:

```text
Negative
```

but prediction is wrong.

---

## Example

Fraud Transaction

↓

Predicted Legitimate

---

Wrong.

---

## Business Impact

Fraud passes through the system.

---

## Interview Takeaway

FN means missed detection.

---

# Accuracy

## Definition

Percentage of correct predictions.

---

## Formula

```text
(TP + TN)

/

Total Predictions
```

---

## Example

100 Predictions

```text
90 Correct
```

---

Accuracy:

```text
90%
```

---

## Limitation

Can be misleading on imbalanced datasets.

---

## Interview Takeaway

Accuracy measures overall correctness.

---

# Precision

## Definition

Among all predicted positives,

how many were actually positive?

---

## Formula

```text
TP

/

(TP + FP)
```

---

## Layman Explanation

Precision answers:

> "When the model says YES, how often is it correct?"

---

## Example

Fraud Detection

Predicted Fraud:

```text
100 Transactions
```

---

Actually Fraud:

```text
90 Transactions
```

---

Precision:

```text
90%
```

---

## Interview Takeaway

Precision measures prediction quality for positive predictions.

---

# Recall

## Definition

Among all actual positives,

how many were found?

---

## Formula

```text
TP

/

(TP + FN)
```

---

## Layman Explanation

Recall answers:

> "How many real positives did we catch?"

---

## Example

100 Fraud Cases Exist.

---

Model Detects:

```text
90
```

---

Recall:

```text
90%
```

---

## Interview Takeaway

Recall measures detection ability.

---

# Precision vs Recall

One of the most common interview questions.

---

# High Precision

Few false alarms.

---

Useful when:

```text
False Positives

Are Expensive
```

---

Example:

```text
Spam Filtering
```

---

# High Recall

Few missed cases.

---

Useful when:

```text
False Negatives

Are Expensive
```

---

Example:

```text
Cancer Detection

Fraud Detection
```

---

## Interview Takeaway

Precision minimizes false positives.

Recall minimizes false negatives.

---

# F1 Score

## Definition

Harmonic mean of Precision and Recall.

---

## Formula

```text
2 ×

(Precision × Recall)

/

(Precision + Recall)
```

---

## Why?

Accuracy ignores imbalance.

---

Precision ignores missed cases.

---

Recall ignores false alarms.

---

F1 balances both.

---

## Interview Takeaway

F1 Score balances Precision and Recall.

---

# Specificity

## Definition

Measures how well negatives are identified.

---

## Formula

```text
TN

/

(TN + FP)
```

---

## Example

Healthy Patients

↓

Correctly Identified

---

## Interview Takeaway

Specificity measures negative-class performance.

---

# Probability Threshold

Most classifiers output:

```text
Probabilities
```

---

Example

```text
0.92

0.75

0.20
```

---

To make predictions:

```text
Threshold
```

is applied.

---

Usually:

```text
0.5
```

---

Changing threshold changes:

```text
Precision

Recall
```

---

## Interview Takeaway

Classification performance depends on threshold selection.

---

# ROC Curve

One of the most asked interview topics.

---

# What Is ROC?

ROC means:

```text
Receiver Operating Characteristic
```

---

ROC shows model performance across:

```text
Different Thresholds
```

---

Instead of evaluating:

```text
One Threshold
```

---

Evaluate:

```text
All Thresholds
```

---

## Interview Takeaway

ROC evaluates classification performance across thresholds.

---

# ROC Axes

X-axis

```text
False Positive Rate
```

---

Y-axis

```text
True Positive Rate

(Recall)
```

---

Better models produce curves closer to:

```text
Top Left Corner
```

---

## Interview Takeaway

Better ROC curves move toward the top-left corner.

---

# AUC (Area Under Curve)

## Definition

Area under ROC Curve.

---

Range:

```text
0 → 1
```

---

# Interpretation

```text
0.5
```

Random Guessing

---

```text
1.0
```

Perfect Classifier

---

## Interview Takeaway

Higher AUC indicates better class separation.

---

# Precision-Recall Curve

Important for imbalanced datasets.

---

Instead of:

```text
ROC
```

focus on:

```text
Precision

Recall
```

---

Especially useful when:

```text
Positive Class Is Rare
```

---

Examples:

```text
Fraud

Disease

Cybersecurity
```

---

## Interview Takeaway

PR Curve is preferred for highly imbalanced datasets.

---

# Regression Metrics

Now move to regression.

---

# MAE (Mean Absolute Error)

## Formula

```text
Average

|Actual - Predicted|
```

---

## Example

Error:

```text
₹10,000

₹5,000

₹15,000
```

---

Average:

```text
₹10,000
```

---

## Interpretation

Average prediction mistake.

---

## Interview Takeaway

MAE measures average absolute prediction error.

---

# MSE (Mean Squared Error)

## Formula

```text
Average

(Actual - Predicted)²
```

---

## Why Square Errors?

Large errors receive:

```text
Large Penalties
```

---

## Interview Takeaway

MSE penalizes large mistakes heavily.

---

# RMSE (Root Mean Squared Error)

## Formula

```text
√MSE
```

---

## Why?

Returns error to:

```text
Original Units
```

---

Example

House Prices

↓

RMSE In Currency

---

## Interview Takeaway

RMSE is easier to interpret than MSE.

---

# MAE vs RMSE

| Topic | MAE | RMSE |
|----------|----------|----------|
| Outlier Sensitivity | Lower | Higher |
| Interpretation | Easier | Good |
| Large Error Penalty | Lower | Higher |

---

## Interview Takeaway

RMSE punishes large errors more aggressively.

---

# R² Score

One of the most common regression interview topics.

---

## Definition

Measures how much variance in the target is explained by the model.

---

## Range

```text
0 → 1
```

usually.

---

# Example

```text
R² = 0.90
```

---

Interpretation:

```text
90%

Of Variability

Explained By Model
```

---

## Special Case

```text
R² < 0
```

means model performs worse than predicting the mean.

---

## Interview Takeaway

R² measures explanatory power.

---

# Cross Validation

One of the most important model evaluation concepts.

---

# Problem

Single train-test split may be unlucky.

---

# Solution

Split data multiple times.

---

Train:

```text
Many Times
```

---

Evaluate:

```text
Average Performance
```

---

## K-Fold Cross Validation

Most common approach.

---

Example

```text
K = 5
```

---

Dataset split into:

```text
5 Parts
```

---

Train:

```text
4 Parts
```

---

Test:

```text
1 Part
```

---

Repeat:

```text
5 Times
```

---

Average results.

---

## Interview Takeaway

Cross Validation provides more reliable performance estimates.

---

# Stratified Sampling

Important for classification.

---

# Problem

Imbalanced classes.

---

Example

```text
99% Non-Fraud

1% Fraud
```

---

Random split may distort proportions.

---

# Solution

Maintain class distribution.

---

Result:

```text
Train And Test

Have Similar Class Ratios
```

---

## Interview Takeaway

Stratified sampling preserves class distribution.

---

# Hyperparameter Tuning

Evaluation is also used for:

```text
Model Selection
```

---

Example

```text
Max Depth

Learning Rate

Number Of Trees
```

---

Try multiple combinations.

---

Choose best performer.

---

## Interview Takeaway

Evaluation metrics guide hyperparameter selection.
