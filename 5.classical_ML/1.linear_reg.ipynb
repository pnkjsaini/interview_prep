# Module 5: Classical Machine Learning

# Chapter 1: Linear Regression

Linear Regression is usually the first machine learning algorithm people learn.

However, its importance goes far beyond being a beginner algorithm.

Linear Regression introduces many concepts that appear repeatedly in Machine Learning:

```text
Prediction

Model Training

Loss Functions

Optimization

Coefficients

Feature Importance
```

Understanding Linear Regression properly makes Logistic Regression, Neural Networks, and many other algorithms much easier to understand.

---

# Why Linear Regression Exists

Suppose a business asks:

> "Can we predict house prices from house characteristics?"

Available data:

```text
Area

Bedrooms

Bathrooms

Location Score
```

Target:

```text
House Price
```

---

We need a model that can learn:

```text
Input Features
        ↓
Target Variable
```

---

This problem is called:

```text
Regression
```

because the target is continuous.

---

## Examples

House Price Prediction

```text
₹25 Lakh

₹40 Lakh

₹90 Lakh
```

---

Sales Forecasting

```text
Future Revenue
```

---

Demand Prediction

```text
Expected Orders
```

---

## Interview Takeaway

Linear Regression predicts continuous numerical values.

---

# Core Idea

Linear Regression assumes:

```text
Input Features

And

Target Variable
```

have approximately a linear relationship.

---

## Example

Area increases:

```text
1000 Sq Ft

↓

1500 Sq Ft

↓

2000 Sq Ft
```

---

House price may also increase.

---

The model tries to discover:

```text
How Strongly Area Influences Price
```

---

## Interview Takeaway

Linear Regression models relationships using a straight-line assumption.

---

# Linear Relationship

## Definition

A linear relationship means changes in input create proportional changes in output.

---

## Example

Suppose:

```text
Every Additional Year Of Experience

Adds ₹1 Lakh Salary
```

---

Experience:

```text
1 Year → ₹5 Lakh

2 Years → ₹6 Lakh

3 Years → ₹7 Lakh
```

---

This follows a linear pattern.

---

## Limitation

Many real-world relationships are not perfectly linear.

---

## Interview Takeaway

Linear Regression works best when relationships are approximately linear.

---

# Equation Of Linear Regression

## Formula

For one feature:

```text
y = mx + c
```

---

Machine Learning Form:

```text
y = β₀ + β₁x
```

---

Where:

```text
y
```

Prediction

---

```text
β₀
```

Intercept

---

```text
β₁
```

Coefficient

---

```text
x
```

Feature

---

## Multiple Features

```text
y = β₀

+ β₁x₁

+ β₂x₂

+ β₃x₃
```

---

## Interview Takeaway

Linear Regression predicts outcomes using weighted feature contributions.

---

# Intercept

## Definition

The intercept is the prediction when all features equal zero.

---

## Example

Salary Prediction

```text
Salary

=

300000

+

100000 × Experience
```

---

Intercept:

```text
300000
```

---

Meaning:

```text
Predicted Salary

When Experience = 0
```

---

## Limitation

May not always have real-world meaning.

---

## Interview Takeaway

The intercept is the model's baseline prediction.

---

# Coefficients

## Definition

Coefficients measure feature influence on predictions.

---

## Layman Explanation

A coefficient answers:

> "How much does the prediction change if this feature changes?"

---

## Example

```text
Price

=

10

+

5 × Area
```

---

Coefficient:

```text
5
```

---

Meaning:

```text
Every Unit Increase In Area

Adds 5 Units To Prediction
```

---

## Why It Matters

Coefficients help interpret feature importance.

---

## Interview Takeaway

Coefficients represent feature impact on the target.

---

# Prediction

Once coefficients are learned:

```text
New Input
      ↓
Model
      ↓
Prediction
```

---

Example

```text
Salary

=

300000

+

100000 × Experience
```

---

Experience:

```text
5 Years
```

---

Prediction:

```text
₹8 Lakh
```

---

## Interview Takeaway

Predictions are generated using learned coefficients.

---

# Residual

## Definition

Residual is the difference between actual and predicted values.

---

## Formula

```text
Residual

=

Actual - Predicted
```

---

## Layman Explanation

Residual measures:

```text
Prediction Error
```

---

## Example

Actual Salary:

```text
₹12 Lakh
```

---

Predicted Salary:

```text
₹10 Lakh
```

---

Residual:

```text
₹2 Lakh
```

---

## Why It Matters

Model training attempts to reduce residuals.

---

## Interview Takeaway

Residuals measure prediction mistakes.

---

# Cost Function

One prediction is not enough.

We need a way to evaluate:

```text
Entire Dataset
```

---

This is the role of:

```text
Cost Function
```

---

# Mean Squared Error (MSE)

## Formula

```text
MSE

=

Σ(Actual - Predicted)²

/

n
```

---

## Layman Explanation

MSE averages squared prediction errors.

---

Large mistakes receive:

```text
Large Penalties
```

---

## Why It Matters

Linear Regression usually minimizes:

```text
MSE
```

---

## Interview Takeaway

MSE measures average squared prediction error.

---

# How Linear Regression Learns

The goal is:

```text
Find Best Coefficients
```

---

Initially:

```text
Random Coefficients
```

---

Predictions:

```text
Poor
```

---

Calculate:

```text
Loss
```

---

Use:

```text
Gradient Descent
```

to improve coefficients.

---

Repeat:

```text
Prediction

↓

Error

↓

Update Coefficients

↓

Better Prediction
```

---

## Interview Takeaway

Linear Regression learns by minimizing prediction error.

---

# Assumptions Of Linear Regression

One of the most common interview topics.

---

# Assumption 1

Linear Relationship

---

Features should have approximately linear relationships with the target.

---

# Assumption 2

Independence Of Errors

---

Residuals should not influence each other.

---

# Assumption 3

Constant Variance (Homoscedasticity)

---

Error variance should remain stable.

---

# Assumption 4

Normal Distribution Of Errors

---

Residuals should be approximately normal.

---

# Assumption 5

No Multicollinearity

---

Features should not contain excessive redundancy.

---

## Interview Takeaway

Violating assumptions can reduce model reliability.

---

# Advantages

## Easy To Understand

Highly interpretable.

---

## Fast Training

Computationally efficient.

---

## Explainable

Coefficients provide clear insights.

---

## Strong Baseline

Often serves as the first model.

---

# Limitations

## Assumes Linearity

May fail on complex patterns.

---

## Sensitive To Outliers

Large outliers can distort coefficients.

---

## Multicollinearity Issues

Highly correlated features create instability.

---

## Underfitting

May struggle with nonlinear relationships.

---

# Common Interview Questions

---

## Question

Why is Linear Regression called linear?

### Answer

```text
Because predictions are a linear combination of features and coefficients.
```

---

## Question

What does a coefficient represent?

### Answer

```text
The expected change in prediction for a unit change in a feature.
```

---

## Question

What is a residual?

### Answer

```text
Actual Value - Predicted Value.
```

---

## Question

Why does Linear Regression use MSE?

### Answer

```text
Because MSE penalizes large errors and is easy to optimize.
```

---

## Question

What happens if assumptions are violated?

### Answer

```text
Predictions may still work, but coefficient interpretation and statistical reliability can suffer.
```

---

## Question

How does Linear Regression learn?

### Answer

```text
By minimizing prediction error using optimization techniques.
```
