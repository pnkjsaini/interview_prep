---

```text
Data

Business Requirements

Interpretability

Scale

Latency
```

---

# The Model Selection Mindset

Instead of asking:

```text
Which Algorithm Is Best?
```

Ask:

```text
What Problem Am I Solving?
```

---

Because every algorithm makes:

```text
Different Assumptions

Different Tradeoffs
```

---

## Interview Takeaway

Choose algorithms based on problem characteristics, not popularity.

# Step 3: Linear Or Nonlinear Relationship

---

# Mostly Linear

Examples:

```text
Salary Prediction

Simple Forecasting
```

---

Algorithms:

```text
Linear Regression

Logistic Regression
```

---

# Nonlinear Relationships

# Step 4: Interpretability Requirements

A very common business constraint.

---

# High Interpretability Needed

Examples:

```text
Banking

Healthcare

Insurance

Government
```

---

Preferred Models:

```text
Linear Regression

Logistic Regression

Decision Trees
```

---

Reason:

```text
Easy To Explain
```

---

# Accuracy More Important

Examples:

---

Preferred Models:

```text
XGBoost

Random Forest

CatBoost
```

---

## Interview Takeaway

Interpretability often competes with predictive power.

---

---

# Step 7: Training Speed

Sometimes speed matters more than accuracy.

---

# Fast Training Needed

Use:

```text
Linear Models

Naive Bayes
```

---

# Moderate Speed

Use:

```text
Decision Trees

Random Forest
```

---

## Interview Takeaway

Training constraints impact model choice.

---

# Step 8: Prediction Latency

Training occurs once.

Predictions may occur millions of times.

---

# Low Latency Required

Examples:

```text
Real-Time Fraud Detection

Real-Time Recommendations
```

---

Prefer:

```text
Logistic Regression

Trees

Boosting
```

---

# High Latency Acceptable

Can consider:

```text
KNN
```

---

## Interview Takeaway

Prediction speed is often more important than training speed.

---

# Understanding Bias And Variance

Model selection often revolves around:

```text
Bias

Variance
```

---

# High Bias

Model too simple.

---

Result:

```text
Underfitting
```

---

Examples:

```text
Linear Regression

Logistic Regression
```

on complex data.

---

# High Variance

Model too flexible.

---

Result:

```text
Overfitting
```

---

Examples:

```text
Deep Decision Trees
```

---

## Interview Takeaway

Choosing a model often means balancing bias and variance.

---

# Algorithm Selection Guide

---

# Linear Regression

Choose when:

```text
Continuous Target

Linear Relationship

Interpretability Required
```

---

Avoid when:

```text
Complex Nonlinear Patterns
```

---

# Logistic Regression

Choose when:

```text
Binary Classification

Interpretability Important

Baseline Model Needed
```

---

Avoid when:

```text
Highly Nonlinear Boundaries
```

---

# Decision Tree

Choose when:

```text
Interpretability Needed

Nonlinear Relationships
```

---

Avoid when:

```text
High Stability Required
```

---

# Random Forest

Choose when:

```text
Strong Baseline Needed

Minimal Tuning

Nonlinear Data
```

---

Avoid when:

```text
Extreme Prediction Speed Needed
```

---

# XGBoost

Choose when:

```text
Maximum Accuracy

Structured Data

Competition Settings
```

---

Avoid when:

```text
Interpretability Is Critical
```

---

# LightGBM

Choose when:

```text
Large Dataset

Fast Training Required
```

---

Avoid when:

```text
Very Small Dataset
```

---

# CatBoost

Choose when:

```text
Many Categorical Features
```

---

Avoid when:

```text
Dataset Is Purely Numerical
```

---

# SVM

Choose when:

```text
Small Dataset

High-Dimensional Data
```

---

Examples:

```text
Text Classification
```

---

Avoid when:

```text
Very Large Dataset
```

---

# KNN

Choose when:

```text
Small Dataset

Simple Baseline Needed
```

---

Avoid when:

```text
Large Dataset

High Dimensions
```

---

# Naive Bayes

Choose when:

```text
Text Classification

Fast Training

Small Dataset
```

---

Avoid when:

```text
Complex Feature Interactions
```

---

# Clustering Selection Guide

---

# K-Means

Choose when:

```text
Clusters Are Roughly Spherical

Large Dataset
```

---

# Hierarchical Clustering

Choose when:

```text
Need Dendrogram

Exploratory Analysis
```

---

# DBSCAN

Choose when:

```text
Unknown Cluster Count

Outlier Detection

Irregular Shapes
```

# Common Interview Questions

---

## Question

Which algorithm is best?

### Answer

```text
There is no universally best algorithm.
Model choice depends on data and business requirements.
```

---

## Question

Why start with Logistic Regression before XGBoost?

### Answer

```text
Simple baselines provide interpretability and establish performance benchmarks.
```

---

## Question

When should Random Forest be preferred?

### Answer

```text
When strong performance is needed with minimal tuning.
```

---

## Question

When should XGBoost be preferred?

### Answer

```text
When maximizing predictive accuracy on structured data.
```

---

## Question

When should CatBoost be preferred?

### Answer

```text
When datasets contain many categorical features.
```

---

## Question

Why is interpretability important?

### Answer

```text
Many industries require explanation of predictions.
```

---

## Question

Why is KNN rarely used in production?

### Answer

```text
Prediction becomes expensive on large datasets.
```
