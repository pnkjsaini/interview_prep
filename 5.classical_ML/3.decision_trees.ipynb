# Module 5: Classical Machine Learning

# Chapter 3: Decision Trees

In the previous chapter we learned:

```text
Logistic Regression
```

which creates:

```text
Linear Decision Boundaries
```

This works well when classes can be separated using a straight line (or hyperplane).

However, many real-world datasets are not linearly separable.

Example:

```text
Loan Approval

Fraud Detection

Customer Churn

Medical Diagnosis
```

often involve complex decision rules.

We need a model that can learn:

```text
Nonlinear Relationships
```

without requiring complicated feature engineering.

This leads to:

```text
Decision Trees
```

---

# Why Decision Trees Exist

Imagine a bank deciding whether to approve a loan.

Instead of using a mathematical equation, a human might think:

```text
Income > ₹10 Lakh?
        ↓

Yes → Check Credit Score

No → Reject
```

---

Then:

```text
Credit Score > 750?
        ↓

Yes → Approve

No → Reject
```

---

This reasoning naturally forms a:

```text
Decision Tree
```

---

## Interview Takeaway

Decision Trees mimic human decision-making using a sequence of rules.

---

# Core Idea

A Decision Tree repeatedly asks questions that split the data into smaller and purer groups.

---

Example:

```text
Income > ₹10 Lakh?
```

---

Then:

```text
Credit Score > 750?
```

---

Then:

```text
Existing Loans < 2?
```

---

Each split attempts to create groups that contain mostly one class.

---

## Interview Takeaway

Decision Trees learn by repeatedly splitting data into purer groups.

---

# Tree Structure

A Decision Tree consists of:

```text
Root Node

Internal Nodes

Leaf Nodes
```

---

# Root Node

## Definition

The first split in the tree.

---

## Example

```text
Income > ₹10 Lakh?
```

---

## Why It Matters

Usually contains the most informative split.

---

## Interview Takeaway

The root node is the first and most important split.

---

# Internal Node

## Definition

A decision point inside the tree.

---

## Example

```text
Credit Score > 750?
```

---

## Interview Takeaway

Internal nodes continue partitioning the data.

---

# Leaf Node

## Definition

The final prediction node.

---

## Example

```text
Approve Loan

Reject Loan
```

---

## Interview Takeaway

Leaf nodes contain final predictions.

---

# How A Decision Tree Learns

A critical question:

> "Which feature should be used for splitting?"

---

Possible Features:

```text
Income

Age

Credit Score

Debt
```

---

We need a way to measure:

```text
How Good A Split Is
```

---

This is where:

```text
Entropy

Information Gain

Gini Impurity
```

become important.

---

# Entropy

## Definition

Entropy measures disorder or uncertainty.

---

## Layman Explanation

Entropy answers:

> "How mixed are the classes?"

---

## Example

Group A

```text
10 Approved

0 Rejected
```

Entropy:

```text
Low
```

---

Group B

```text
5 Approved

5 Rejected
```

Entropy:

```text
High
```

---

## Interpretation

Low Entropy

```text
Pure Group
```

---

High Entropy

```text
Mixed Group
```

---

## Interview Takeaway

Entropy measures class impurity.

---

# Information Gain

## Definition

Information Gain measures how much uncertainty is reduced after a split.

---

## Layman Explanation

A good split should:

```text
Reduce Entropy
```

---

Information Gain answers:

> "How much cleaner did the data become?"

---

## Example

Before Split

```text
50 Approved

50 Rejected
```

---

After Split

Group 1

```text
45 Approved

5 Rejected
```

---

Group 2

```text
5 Approved

45 Rejected
```

---

Huge reduction in uncertainty.

---

High Information Gain.

---

## Why It Matters

Traditional Decision Trees choose splits with:

```text
Maximum Information Gain
```

---

## Interview Takeaway

Information Gain measures improvement produced by a split.

---

# Gini Impurity

## Definition

Gini Impurity measures how often a randomly selected observation would be misclassified.

---

## Layman Explanation

Gini answers:

> "How mixed is this node?"

similar to entropy.

---

## Example

Node:

```text
100 Approved

0 Rejected
```

Gini:

```text
0
```

Perfectly pure.

---

Node:

```text
50 Approved

50 Rejected
```

High Gini.

---

## Why It Matters

Most modern Decision Tree implementations use:

```text
Gini Impurity
```

instead of entropy.

---

## Why?

Faster computation.

---

## Interview Takeaway

Gini Impurity is an alternative impurity measure commonly used in Decision Trees.

---

# Entropy vs Gini

| Topic | Entropy | Gini |
|----------|----------|----------|
| Measures Impurity | Yes | Yes |
| Uses Logarithms | Yes | No |
| Computational Cost | Higher | Lower |
| Most Common In Practice | Less | More |

---

## Interview Takeaway

Both measure impurity, but Gini is generally faster.

---

# Splitting Criteria

The tree evaluates many possible splits.

---

Example

```text
Income > ₹5 Lakh

Income > ₹10 Lakh

Income > ₹15 Lakh
```

---

Each split receives an impurity score.

---

Best split:

```text
Maximum Information Gain

Or

Minimum Gini Impurity
```

---

## Interview Takeaway

Decision Trees choose the split that produces the purest child nodes.

---

# Recursive Splitting

After the first split:

```text
Split Again
```

---

Then:

```text
Split Again
```

---

Then:

```text
Split Again
```

---

This process continues recursively.

---

## Interview Takeaway

Decision Trees grow by recursively partitioning data.

---

# Tree Depth

## Definition

Tree depth is the number of levels from root to leaf.

---

## Example

```text
Root

↓

Level 1

↓

Level 2

↓

Leaf
```

---

Depth:

```text
3
```

---

## Why It Matters

Depth controls model complexity.

---

## Interview Takeaway

Deeper trees are more flexible but more prone to overfitting.

---

# Overfitting In Decision Trees

One of the most important interview topics.

---

Suppose the tree keeps splitting until:

```text
Every Training Record
```

is perfectly classified.

---

Training Accuracy:

```text
100%
```

---

Test Accuracy:

```text
Poor
```

---

The tree memorized noise.

---

## Interview Takeaway

Deep trees often overfit training data.

---

# Pruning

## Definition

Pruning removes unnecessary branches from a tree.

---

## Layman Explanation

If a branch contributes little predictive value:

```text
Remove It
```

---

Result:

```text
Simpler Tree

Better Generalization
```

---

## Types

### Pre-Pruning

Stop growth early.

---

### Post-Pruning

Grow tree first.

Then remove weak branches.

---

## Interview Takeaway

Pruning reduces overfitting.

---

# Decision Tree For Regression

Decision Trees are not limited to classification.

---

Example:

```text
House Price Prediction
```

---

Instead of predicting classes:

```text
Predict Average Value
```

inside each leaf.

---

## Interview Takeaway

Decision Trees support both classification and regression.

---

# Feature Importance

## Definition

Feature Importance measures how much a feature contributes to tree decisions.

---

## Example

Features:

```text
Income

Credit Score

Age
```

---

Tree may use:

```text
Income
```

frequently.

---

Result:

```text
High Feature Importance
```

---

## Why It Matters

Provides interpretability.

---

## Interview Takeaway

Feature importance reflects a feature's contribution to prediction.

---

# Advantages

## Easy To Explain

Human-readable decisions.

---

## Handles Nonlinear Relationships

No linear assumption required.

---

## Works With Numerical And Categorical Data

Very flexible.

---

## Little Feature Scaling Needed

Unlike many algorithms.

---

# Limitations

## Overfitting

Can memorize training data.

---

## High Variance

Small data changes can create different trees.

---

## Unstable

Sensitive to training samples.

---

## Poor Extrapolation

Cannot predict beyond observed patterns.

---

# Decision Tree Workflow

```text
Dataset
      ↓
Evaluate Splits
      ↓
Choose Best Split
      ↓
Create Child Nodes
      ↓
Repeat Recursively
      ↓
Leaf Nodes
      ↓
Prediction
```

# Common Interview Questions

---

## Question

Why do Decision Trees exist?

### Answer

```text
To learn nonlinear decision boundaries using rule-based splitting.
```

---

## Question

What is Entropy?

### Answer

```text
A measure of uncertainty or impurity.
```

---

## Question

What is Information Gain?

### Answer

```text
The reduction in entropy achieved by a split.
```

---

## Question

What is Gini Impurity?

### Answer

```text
A measure of class impurity commonly used in Decision Trees.
```

---

## Question

Why do Decision Trees overfit?

### Answer

```text
Because they can keep splitting until they memorize training data.
```

---

## Question

What is Pruning?

### Answer

```text
Removing unnecessary branches to improve generalization.
```

---

## Question

Can Decision Trees perform regression?

### Answer

```text
Yes, by predicting numerical values within leaf nodes.
```

# Decision Tree vs Logistic Regression

| Topic | Logistic Regression | Decision Tree |
|----------|----------|----------|
| Decision Boundary | Linear | Nonlinear |
| Interpretability | High | High |
| Feature Scaling | Often Helpful | Usually Not Needed |
| Overfitting Risk | Lower | Higher |
| Complex Relationships | Weak | Strong |
